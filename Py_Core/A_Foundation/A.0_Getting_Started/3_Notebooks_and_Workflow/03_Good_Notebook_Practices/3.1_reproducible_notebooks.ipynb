# Reproducible Notebooks

A notebook is a scientific document. It must be reproducible.

## 🎯 Learning Objectives
- Linear Execution.
- Handling External Data Paths.

## 1. Linear Execution

Never write code that requires the user to scroll UP to run a cell again.

**Bad:**
- Cell 1: `x = 10`
- Cell 2: `print(x)`
- Cell 3: `x = 20` (User must scroll up and run Cell 2 again to see 20? No!)

**Good:**
The flow is strictly top to bottom.

## 2. Paths

Avoid hardcoding paths like `C:\Users\Mike\Desktop\data.csv`.

Use relative paths or libraries to download data automatically if missing.
