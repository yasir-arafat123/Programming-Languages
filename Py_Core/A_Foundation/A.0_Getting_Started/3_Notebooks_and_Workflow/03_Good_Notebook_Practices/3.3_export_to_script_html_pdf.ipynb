# Exporting Notebooks

Don't send `.ipynb` files to your boss. Send a PDF or HTML report.

## 🎯 Learning Objectives
- `nbconvert`.
- Export as Python Script.

## 1. Export as HTML

File -> Save and Export Notebook As -> HTML.

Or CLI:
```bash
jupyter nbconvert --to html notebook.ipynb
```
This creates a standalone web page with your analysis.

## 2. Export as Script

If you want to run your notebook in production, convert it to a `.py` file.

```bash
jupyter nbconvert --to script notebook.ipynb
```
