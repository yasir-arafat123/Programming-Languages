# Clear Outputs & Version Control

Notebooks with large plots or dataframes can be huge (MBs). Git hates huge text files with binary blobs.

## 🎯 Learning Objectives
- Clearing Output.
- Git Hooks (nbstripout).

## 1. Clear Output

Before committing to GitHub, go to `Kernel -> Restart & Clear Output`.

This removes all images and tables, reducing the file size to just the code (KBs).

## 2. Nbstripout

A pro tool that does this automatically when you `git commit`.

```bash
pip install nbstripout
nbstripout --install
```

Now Git will never see your outputs, only code.
