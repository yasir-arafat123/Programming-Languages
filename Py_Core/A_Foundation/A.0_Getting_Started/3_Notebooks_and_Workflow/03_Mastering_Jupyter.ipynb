## 1. The Two Modes

1.  **Command Mode (Press `Esc`)**: The border turns **BLUE**. You operate on *cells* (move, delete, create).
2.  **Edit Mode (Press `Enter`)**: The border turns **GREEN**. You type *code* inside a cell.

### ⚡ Vital Shortcuts (Memorize These)

**Command Mode (Blue):**
- `A`: Insert cell **A**bove
- `B`: Insert cell **B**elow
- `D`, `D`: **D**elete current cell
- `M`: Convert to **M**arkdown
- `Y`: Convert to Code

**Edit Mode (Green):**
- `Shift + Enter`: Run and move to next
- `Ctrl + Enter`: Run and stay

---
## 2. Magic Commands (`%` and `%%`)

Magics are special commands provided by the IPython kernel.
- `%line_magic`: Affects one line.
- `%%cell_magic`: Affects the entire cell.

### ⏱️ Profiling Time: `%timeit`
The most useful magic. It runs code loops thousands of times to get an accurate execution time benchmark.

```python
# Which is faster: list comprehension or loop?

print("List Comprehension:")
%timeit [x**2 for x in range(1000)]

print("\nFor Loop:")
%%timeit
L = []
for x in range(1000):
    L.append(x**2)
```

### 🛠️ System Commands: `!`
You can run any terminal command using `!`.

```python
!echo "Hello from the Operating System!"
!python --version
```

### 📂 File Operations: `%%writefile`
We used this in the previous notebook. It writes the cell contents to a file.

## 3. The Kernel & State

The "Kernel" is the Python process running in the background. It remembers all variables you've defined.

**Problem**: Sometimes you delete code, but the variable stays in memory. This causes bugs.
**Solution**: `Kernel -> Restart & Run All`.

ALWAYS do this before sharing a notebook to ensure it runs top-to-bottom.
