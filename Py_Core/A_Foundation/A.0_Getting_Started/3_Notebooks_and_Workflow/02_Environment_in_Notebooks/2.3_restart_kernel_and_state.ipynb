# Restart Kernel & State

Notebooks have a memory. It remembers variables even if you delete the cell that created them. This is dangerous.

## 🎯 Learning Objectives
- The Hidden State Problem.
- Restart & Clear Output.

```python
x = 100
```

Imagine you delete the cell above. `x` is still 100 in memory.

```python
print(x)  # This works!
```

Now imagine you send this to a colleague. They open it, try to run `print(x)`, and it FAILS because they never ran the deleted cell.

## The Golden Rule

Before finishing your work, **ALWAYS**:
1.  Kernel -> Restart & Clear Output.
2.  Kernel -> Restart & Run All.

If it crashes, your notebook is broken.
