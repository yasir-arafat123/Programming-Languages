# Selecting the Kernel

A "Kernel" is the specific Python installation running your code. You can have multiple kernels (Python 3.9, Python 3.12, Conda Env, etc.).

## 🎯 Learning Objectives
- Understanding Kernels.
- Installing a Virtual Environment as a Kernel.

## 1. The Kernel Spec

Jupyter doesn't just "run Python". It looks for registered kernels.

To see available kernels in CLI:
```bash
jupyter kernelspec list
```

## 2. Adding your Virtual Env to Jupyter

This is a common interview question/practical problem.

1.  Activate your venv: `source .venv/bin/activate`
2.  Install ipykernel: `pip install ipykernel`
3.  Register it:
```bash
python -m ipykernel install --user --name=my_project_env --display-name "Python (My Project)"
```
4.  Refresh Jupyter web page.
5.  Click "Kernel" -> "Change Kernel" -> Select "Python (My Project)".
