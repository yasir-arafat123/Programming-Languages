# Installing Packages in Notebooks

There is a right way and a wrong way. The wrong way breaks your environment.

## 🎯 Learning Objectives
- `%pip` vs `!pip`.
- Handling restarts.

## 1. The Wrong Way (`!pip`)

```python
!pip install pandas
```
**Why?** `!` executes in the underlying shell. If your Jupyter Kernel is running Python A, but the global shell `pip` maps to Python B, you install the package in the WRONG place.

## 2. The Right Way (`%pip`)

```python
%pip install pandas
```
**Why?** Magic commands like `%pip` are aware of the CURRENT running kernel. It guarantees installation into the correct environment.

````python
# Try it (it might say Requirement already satisfied)
%pip install requests
````

## 3. The Restart Rule

After installing a package, you often need to **Restart the Kernel** to use it, because Python only scans for packages at startup or when `import` is called (but cached imports can be tricky).
