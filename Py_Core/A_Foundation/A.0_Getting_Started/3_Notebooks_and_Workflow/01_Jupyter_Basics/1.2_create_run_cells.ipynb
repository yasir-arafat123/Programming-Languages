## 1. The Modes

- **Edit (Green)**: Enter text. Press `Enter` to access.
- **Command (Blue)**: Modify structure. Press `Esc` to access.

## 2. Essential Shortcuts

| Action | Shortcut |
| :--- | :--- |
| Run + New | `Shift + Enter` |
| Run + Stay | `Ctrl + Enter` |
| New Above | `A` (Command Mode) |
| New Below | `B` (Command Mode) |
| Delete | `D, D` (Command Mode) |
| Markdown | `M` (Command Mode) |
| Code | `Y` (Command Mode) |
