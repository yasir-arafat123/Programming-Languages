# What is Jupyter?

Jupyter (JUlia, PYThon, and R) is an open-source web application that allows you to create and share documents that contain live code, equations, visualizations and narrative text.

## 🎯 Learning Objectives
- The Frontend vs The Kernel.
- The `.ipynb` file format (it's just text!).

## 1. Architecture

1.  **Frontend**: The web interface you are looking at. It sends code to the kernel.
2.  **Kernel**: The engine processing the code (IPython). It returns output to the frontend.

This separation means you can connect a single frontend to a remote supercomputer kernel.

## 2. The `.ipynb` format

Notebooks are actually ugly JSON files under the hood. They contain metadata about every cell's execution count and output.

```python
import json

# Let's read THIS file effectively (meta-programming!)
# Note: This only works if the file is saved on disk.
try:
    with open("1.1_what_is_jupyter.ipynb", "r", encoding="utf-8") as f:
        data = json.load(f)
        print("Notebook Keys:", list(data.keys()))
        print("First Cell content:", data["cells"][0]["source"][:3])
except FileNotFoundError:
    print("File not saved to disk yet!")
```
