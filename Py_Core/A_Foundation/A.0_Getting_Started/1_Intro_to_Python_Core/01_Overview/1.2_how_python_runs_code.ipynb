# How Python Runs Code

Understanding the execution model distinguishes a user from a developer. Python is not magic; it follows a strict compile-interpret process.

## 🎯 Learning Objectives
- Understand the **Compilation** step (Bytecode).
- Understand the **Execution** step (PVM).
- Learn about `.pyc` files and `__pycache__`.

## 1. The Execution Flow

When you run `python script.py`, two things happen:

1.  **Compilation**: The source code (`.py`) is translated into **Bytecode**.
    - Bytecode is a low-level, platform-independent representation.
    - It is NOT machine code (binary zeros and ones for the CPU).
    
2.  **Interpretation**: The **Python Virtual Machine (PVM)** loops through the bytecode instructions and executes them.

## 2. Inspecting Bytecode

We can peek at the bytecode using the `dis` (disassembler) module.

````python
import dis

def hello():
    print("Hello World")

print("--- Bytecode for 'hello' function ---")
dis.dis(hello)
````

## 3. __pycache__ folders

To speed up loading, Python saves the compiled bytecode to `.pyc` files inside a `__pycache__` directory. 

- If you run the script again and the source hasn't changed, Python skips the compilation step and loads the `.pyc` directly.
- This makes **startup** faster, but does not make the code **run** faster.
