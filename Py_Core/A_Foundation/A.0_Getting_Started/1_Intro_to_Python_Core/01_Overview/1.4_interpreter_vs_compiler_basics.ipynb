# Interpreter vs Compiler Basics

Is Python interpreted or compiled? The answer is... complicated.

## 🎯 Learning Objectives
- Differentiate between **Compilation** (C++) and **Interpretation** (Bash).
- Learn about Python's Hybrid approach.
- Explore alternative implementations like PyPy and Jython.

## 1. Pure Compilation (e.g., C++)
*Source Code* → *Machine Code (Binary)*
- **Pros**: Extremely fast execution.
- **Cons**: Must re-compile for every operating system (Windows exe, Linux ELF).

## 2. Pure Interpretation (e.g., Bash)
*Source Code* → *Interpreter reads line* → *Executes*
- **Pros**: Runs anywhere the interpreter exists.
- **Cons**: Very slow (re-evaluates loops every time).

## 3. Python's Hybrid (Bytecode)
*Source Code* → *Bytecode* → *Virtual Machine*
- **Pros**: Portable (Write Once, Run Anywhere) and reasonably fast.
- **Cons**: Slower than C++.

### JIT (Just-In-Time) Compilation
Some Python implementations, like **PyPy**, use a JIT compiler. It watches the running code, identifies "hot spots" (code run often), and compiles them to machine code *while the program is running*. This can make Python as fast as C in some cases.

## 4. Implementations

1.  **CPython**: The standard version (C). Stable, compatible.
2.  **PyPy**: The fast version (RPython + JIT). Focus on speed.
3.  **Jython**: Runs on Java Virtual Machine.
4.  **IronPython**: Runs on .NET Framework.
