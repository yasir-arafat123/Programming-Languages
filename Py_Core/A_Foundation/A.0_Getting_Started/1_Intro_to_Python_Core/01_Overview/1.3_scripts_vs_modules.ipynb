# Scripts vs Modules

Any `.py` file can be a script or a module. The difference lies in **how** it is used.

## 🎯 Learning Objectives
- Distinguish between standalone execution (Script) and importing (Module).
- Master the `if __name__ == "__main__":` block.

## 1. Definitions

- **Script**: A Python file designed to be executed directly from the command line (`python app.py`). It typically performs a task.
- **Module**: A Python file designed to be imported by other files (`import app`). It typically defines functions, classes, and variables for reuse.

## 2. The `__name__` Variable

When Python runs a file, it sets a special built-in variable called `__name__`.

- If run directly: `__name__` is set to `"__main__"`.
- If imported: `__name__` is set to the filename (e.g., `"math_utils"`).

````python
print(f"The name of this scope is: {__name__}")
````

## 3. The Best Practice Pattern

To make a file usable as **both** a script and a module, wrap the execution logic in the main block.

```python
def calculate(points):
    return len(points) * 10

if __name__ == "__main__":
    # This only runs if I execute 'python this_file.py'
    # It does NOT run if I do 'import this_file'
    print(calculate([1, 2, 3]))
```
