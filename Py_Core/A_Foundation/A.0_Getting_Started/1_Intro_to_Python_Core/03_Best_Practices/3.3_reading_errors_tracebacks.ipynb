# Reading Errors & Tracebacks

Don't panic! Red text is helpful. It tells you exactly what went wrong and where.

## 🎯 Learning Objectives

## 1. Anatomy of an Error

When Python crashes, it prints a "Stack Trace" or "Traceback".

**Read it from the BOTTOM up.**

1.  **Last line**: The actual error type and message.
2.  **Lines above**: The specific file and line number where it happened.

```python
# Let's generate an error on purpose
def division(a, b):
    return a / b

def calculate():
    division(10, 0)

try:
    calculate()
except Exception as e:
    # We print the error here just to show it safely in the notebook
    import traceback
    traceback.print_exc()
```

## 2. Common Errors

- **SyntaxError**: You typed valid English but invalid Python (missing `:`, unmatched `)`).
- **NameError**: You used a variable that hasn't been defined yet (`print(unknown_var)`).
- **TypeError**: You tried to combine incompatible types (`"Hello" + 5`).
- **IndexError**: You tried to access a list item that doesn't exist (`list[99]` when length is 3).
