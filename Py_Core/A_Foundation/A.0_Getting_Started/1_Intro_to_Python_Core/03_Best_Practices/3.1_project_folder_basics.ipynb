# Project Folder Structure

Professional Python projects don't just dump files on the desktop. They follow a predictable structure.

## 🎯 Learning Objectives
- Standard Directory Layout.
- Virtual Environments (`venv`).
- Ignoring files (`.gitignore`).

## 1. The Anatomy of a Project

A typical Python project looks like this:

```text
my_project/
│
├── .venv/               # Virtual Environment (Dependencies)
├── src/                 # Source Code
│   └── main.py
├── tests/               # Unit Tests
├── .gitignore           # Files to ignore in Git
├── README.md            # Documentation
└── requirements.txt     # Dependency list
```

## 2. Virtual Environments

**NEVER install packages globally.** Always use a virtual environment.

### Creating one
```bash
python -m venv .venv
```

### Activating one
- **Windows**: `.venv\Scripts\activate`
- **Mac/Linux**: `source .venv/bin/activate`

## 3. .gitignore

You should never commit your virtual environment or compiled python files to Git. A `.gitignore` file prevents this.

**Standard Python .gitignore:**
```text
__pycache__/
*.py[cod]
.venv/
.env
.ipynb_checkpoints/
```
