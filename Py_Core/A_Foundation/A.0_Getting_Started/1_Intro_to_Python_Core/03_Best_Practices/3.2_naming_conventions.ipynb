# Naming Conventions (PEP 8)

Code is read much more often than it is written. Following standard naming conventions ensures your code is readable by other Python developers.

## 1. The Rules

| Type | Convention | Example |
| :--- | :--- | :--- |
| **Variables** | `snake_case` | `user_name`, `item_count` |
| **Functions** | `snake_case` | `calculate_total()`, `get_data()` |
| **Classes** | `PascalCase` | `UserProfile`, `DataFrame` |
| **Constants** | `UPPER_SNAKE_CASE` | `MAX_RETRIES`, `DEFAULT_TIMEOUT` |
| **Modules** | `snake_case` | `math_utils.py` |

```python
# Bad Naming
x = 10
Val = 20
def ProcessData(D):
    return D * 2

# Good Naming
item_count = 10
item_value = 20

def calculate_double(data):
    return data * 2
```
