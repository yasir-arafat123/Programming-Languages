# Verifying Installation

Once installed, we must confirm everything works correctly.

## 🎯 Learning Objectives
- Check Python Version.
- specific check for `pip` (Package Installer).

## 1. Check Python Version

Open your terminal (Command Prompt on Windows) and run:

```python
!python --version
```

**Note**: If you see Python 2.x, try `python3 --version`. On some systems, `python` refers to legacy Python 2.

## 2. Check Pip

**Pip** is the tool you'll use to install third-party libraries (like pandas, requests). It should come installed with Python.

```python
!pip --version
```

## 3. Verify Programmatic Access

Let's ask Python itself where it lives on your hard drive.

```python
import sys
print("Python Executable Location:")
print(sys.executable)

print("\nPython Version Detail:")
print(sys.version)
```
