## 1. The Interactive Shell (REPL)

1.  Open Terminal/Command Prompt.
2.  Type `python` and hit Enter.
3.  Type `print("Hello from the Shell!")` and hit Enter.
4.  Type `exit()` to leave.

## 2. The Script File

1.  Create a file named `hello.py`.
2.  Write `print("Hello from the File!")` inside it.
3.  Save it.
4.  Open Terminal in that folder.
5.  Run `python hello.py`.

```python
%%writefile hello.py
print("Hello from the File, created by Jupyter!")
```
