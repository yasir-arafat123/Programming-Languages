# Choosing an IDE / Editor

Writing code in Notepad is possible, but painful. A good Integrated Development Environment (IDE) is your superpower.

## 🎯 Learning Objectives
- Compare Top Editors: **VS Code**, **PyCharm**, **Sublime Text**.
- Understand the difference between an Editor and an IDE.

## 1. VS Code (Visual Studio Code)

- **Type**: Lightweight Editor + Powerful Extensions.
- **Cost**: Free (Open Source).
- **Verdict**: **The Industry Standard**. It's fast, has an amazing Python extension, and works for almost any language.
- **Key Extensions**: Python (Microsoft), Pylance, Jupyter.

## 2. PyCharm

- **Type**: Full-Blown IDE.
- **Cost**: Community (Free), Professional (Paid).
- **Verdict**: The "Heavy Lifter". Best code navigation, refactoring tools, and debugger out of the box. Can be heavy on RAM.
- **Best For**: Large, complex projects with Django/Web frameworks.

## 3. Sublime Text / Atom / Notepad++

- **Type**: Text Editors.
- **Cost**: Free / Freemium.
- **Verdict**: Fast, distraction-free. Great for quick edits, but lacks deep introspection/debugging without much configuration.

## 4. Jupyter Notebook / Lab

- **Type**: Interactive Computing Platform.
- **Best For**: Data Science, Exploration, Teaching (Like this!).
- **Not Good For**: Building large applications or web servers.
