# Installing Python

Before you can cook, you need a kitchen. Installing Python correctly is the first step to becoming a pro.

## 🎯 Learning Objectives
- Install Python on Windows, macOS, or Linux.
- Understand the **PATH** environment variable (Crucial!).

## 1. Windows Installation

1.  Go to [python.org/downloads](https://www.python.org/downloads/).
2.  Download the **executable installer**.
3.  **⚠️ CRITICAL STEP**: Check the box that says **"Add Python 3.x to PATH"**.
    - If you miss this, typing `python` in the terminal won't work.
4.  Click "Install Now".

![Add to Path](https://miro.medium.com/max/1318/1*uH-pM4jOAA-K_YFcwMv7-g.png)
*(Make sure that box is checked!)*

## 2. macOS Installation

macOS comes with an old version of Python (usually). Do **not** use the system Python.

**Method A: Official Installer**
1.  Download distinct macOS installer from python.org.

**Method B: Homebrew (Recommended for Pros)**
1.  Open Terminal.
2.  Install Homebrew: `brew install python`.

## 3. Linux Installation

Most distros have it installed. To upgrade:

```bash
sudo apt update
sudo apt install python3 python3-pip
```

## 4. What is PATH?

The **PATH** is a list of folders where your Operating System looks for executable files.

- When you type `python`, Windows looks in `C:\Windows`, `C:\System32`, etc.
- If your Python folder isn't in that list, Windows says: `'python' is not recognized`.
- **Fix**: Reinstall Python and check the "Add to PATH" box.
