# Environment Variables Basics

Environment variables are global system variables that store configuration (like Database URLs, API Keys, or system paths).

## 🎯 Learning Objectives
- What is an ENV Variable?
- Accessing them in Python (`os.environ`).
- Critical Variables: `PATH` and `PYTHONPATH`.

## 1. Accessing ENV Variables

We use the `os` module.

````python
import os

# List all variables (Print just 5 for brevity)
for key in list(os.environ.keys())[:5]:
    print(f"{key}: {os.environ[key]}")
````

## 2. Setting ENV Variables

You usually set these in your OS terminal or a `.env` file (using `python-dotenv`).

**Terminal (PowerShell):**
```powershell
$env:MY_API_KEY="12345"
python script.py
```

**Terminal (Bash):**
```bash
export MY_API_KEY="12345"
python script.py
```

## 3. PYTHONPATH

Often overlooked. This variable tells Python where *else* to look for modules to import, usually if you have a custom library in a non-standard location.
