# Files, Folders & Navigation

Understanding paths is the #1 reason beginners fail to open files in Python (`FileNotFoundError`).

## 🎯 Learning Objectives
- Absolute vs Relative Paths.
- Dealing with Windows Backslashes (`\`) vs Standard Slashes (`/`).

## 1. Absolute Paths

The full GPS address of a file. It starts from the root drive.

- Windows: `C:\Users\Yasir\Documents\Project\data.csv`
- Mac/Linux: `/home/yasir/documents/project/data.csv`

**Pros**: Always works.
**Cons**: Breaks if you move the project to another computer.

## 2. Relative Paths

The address relative to **where the script is running**.

- `data.csv`: Look in the same folder.
- `data/file.txt`: Look in the `data` subfolder.
- `../data.csv`: Look in the parent folder.

**Pros**: Portable (works on any computer).
**Cons**: Breaks if you run the script from the wrong directory.

## 3. The Pathlib Solution

In Python, **ALWAYS** use `pathlib` to handle paths. It solves the `/` vs `\` problem automatically.

```python
from pathlib import Path

# Create a path compatible with ANY OS
p = Path("data") / "subdirectory" / "file.txt"
print(p)
```
