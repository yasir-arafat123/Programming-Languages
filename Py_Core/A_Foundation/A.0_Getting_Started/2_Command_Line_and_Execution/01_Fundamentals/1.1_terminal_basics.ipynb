# Terminal Basics

You cannot be a developer without knowing the Command Line Interface (CLI). It is the cockpit of your computer.

## 🎯 Learning Objectives
- Navigation (`cd`, `ls/dir`, `pwd`).
- File Manipulation (`mkdir`, `touch`, `rm`).

## 1. Where am I? (`pwd`)

**P**rint **W**orking **D**irectory.
- Windows (PowerShell): `pwd` or `Get-Location`
- Mac/Linux: `pwd`

```python
!pwd
```

## 2. What is here? (`ls`)

**L**i**s**t content.
- Windows (CMD): `dir`
- Windows (PowerShell): `ls` or `dir`
- Mac/Linux: `ls`

```python
!ls
```

## 3. Go somewhere else (`cd`)

**C**hange **D**irectory.
- `cd foldername`: Go inside a folder.
- `cd ..`: Go UP one level.
- `cd ~`: Go to user home directory.

## 4. Make things (`mkdir`)

**M**a**k**e **Dir**ectory.
- `mkdir project_alpha`
