## 1. The REPL (Read-Eval-Print Loop)

When you type `python` in your terminal without a file, you enter the REPL. It reads your input, evaluates it, and prints the result.

**Key variable: `_` (underscore)**
In the REPL, `_` stores the result of the last executed expression.

````python
# Simulating REPL behavior programmatically (REPL-only feature)
# Note: '_' works in Jupyter too, behaving like the standard REPL

10 + 20
````

````python
result = _ + 5  # Uses the result of the previous cell (30)
print(f"30 + 5 = {result}")
````

## 2. Scripts and `__main__`

When you act as a pro developer, you will write `.py` files. A common pattern you will see is:

```python
if __name__ == "__main__":
    main()
```

### Why?
This prevents code from running when the file is **imported** as a module by another script, but allows it to run when executed directly.

Let's simulate this by writing a small temporary script.

````python
%%writefile demo_script.py
# This magic command saves the cell content to 'demo_script.py'

def useful_function():
    return "I calculate important things."

print("Global scope: I run whenever this file is loaded!")

if __name__ == "__main__":
    print("Main block: I ONLY run if you called 'python demo_script.py'")
    print(useful_function())
````

### Experiment 1: Running Directly
The `if` block SHOULD run.

````python
!python demo_script.py
````

### Experiment 2: Importing
The `if` block SHOULD NOT run, but "Global scope" will.

````python
import demo_script
````

## 3. Command Line Arguments (`sys.argv`)

Scripts often need input parameters (e.g., `python image_processor.py photo.jpg`).
Python stores these in `sys.argv`, a **list of strings**.

- `sys.argv[0]`: The name of the script itself.
- `sys.argv[1]`: The first argument passed by the user.

````python
%%writefile args_demo.py
import sys

print(f"Script Name: {sys.argv[0]}")
print(f"Arguments: {sys.argv[1:]}")

if len(sys.argv) > 1:
    print(f"Hello, {sys.argv[1]}!")
else:
    print("Hello, Stranger!")
````

````python
!python args_demo.py --verbose "Alice" 42
````

### Pro Tip: `argparse`
For complex CLIs, never use raw `sys.argv`. Use the `argparse` module (covered in depth later), which handles help messages (`--help`) and type validation automatically.

## 4. Shebangs & Encoding

At the top of Python files, you may see:

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
```

1.  **Shebang (`#!...`)**: On Unix/Linux/macOS, this tells the OS *which interpreter* to use to run the file.
2.  **Encoding Cookie**: Tells Python the source file encoding. Since Python 3 converts defaults to UTF-8, this is rarely strictly necessary now unless you use special characters in string literals or comments, but good for legacy compatibility.
