# Permissions & Execution Policy

Sometimes execution is blocked for security reasons.

## 🎯 Learning Objectives
- **Linux/Mac**: `chmod +x`.
- **Windows**: PowerShell Execution Policy.

## 1. Linux/Mac: `chmod`

If you try to run `./script.py` and get `Permission denied`.

**Fix**:
```bash
chmod +x script.py
```
This adds the **eXecutable** permission flag.

## 2. Windows: ExecutionPolicy

If you try to activate a virtual environment (`.venv\Scripts\activate`) and get `...cannot be loaded because running scripts is disabled`.

**Fix**:
Open PowerShell as **Administrator** and run:
```powershell
Set-ExecutionPolicy RemoteSigned
```
Select `A` (Yes to All).
