# Troubleshooting: Python Not Found

The most common Error 1 for beginners.

```text
'python' is not recognized as an internal or external command...
```

## 🎯 Learning Objectives
- Diagnose PATH issues.
- Fix it permanently.

## 1. The Diagnosis

Open Terminal and type:
```bash
echo %PATH%  # Windows (CMD)
$env:Path    # Windows (PowerShell)
echo $PATH   # Mac/Linux
```

Look closely at the output. Does it contain `C:\Python39\` or `C:\Users\...\AppData\Local\Programs\Python...`?

If **NO**, then Python is not in your PATH.

## 2. The Fix (Windows)

1.  Run the Python installer again.
2.  Select **Modify**.
3.  Select **Next**.
4.  Check **"Add Python to environment variables"**.
5.  Click **Install**.
