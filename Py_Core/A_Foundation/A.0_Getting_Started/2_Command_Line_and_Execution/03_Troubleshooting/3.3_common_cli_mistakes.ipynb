# Common CLI Mistakes

Even seniors make these mistakes. Learn to spot them.

## 🎯 Learning Objectives
- Trying to run Python code in the Shell prompt.
- `python` vs `python3` confusion.
- File extension errors.

## 1. The "Python Inception" Mistake

**Mistake**:
Typing `pip install pandas` INSIDE the Python REPL (`>>>`).

**Result**:
`SyntaxError: invalid syntax`

**Fix**: Exit Python (`exit()`) back to the terminal (e.g., `C:\User>`) before running pip.

## 2. The Version Confusion

**Mistake**:
Running `python script.py` but relying on Python 3 features, while `python` points to Python 2.7.

**Fix**:
Explicitly use `python3` or check `python --version`.

## 3. The Extension Error

**Mistake**:
Saving a file as `script.py.txt` because Notepad hid the extension.

**Fix**:
Enable "Show file name extensions" in File Explorer.
