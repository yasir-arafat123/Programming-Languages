# Run Python Interactive (REPL)

The **Read-Eval-Print Loop** is your scratchpad. Use it for quick math (`135 * 42`), checking types, or testing small functions.

## 🎯 Learning Objectives
- Access to the REPL.
- The `_` variable trick.
- Exiting the REPL.

## 1. Entering the REPL

Type `python` (or `python3`) and hit Enter.

You will see:
```text
>>>
```
The triple chevrons mean "I am listening".

## 2. The `_` Variable

This holds the result of the LAST executed expression.

````python
100 + 20
````

````python
_ * 2  # This uses 120 * 2
# Note: In Jupyter, this sometimes behaves differently if cells are run out of order,
# but in the standard terminal REPL, it is strict.
````

## 3. Exiting

To leave the matrix:
- **Windows**: `Ctrl + Z`, then Enter. Or type `exit()`.
- **Mac/Linux**: `Ctrl + D`. Or `exit()`.
