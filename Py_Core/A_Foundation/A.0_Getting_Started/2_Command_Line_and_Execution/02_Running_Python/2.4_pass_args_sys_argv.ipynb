# Passing Arguments

Scripts are boring if they do the same thing every time. Let's send them data from the outside world.

## 🎯 Learning Objectives
- Reading command line arguments with `sys.argv`.
- Understanding argument types (everything is a string!).

## 1. The `sys.argv` List

It contains strings.
- Index `0`: The script name.
- Index `1`: The first argument.

**Example**:
`python echo.py hello`
`sys.argv` -> `['echo.py', 'hello']`

```python
%%writefile adder.py
import sys

if len(sys.argv) < 3:
    print("Usage: python adder.py <num1> <num2>")
    sys.exit(1)

try:
    num1 = float(sys.argv[1])
    num2 = float(sys.argv[2])
    print(f"Result: {num1 + num2}")
except ValueError:
    print("Error: Arguments must be numbers!")
```
