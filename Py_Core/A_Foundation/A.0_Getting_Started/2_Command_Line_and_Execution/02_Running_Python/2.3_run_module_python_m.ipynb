# Run Module (`-m`)

You will often see commands like `python -m http.server` or `python -m pip install`. What is this `-m` switch?

## 🎯 Learning Objectives
- Understanding Module execution.
- Common modules to run (`json.tool`, `http.server`, `pip`).

## 1. What does it do?

`python -m <module_name>` means:
"Search for the module named module_name in `sys.path` and execute it as the `__main__` script."

It allows you to run library code without knowing exactly where the file is located on your hard drive.

## 2. Examples

### A. Simple HTTP Server
Host the current folder on localhost:8000.
```bash
python -m http.server
```

### B. JSON Formatter
Prettify JSON output.
```bash
echo '{"x":1}' | python -m json.tool
```

### C. Pip
Why `python -m pip` instead of just `pip`?
Because `pip` might belong to a different Python version. `python -m pip` guarantees you install packages for *that specific python interpreter*.
