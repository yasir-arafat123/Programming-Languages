## 1. The Command

```bash

```

```bash
python3.12 filename.py

```

````python
!python demo.py
````

## 2. Redirection (Advanced)

You can redirect input and output.

- `python script.py > output.txt`: Saves print statements to a file instead of the screen.

````python
!python demo.py > run_output.txt
````

````python
# Read it back
with open("run_output.txt", "r") as f:
    print(f.read())
````

````python
# Cleanup
import os
try:
    os.remove("demo.py")
    os.remove("run_output.txt")
except FileNotFoundError:
    pass
````
