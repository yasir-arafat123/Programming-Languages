# Using Help & Dir

The most powerful tool in Python is Python itself.

## 🎯 Learning Objectives
- `dir()` for discovery.
- `help()` for reading manuals.

## 1. Discovery with `dir`

Pass ANY object to `dir()` to see what it can do.

```python
s = "Hello"
print(dir(s))
```

## 2. Reading format with `help`

Pass any object to `help()` to read its documentation.

```python
help(s.upper)
```
