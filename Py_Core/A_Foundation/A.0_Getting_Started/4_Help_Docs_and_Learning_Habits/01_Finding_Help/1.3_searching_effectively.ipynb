# Searching Effectively

Programming is 90% Googling.

## 🎯 Learning Objectives
- Selecting Keywords.
- Reading StackOverflow.

## 1. Keywords

**Bad**: "My python code is broken it shows red text"
**Good**: "Python IndexError list assignment out of range"

## 2. Reading StackOverflow

1.  Look at the **Accepted Answer** (Green Check).
2.  Look at the **Highest Voted Answer** (Sometimes better than accepted).
3.  Check the **Date**. Authentication answers from 2013 are likely wrong in 2025.
