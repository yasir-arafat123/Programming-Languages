# Practice Routine

You cannot learn to swim by reading books about water. You must code.

## 🎯 Learning Objectives
- The 100 Days of Code concept.
- Coding vs Reading.

## 1. The Ratio

Aim for **30% Reading, 70% Coding**.

If you watch a 10-minute tutorial, spend 20 minutes playing with the code you just saw.

## 2. Consistency > Intensity

Coding for 20 minutes every day is infinitely better than coding for 10 hours once a week.

- It builds muscle memory.
- It keeps concepts fresh in your "RAM".
