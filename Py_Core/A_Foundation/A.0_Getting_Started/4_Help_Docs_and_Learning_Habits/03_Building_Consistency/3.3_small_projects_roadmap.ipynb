# Small Projects Roadmap

Don't build Facebook as your first project. You will quit.

## 🎯 Learning Objectives
- Progressive Overload.
- Project Ideas.

## Level 1: The Basics (Week 1-2)

1.  **Tip Calculator**: Input bill amount, output total with tip.
2.  **Unit Converter**: Miles to Km, C to F.
3.  **Guess the Number**: Loop until user guesses correctly.

## Level 2: Data & Files (Week 3-4)

1.  **To-Do List (CLI)**: Save tasks to a `.txt` file.
2.  **Hangman**: Read words from a file.
3.  **CSV Analyzer**: Read a spreadsheet and print the average.

## Level 3: APIs & Web (Month 2)

1.  **Weather App**: Fetch data from OpenWeatherMap API.
2.  **Portfolio Website**: Build a static site with Flask.
