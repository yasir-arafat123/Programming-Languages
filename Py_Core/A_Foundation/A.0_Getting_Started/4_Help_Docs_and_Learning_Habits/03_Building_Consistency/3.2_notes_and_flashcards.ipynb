# Notes & Flashcards

How to retain what you learn.

## 🎯 Learning Objectives
- Spaced Repetition (Anki).
- Second Brain (Obsidian).

## 1. Syntax is for Flashcards

Don't try to "understand" why `len(list)` is the syntax. Just memorize it.

**Tool**: Anki.
**Card Example**:
- Front: How to get the length of a list?
- Back: `len(my_list)`

## 2. Concepts are for Notes

Understanding "Object Oriented Programming" requires a note, not a flashcard.

**Tool**: Obsidian / Notion.
Write it in your own words. If you can't explain it simply, you don't understand it.
