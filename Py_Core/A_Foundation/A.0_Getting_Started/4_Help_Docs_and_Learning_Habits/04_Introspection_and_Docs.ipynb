# Introspection & Documentation

A "Pro" developer doesn't memorize every function. They know how to **find** answers instantly without leaving their editor. Python enables this via **Introspection**: the ability of the code to examine itself.

## 🎯 Learning Objectives
- Use the "Trinity of Inspection": `type()`, `id()`, `dir()`.
- Master `help()` and Docstrings (`__doc__`).
- Navigate source code from inside Python.

---

## 1. The Trinity: `type`, `id`, `dir`

Whenever you encounter a mysterious object, these three functions reveal its nature.

```python
s = "Hello Code"
```

### A. What is it? (`type`)

```python
print(type(s))
```

### B. Where is it? (`id`)
Returns the memory address (unique identifier).

```python
print(hex(id(s)))
```

### C. What can I do with it? (`dir`)
Possibly the most important function in Python. It returns a list of all valid attributes and methods.

```python
print(dir(s))
# Notice the methods like 'upper', 'split', AND special 'dunder' methods like '__add__'
```

---

## 2. Help and Docstrings

Every well-written Python object carries its own documentation string (`__doc__`).

### `help()` function
Provides a pager interface to read documentation.

```python
help(len)  # Pass the object itself, NOT len()
```

### Writing your own Docstrings
You should document your code using triple quotes `"""` right after the definition.

```python
def calculate_momentum(mass, velocity):
    """
    Calculates the momentum of an object.
    
    momentum = mass * velocity
    
    Args:
        mass (float): Mass in kg.
        velocity (float): Velocity in m/s.
        
    Returns:
        float: Momentum in kg*m/s.
    """
    return mass * velocity
```

```python
# Accessing it programmatically
print(calculate_momentum.__doc__)
```

---

## 3. Advanced Inspection: `inspect` module

For deep diving (e.g., getting source code or argument specs), use `inspect`.

```python
import inspect

# Get the arguments of a function
sig = inspect.signature(calculate_momentum)
print(f"Signature: {sig}")
print(f"Parameters: {list(sig.parameters.keys())}")

# Get the actual source code (if available)
print("\nSource Code:")
print(inspect.getsource(calculate_momentum))
```
