# Minimal Reproducible Example (MRE)

To get help, you must help others reproduce your bug.

## 🎯 Learning Objectives
- Reducing code to the smallest failing unit.

## 1. Eliminate Dependencies

If your bug is in a math calculation, remove the database code, remove the UI code. Just keep the math.

**Bad MRE**:
"Here is my 300 line Django project zip file."

**Good MRE**:
"When I run `x = 10 / 0`, it crashes."

## 2. Hardcode Data

Don't say "It fails when reading my local CSV". Provide the CSV content as a string or list in the code so I can run it too.
