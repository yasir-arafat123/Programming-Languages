# Logging vs Print

Beginners use `print()`. Pros use `logging`.

## 🎯 Learning Objectives
- Why `print` fails in production.
- Using the `logging` module.

## 1. The Problem with Print

1.  You can't turn it off easily.
2.  You don't know WHERE it came from (file, line, time).
3.  It always goes to Standard Output (stdout).

## 2. The Solution: Logging

It has levels: DEBUG, INFO, WARNING, ERROR, CRITICAL.

```python
import logging

# Simple config
logging.basicConfig(level=logging.INFO)

logging.debug("This is for developers (won't show because level is INFO)")
logging.info("System started successfully")
logging.warning("Disk space low")
logging.error("Database connection failed")
```
