# Understanding Tracebacks (Deep Dive)

We covered this briefly, but let's practice.

## 🎯 Learning Objectives
- Identifying the offending line.
- Parsing nested function calls.

```python
def function_a():
    function_b()

def function_b():
    raise ValueError("Something bad happened!")

function_a()
```

## Analysis

1.  **Bottom**: `ValueError: Something bad happened!`.
2.  **Up one**: `function_b`, line 5 (`raise...`).
3.  **Up one**: `function_a`, line 2 (`function_b()`).
4.  **Up one**: `<module>`, line 7 (`function_a()`).

This "Stack" shows the path Python took to get to the error.
