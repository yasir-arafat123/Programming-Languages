# Match Case (Pattern Matching)

*Introduced in Python 3.10*

This is NOT just a `switch` statement. It is **Structural Pattern Matching**.

## 🎯 Learning Objectives

- **Syntax**: `match` and `case`.
- **Sequence Matching**: Matching lists/tuples patterns.
- **Mapping Matching**: Matching dictionaries.
- **Guards**: Adding `if` conditions to cases.
- **Wildcards**: Using `_` as a catch-all.

## 1. Basic Switch-like Behavior

Matching literal values.

```python
status = 404

match status:
    case 200:
        print("OK")
    case 400 | 404:  # Or operator
        print("Client Error")
    case 500:
        print("Server Error")
    case _:
        print("Unknown status")
```

## 2. Sequence Matching (The Power)

You can match the *shape* and *content* of a list.

```python
command = ["move", "north", 10]

match command:
    case ["quit"]:
        print("Quitting")
    case ["move", direction, distance]:
        # Variables 'direction' and 'distance' are bound here!
        print(f"Moving {direction} by {distance} units")
    case ["move", *rest]:
        print(f"Incomplete move command: {rest}")
    case _:
        print("Invalid command")
```

## 3. Mapping Matching

Matching keys in dictionaries. Note: It ignores extra keys.

```python
user = {"name": "Alice", "role": "admin", "id": 123}

match user:
    case {"role": "admin", "name": name}:
        print(f"Welcome Admin {name}")
    case {"role": "user"}:
        print("Welcome User")
    case _:
        print("Access Denied")
```

## 4. Guards

Adding logic to a match.

```python
point = (10, 20)

match point:
    case (x, y) if x == y:
        print("On the diagonal")
    case (x, y):
        print(f"Point at {x}, {y}")
```
