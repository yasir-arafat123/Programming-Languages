## 🎯 Learning Objectives

- **SyntaxError**: The code is gibberish to the parser.
- **IndentationError**: The blocks are misaligned.
- **TabError**: You mixed spaces and tabs.
- **Unexpected EOF**: You forgot to close a parenthesis.

## 5. Pro Tip: Reading Tracebacks

Always read valid Stack Traces from **bottom to top**.
1.  **Bottom Line**: What is the error type and message?
2.  **Line Above**: Where did it happen (Line number)?
3.  **Above that**: The call stack (how did we get here?).
