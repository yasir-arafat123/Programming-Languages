# F-strings (Formatted String Literals)

The modern standard for string interpolation (Python 3.6+).

## 🎯 Learning Objectives

- **Syntax**: Embedding expressions `{}`.
- **Format Specifiers**: Controlling precision (`:.2f`).
- **Debugging**: The `=` magic.
- **Date Formatting**: Direct date support.
- **Performance**: Why f-strings are faster.

## 1. Basic Syntax

Prefix the string with `f` or `F`. Expressions go inside curly braces.

```python
name = "Alice"
age = 30
print(f"Hello {name}, you are {age} years old.")

# Expressions work too
print(f"Next year you will be {age + 1}.")
```

## 2. Format Specifiers (`:`) 

Control how values are converted to string.

- `f`: Fixed point (float).
- `%`: Percentage.
- `,`: Thousands separator.
- `>N`: Right align with width N.

```python
pi = 3.14159265
price = 12345.6789

print(f"Pi (2 decimals): {pi:.2f}")
print(f"Price (money): ${price:,.2f}")
print(f"Percentage: {0.125:.1%}")
print(f"Aligned: |{name:>10}|")
```

## 3. Debugging Magic (`=`)

If you put an `=` sign after your expression, Python prints both the expression text AND the value.

```python
x = 10
y = 20

# Old way
print(f"x={x}, y={y}")

# New way (Python 3.8+)
print(f"{x=}, {y=}, {x+y=}")
```

## 4. Date Formatting

Objects that implement `__format__` (like `datetime`) can be formatted directly.

```python
from datetime import datetime
now = datetime.now()

print(f"Today is: {now:%Y-%m-%d}")
print(f"Time is: {now:%H:%M}")
```

## 5. Performance Note

F-strings are converted to specific bytecode instructions that are faster than `.format()` or `%` concatenation because they are evaluated at runtime as part of the string parsing optimization, rather than calling a method.
