# Comprehension Syntax Overview

Python's most loved feature.

## 🎯 Learning Objectives

- **List Comprehensions**: `[x for x in data]`.
- **Set/Dict Comprehensions**: `{x}` vs `{k:v}`.
- **Filtering**: `if` clauses.
- **Nested Loops**: `for x for y`.
- **Performance**: Why they are faster than loops.

## 1. List Comprehension

Concise way to create lists.

**Syntax**: `[expression for item in iterable if condition]`

```python
nums = [1, 2, 3, 4, 5]
squares = [x**2 for x in nums]
print(squares)

# With Filter
evens = [x for x in nums if x % 2 == 0]
print(evens)
```

## 2. Set & Dict Comprehension

Uses braces `{}`. Distinction is based on presence of colon `:`.

```python
# Set (Unique values)
unique_lengths = {len(w) for w in ["apple", "banana", "pear", "apple"]}
print(f"Set: {unique_lengths}")

# Dict (Key: Value)
word_map = {w: len(w) for w in ["apple", "banana"]}
print(f"Dict: {word_map}")
```

## 3. Nested Comprehensions

Equivalent to nested for loops. The order matches the loops.

```python
matrix = [[1, 2], [3, 4]]

# Flatten matrix
flat = [num for row in matrix for num in row]
print(flat)

# Equivalent loop:
# for row in matrix:
#     for num in row:
#         append(num)
```

## 4. Under the Hood: Performance

Comprehensions are faster because they use a specialized bytecode instruction `LIST_APPEND` that avoids the overhead of looking up the `.append` method attribute on every iteration.

```python
import dis

def using_loop():
    res = []
    for i in range(3):
        res.append(i)
    return res

def using_comp():
    return [i for i in range(3)]

print("--- Loop Bytecode ---")
# You'll see LOAD_METHOD (append) and CALL_METHOD
dis.dis(using_loop)

print("\n--- Comprehension Bytecode ---")
# You'll see LIST_APPEND (Faster)
dis.dis(using_comp)
```
