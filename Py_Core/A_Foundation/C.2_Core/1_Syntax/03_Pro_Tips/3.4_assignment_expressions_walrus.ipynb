# Assignment Expressions (The Walrus `:=`)

Assign and evaluate in one go.

## 🎯 Learning Objectives

- **The Problem**: Repetitive calls (calls in condition + calls in body).
- **The Solution**: Capture the value while checking it.
- **Scope Rules**: Unlike comprehensions, walrus variables leak.
- **Precedence**: Why you often need parentheses.

## 1. The Classic Use Case: `while` Loops

Reading data until it's empty.

```python
# Pre-Walrus (Repetitive)
# input_val = input("Type something (q to quit): ")
# while input_val != 'q':
#     print(f"You said: {input_val}")
#     input_val = input("Type something (q to quit): ")

# With Walrus (DRY - Don't Repeat Yourself)
# We mock input for this notebook
inputs = iter(["Hello", "World", "q"])

while (value := next(inputs)) != 'q':
    print(f"You said: {value}")
```

## 2. List Comprehensions + Filtering

Calculating a value, checking it, and then keeping it.

```python
import math

numbers = [1, 4, 9, -1, 16]

# Task: Get square root, but only if it's integer, and ignore invalid inputs.

# Without Walrus (Need to call sqrt twice or write nested loop)
res = [math.sqrt(x) for x in numbers if x > 0 and math.sqrt(x) > 2]

# With Walrus
# We calculate root once, assign to 'r', check 'r', and then yield 'r'
res_optimized = [
    r 
    for x in numbers 
    if x > 0 and (r := math.sqrt(x)) > 2
]
print(res_optimized)
```

## 3. Scope Leaking

Variables defined in a walrus expression remain in the scope *after* the expression finishes. This is different from loop variables in comprehensions (which don't leak in Python 3).

```python
if (n := 10) > 5:
    pass

print(n)  # 10 (It leaked!)
```
