# Unpacking Basics

assigning multiple variables at once.

## 🎯 Learning Objectives

- **Basic Unpacking**: 1-to-1 mapping.
- **Extended Unpacking**: Using `*` to grab the rest.
- **Swapping**: The Pythonic `a, b = b, a`.
- **Ignorables**: Using `_` to discard values.
- **Nested Unpacking**: Digging deep into structures.

## 1. The Basics (Sequence Unpacking)

You can assign an iterable of size N to N variables directly.

```python
coordinates = (10, 20)
x, y = coordinates
print(f"x: {x}, y: {y}")

# Works for lists and strings too
a, b, c = "XYZ"
print(f"{a}-{b}-{c}")
```

## 2. Extended Unpacking (`*`)

Python 3 introduced the star operator to grab "everything else". It returns a **list**.

```python
numbers = [1, 2, 3, 4, 5]

first, *rest = numbers
print(f"First: {first}")
print(f"Rest: {rest}")

# Middle extraction (Very powerful)
first, *middle, last = numbers
print(f"Middle: {middle}")
print(f"Last: {last}")
```

## 3. Discarding Values (`_`)

By convention, the underscore `_` is used for "I don't care about this variable".

```python
filename = "image.png"
name, _ = filename.split(".")
print(f"Name: {name}")

# Combined with star
first, *_, last = [1, 2, 3, 4, 5, 6]
print(f"Edges: {first}, {last}")
```

## 4. Nested Unpacking

You can match the shape of the data structure exactly.

```python
data = ["Alice", (1990, 5, 12)]
name, (year, month, day) = data

print(f"{name} was born in {year}")
```

## 5. Under the Hood: Swapping

Python has a dedicated bytecode optimization for `a, b = b, a` called `ROT_TWO`. It happens on the stack, without creating a temporary tuple object.

```python
import dis

def swap():
    a = 1
    b = 2
    a, b = b, a

dis.dis(swap)
# Look for ROT_TWO in the output!
```
