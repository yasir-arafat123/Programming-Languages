# Statements vs Expressions

Understanding the fundamental atoms of Python code.

## 🎯 Learning Objectives

- **The Distinction**: Statement (Action) vs Expression (Value).
- **Evaluation**: Why `eval()` takes expressions but `exec()` takes statements.
- **Modern Python**: Using the Walrus Operator (`:=`) to bridge the gap.
- **Lambdas**: Why they are limited to single expressions.

## 1. The Core Difference

### Expressions
Something that **evaluates to a value**. It can be printed, assigned, or used as function arguments.
- `1 + 1`
- `"Hello"`
- `my_list[0]`
- `func()` (if it returns something)

### Statements
An instruction that **performs an action**. It does not return a value itself.
- `x = 5` (Assignment)
- `if x:` (Control flow)
- `import os` (Import)
- `return`

> **Rule of Thumb**: If you can't print it (`print(X)`), it's likely a statement.

```python
# Expression: Returns 15
print(10 + 5) 

# Statement: Returns Nothing (SyntaxError if you try to print it)
# print(x = 10)  # Error!
```

## 2. `eval()` vs `exec()`

Python exposes this distinction in its dynamic execution functions.

- `eval(str)`: Evaluates a single **expression** and returns the value.
- `exec(str)`: Executes a block of **statements** (ignores return values).

```python
# EVAL examples
result = eval("2 * 10")
print(f"Eval result: {result}")

try:
    eval("x = 10")  # This is a statement, so eval fails!
except SyntaxError as e:
    print(f"Eval failed: {e}")

# EXEC examples
import math
exec("y = math.sqrt(16)")  # This works, but returns None
print(f"Exec side effect (y): {y}")
```

## 3. The Walrus Operator (`:=`) 

*Introduced in Python 3.8*

This operator allows you to perform an assignment (normally a statement) **inside** an expression. This is formally called an "Assignment Expression".

**Syntax**: `NAME := EXPRESSION`

```python
# Without Walrus (Verbose)
values = [1, 2, 3]
n = len(values)
if n > 2:
    print(f"List is too long: {n} items")

# With Walrus (Concise)
# We assign 'n' AND check it in one line
if (n := len(values)) > 2:
    print(f"List is too long: {n} items")
```

## 4. Lambda Limitations

Python's `lambda` functions are restricted to a **single expression**. You cannot put statements (like `if` blocks, `try/except`, or `return`) inside them.

This is why lambdas are often called "anonymous functions" but are less powerful than `def`.

```python
# Valid Lambda (Expression)
square = lambda x: x ** 2
print(square(5))

# Invalid Lambda (Statement)
# my_func = lambda x: (y = x + 1)  # SyntaxError!
```

## 5. Under the Hood: AST (Abstract Syntax Tree)

We can peek at how Python parses code to see the node types.

```python
import ast

code_expr = "1 + 1"
code_stmt = "x = 5"

tree_expr = ast.parse(code_expr)
tree_stmt = ast.parse(code_stmt)

print("--- Expression AST ---")
for node in ast.walk(tree_expr):
    print(type(node).__name__) # Look for 'Expr'

print("\n--- Statement AST ---")
for node in ast.walk(tree_stmt):
    print(type(node).__name__) # Look for 'Assign'
```
