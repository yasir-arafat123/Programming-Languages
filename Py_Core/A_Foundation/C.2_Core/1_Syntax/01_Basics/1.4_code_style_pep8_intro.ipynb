# Code Style (PEP 8)

**"Code is read much more often than it is written."** — *Guido van Rossum*

PEP 8 is the Style Guide for Python Code. It is not about personal preference; it is about community consistency.

## 🎯 Learning Objectives

- **The Why**: Readability and maintainability.
- **Key Rules**: Whitespace, naming, imports.
- **Ignoring Rules**: When practical usage beats purity.
- **Tooling**: `flake8`, `black`, `ruff`, `pylint`.

## 1. Naming Conventions

Python has strict conventions for different object types:

| Type | Convention | Example |
| :--- | :--- | :--- |
| Variable | snake_case | `user_count`, `is_valid` |
| Function | snake_case | `calculate_total()` |
| Class | PascalCase | `UserProfile`, `HttpRequest` |
| Constant | SCREAMING_SNAKE | `MAX_RETRIES`, `DEFAULT_TIMEOUT` |
| Private | _leading_underscore | `_internal_helper` |

## 2. Whitespace Rules

### Around Operators
Use spaces around binary operators (`=`, `+`, `==`), but not for keyword arguments.

```python
# ✅ Good
x = 5 + y
def func(key=value):
    pass

# ❌ Bad
x=5+y
def func( key = value ):
    pass
```

### Imports
Imports should be at the top, grouped in order:
1.  Standard Library (`os`, `sys`)
2.  Third Party (`numpy`, `requests`)
3.  Local Application (`myapp.models`)

## 3. The 79 Character Limit

PEP 8 suggests lines should not exceed 79 characters.

**Why?**
- Historically: Terminal width.
- Modernly: Allows opening 3 files side-by-side on a wide monitor.

**Industry Standard**: Many teams relax this to **88** (Black default) or **100**.

## 4. Modern Tooling

Nobody memorizes PEP 8. We use robots.

### Linters (The Police)
They yell at you when you break rules.
- `flake8`: The classic lightweight linter.
- `pylint`: Strict, opinionated, catches logical bugs too.
- `ruff`: Rust-based, extremely fast replacement for flake8/isort.

### Formatters (The Janitor)
They fix the code for you on save.
- `black`: The uncompromizing code formatter. It gives you no choice, zero config.
- `isort`: Sorts your imports automatically.

````python
# Example: Ugly code
def ugly(   x,y ):
    result=x+y
    return result

# Example: What 'Black' would transform it to
def pretty(x, y):
    result = x + y
    return result
````
