# Line Continuation

Managing long lines of code gracefully.

## 🎯 Learning Objectives

- **Implicit Continuation**: The preferred, robust way.
- **Explicit Continuation**: The fragile backslash `\`.
- **String Concatenation**: Automatic string merging.
- **PEP 8 Rules**: Where to break lines.

## 1. Implicit Continuation (The Good Way)

Any code inside parentheses `()`, brackets `[]`, or braces `{}` can be split across multiple lines without any special character. This is preferred because it's robust against trailing whitespace.

**Preferred for**: Function calls, list definitions, complex arithmetic.

```python
# Function Call
my_list = [
    "Item 1",
    "Item 2",
    "Item 3",  # Trailing commas are good practice!
]

# Math Grouping
result = (
    10 + 20 +
    30 + 40
)
print(result)
```

## 2. Explicit Continuation (The Dangerous Way)

Using a backslash `\` at the end of a line.

**Risk**: If you accidentally add a space *after* the backslash, it breaks the syntax and causes a confusing error.

Use only when parentheses aren't an option (e.g., long `with` statements in older Python).

```python
x = 10 + \
    20 + \
    30

# Try adding a space after one of the backslashes above -> PRO TRIP: Don't use this style.
```

## 3. Implicit String Concatenation

Python automatically merges adjacent string literals. This is useful for splitting long messages.

```python
error_msg = (
    "This is a very long error message "
    "that we want to split across lines "
    "without using plus signs."
)
print(error_msg)
# Note: This only works with literals, not variables!
```

## 4. Where to Break (PEP 8)

**Rule**: Break *before* the operator, not after. This lines up the operators vertically for easier reading (Knuth's style).

```python
# ✅ Good (Operator at start of line)
total = (
    variable_one
    + variable_two
    - variable_three
)

# ❌ Bad (Operator at end of line)
total = (
    variable_one +
    variable_two -
    variable_three
)
```

## 5. Pro Tip: Using `Black`

Don't waste mental energy formatting lines manually. Use an auto-formatter like **Black** or **Ruff**. They enforce these rules automatically on save.
