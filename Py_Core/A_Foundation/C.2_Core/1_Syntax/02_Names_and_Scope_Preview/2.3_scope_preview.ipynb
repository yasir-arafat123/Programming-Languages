# Scope Preview (LEGB)

Where can I see my variables?

## 🎯 Learning Objectives

- **The LEGB Rule**: Standard hierarchy of name resolution.
- **Shadowing**: Hiding variables (and why it's dangerous).
- **UnboundLocalError**: The most common scope bug.
- **Internals**: `locals()` vs `globals()`.

## 1. The LEGB Rule

When you refer to a name `x`, Python searches in this order:

1.  **L**ocal: Inside the current function.
2.  **E**nclosing: Inside enclosing functions (nested).
3.  **G**lobal: At the top level of the module.
4.  **B**uilt-in: Python's pre-defined names (`len`, `print`, `list`).

```python
x = "Global"

def outer():
    x = "Enclosing"
    
    def inner():
        x = "Local"
        print(f"Inside inner: {x}")
    
    inner()
    print(f"Inside outer: {x}")

outer()
print(f"Global scope: {x}")
```

## 2. Shadowing Built-ins (Danger)

Because **Built-in** is last in LEGB, you can accidentally overwrite functions like `list` or `str`.

```python
# BAD IDEA:
# list = [1, 2, 3]

# Now if you try to use list() constructor:
# empty = list()  # TypeError: 'list' object is not callable

# FIX: Never name variables after built-in types.
```

## 3. The UnboundLocalError Trap

If you assign to a variable *anywhere* in a function, Python treats it as **Local** for the entire function.

```python
score = 10

def update_score():
    # We want to print global score, then update it
    try:
        print(f"Current score: {score}") # Crash here!
        score = 20 
    except UnboundLocalError as e:
        print(f"Caught error: {e}")

# Why? Because 'score = 20' makes 'score' local.
# So the print statement tries to look up the LOCAL score, which hasn't been assigned yet.
update_score()
```

## 4. Internals: `locals()` and `globals()`

Scopes are essentially dictionaries mapping names to objects. You can inspect them.

```python
my_global_var = 99

def inspect_me():
    my_local_var = 1
    print("--- Locals ---")
    print(locals())

inspect_me()

print("\n--- Globals (Partial) ---")
# keys() just to avoid printing massive output
print([k for k in globals().keys() if "my_" in k])
```
