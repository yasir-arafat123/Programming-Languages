## 1. The Rules

An identifier is a name used to identify a variable, function, class, module, or other object.

1.  **Must start with**: A letter (`a-z`, `A-Z`) or an underscore (`_`).
2.  **Can contain**: Letters, underscores, and digits (`0-9`).
3.  **Cannot start with**: A digit.
4.  **Case Sensitive**: `myVar` and `myvar` are different.

```python
# Valid
my_var = 1
_hidden = 2
var3 = 3

# Invalid
# 2var = 1      (SyntaxError: invalid syntax)
# my-var = 1    (SyntaxError: cannot assign to operator)
```

## 2. Unicode Support (Python 3)

Python 3 source code is UTF-8 by default. This means you can use non-ASCII characters in identifiers.
This is useful for mathematical notation but generally discouraged for production code (hard to type).

````python
# Mathematical notation (Sometimes useful)
π = 3.14159
r = 5
area = π * r**2
print(f"Area: {area}")

# Emojis (Please don't do this in production)
🚀 = "Launch!"
print(🚀)  # Yes, this works!
````

## 3. Keywords

Keywords are reserved words that have special meaning to the compiler. You cannot use them as variable names.

We can list them programmatically.

````python
import keyword

print("Python Keywords:")
print(keyword.kwlist)

print(f"\nTotal keywords: {len(keyword.kwlist)}")

# Check if something is a keyword
print(f"Is 'class' a keyword? {keyword.iskeyword('class')}")
print(f"Is 'print' a keyword? {keyword.iskeyword('print')}") # False! It's a built-in function.
````

## 4. Soft Keywords

Introduced in Python 3.10 for pattern matching (`match`, `case`, `_`).

These words are keywords *only in specific contexts*. You can still use them as variable names (though it's confusing).

````python
# This is valid, but confusing:
match = 10
case = 20
print(match + case)
````

## 5. Pro Tip: `isidentifier()`

You can check if a string is a valid identifier programmatically. Useful when parsing user input or generating code.

````python
print("hello".isidentifier())     # True
print("123".isidentifier())       # False
print("class".isidentifier())     # True (Wait, what?)

# Important: isidentifier() checks SYNTAX, not RESERVATION.
# 'class' is syntactically valid (letters), but it's a keyword.
# To fully validate a variable name:

def is_valid_variable_name(name):
    return name.isidentifier() and not keyword.iskeyword(name)

print(f"Can I name a var 'class'? {is_valid_variable_name('class')}")
````

## 6. Naming Conventions (Internal Meanings)

- `_single_leading_underscore`: Weak "internal use" indicator. `from M import *` ignores these.
- `__double_leading_underscore`: triggers name mangling in classes (becomes `_Class__var`).
- `__double_leading_and_trailing__`: Magic methods (`__init__`). Don't invent your own.
