# Name Binding (Variables)

**"Variables are labels, not boxes."**

In C++, `int a = 10` creates a box named `a` and puts `10` inside. 
In Python, `a = 10` creates an object `10` somewhere in memory, and sticks a label `a` on it.

## 🎯 Learning Objectives

- **References**: How assignment actually works.
- **Aliasing**: Two names, one object (`a is b`).
- **Identity**: Using `id()` to see memory addresses.
- **Garbage Collection**: Reference counting explained.
- **Mutability Trap**: Why `b = a` usually surprises beginners.

## 1. The Post-It Note Model

Think of objects as balloons floating in memory. Think of variables as strings tied to those balloons.

```python
a = [1, 2, 3]
```
We created a list object. We tied string `a` to it.

````python
a = [1, 2, 3]
b = a

# We didn't copy the list. We just tied another string 'b' to the SAME balloon.
print(f"a: {a}")
print(f"b: {b}")

# Let's check the memory addresses
print(f"Address of a: {id(a)}")
print(f"Address of b: {id(b)}")
print(f"Are they the same? {a is b}")
````

## 2. Mutation & Aliasing

Since `a` and `b` touch the same object, changes via one label are visible via the other. This is called **aliasing**.

````python
b.append(99)
print(f"a is now: {a}")  # variable 'a' changed even though we touched 'b'!
````

## 3. Rebinding

If we assign `a` to something else, we cut the string.

````python
a = "New Object"
# Now 'a' is tied to a string object.
# 'b' is STILL tied to the list object.

print(f"a: {a}")
print(f"b: {b}")
print(f"Are they same? {a is b}")
````

## 4. Under the Hood: Reference Counting

Python's Garbage Collector (GC) runs automatically. The primary mechanism is **Reference Counting**.

Every object keeps a count of how many labels (variables) point to it.
- `a = []` -> Ref count 1
- `b = a`  -> Ref count 2
- `del a`  -> Ref count 1
- `del b`  -> Ref count 0 -> **DESTROYED (Memory Freed)**

````python
import sys

x = []
# getrefcount returns usually ref count + 1 (because the argument itself is a reference)
print(f"Ref count of x: {sys.getrefcount(x)}")

y = x
print(f"Ref count after aliasing: {sys.getrefcount(x)}")
````

## 5. Pro Tip: Immutable Objects

Integers and strings are immutable. You cannot change them in place, so aliasing is safe.

````python
x = 10
y = x
y = y + 1  # This creates a NEW integer 11 and binds y to it.

print(f"x: {x}")  # Unchanged
print(f"y: {y}")
````
