# Docstring Conventions (PEP 257)

The official style guide for documentation strings.

## 🎯 Learning Objectives

- **Triple Quotes**: Always use `"""` even for one-liners.
- **Imperative Mood**: "Return" vs "Returns".
- **Multi-line Structure**: Summary, blank line, details.

## 1. The One-Liner

For simple functions. Keep the closing quotes on the same line.

**Rule**: Use the imperative mood (command form).
- ✅ "Do this"
- ❌ "Does this"
- ❌ "Will do this"

```python
# Good
def add(a, b):
    """Return the sum of a and b."""
    return a + b

# Bad (Descriptive)
def sub(a, b):
    """Returns the difference."""

# Bad (Future)
def mul(a, b):
    """Will multiply a and b."""
```
