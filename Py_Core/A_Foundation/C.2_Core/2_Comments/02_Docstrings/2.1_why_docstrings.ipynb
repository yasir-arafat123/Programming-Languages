# Why Docstrings?

Comments for the Machine (and the User).

Unlike comments (`#`), which are ignored by Python, **Docstrings** are retained at runtime. They become part of the object's metadata.

## 🎯 Learning Objectives

- **Runtime Access**: The `__doc__` attribute.
- **Tooling**: How IDEs (IntelliSense) use them.
- **Philosophy**: Documentation-Driven Development.

## 1. Comments vs Docstrings

- **Comment**: For the *developer* reading the source code.
- **Docstring**: For the *user* importing the function (who might never see the source code).

```python
def calculate_area(radius):
    """Calculates the area of a circle given its radius."""
    return 3.14 * radius * radius
```

## 2. Runtime Access (`__doc__`)

Python stores the string literals at the top of a function/class/module in a special attribute.

```python
print(calculate_area.__doc__)

# This allows tools to inspect code they didn't write.
```

## 3. Integration with Built-ins

Even built-in functions have docstrings (written in C, but exposed in Python).

```python
print(len.__doc__)
```

## 4. Documentation Generators

Tools like **Sphinx** and **MkDocs** read these attributes to generate beautiful HTML websites (like the official Python docs). If you write good docstrings, you get a manual for free.
