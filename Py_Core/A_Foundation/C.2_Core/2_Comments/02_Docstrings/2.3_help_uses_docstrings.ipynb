# Help & Introspection

Asking Python for documentation.

## 🎯 Learning Objectives

- **`help()`**: The built-in interactive manual.
- **Jupyter Magic**: Using `?` and `??`.
- **Programmatic Access**: Reading `__doc__`.

## 1. The `help()` Function

Call this on any object to see its docstring formatted nicely.

```python
def greet(name):
    """Say hello to someone."""
    pass

help(greet)
```

## 2. Using `?` (Jupyter Exclusive)

In notebooks, adding a question mark is a shortcut for help. It opens a pager at the bottom of the screen.

```python
greet? # Run this cell to see the docstring popup
```

## 3. Using `??` (Source Code)

Two question marks try to show the **Source Code** of the function, not just the docs. Very useful for debugging open source libraries.

```python
greet?? # Run this to see the source definition
```
