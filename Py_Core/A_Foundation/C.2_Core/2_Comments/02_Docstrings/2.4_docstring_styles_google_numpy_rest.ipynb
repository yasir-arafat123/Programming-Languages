# Docstring Styles

Three major standards compete for dominance. Choose one and stick to it.

## 🎯 Learning Objectives

- **reStructuredText (reST)**: The official Python standard (verbose).
- **Google Style**: Clean, indentation-based.
- **NumPy Style**: Vertical, great for heavy math docs.
