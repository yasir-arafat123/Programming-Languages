# Inline Comments

Comments on the same line as code.

## 🎯 Learning Objectives

- **Usage**: Sparse and specific.
- **Formatting**: Two spaces before the `#`.
- **Use Cases**: Magic numbers, bug IDs, weird logic.
- **Anti-patterns**: Cluttering simple lines.

## 1. Formatting (PEP 8)

Use **at least two spaces** between the code and the comment.

```python
x = x + 1  # Correct (2 spaces)
x = x + 1 # Incorrect (1 space is cramped)
```

Inline comments are great for explaining constants, regex, or bug fixes.
