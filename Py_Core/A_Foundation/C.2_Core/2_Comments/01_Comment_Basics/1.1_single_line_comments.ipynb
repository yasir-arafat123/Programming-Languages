## 1. Syntax & Formatting

A comment starts with `#` and continues to the end of the line. Python ignores it completely.

**PEP 8 Rules**:
1.  Start with a capital letter.
2.  Use a space after the `#`.

## 2. The "Why" vs "How" Rule

**Bad Comment (How)**: Explains what the code is doing. The code already tells us that.
**Good Comment (Why)**: Explains business logic, edge cases, or history.

```python
x = 10
y = 0

# BAD: Increment x by 1 (We can see that!)
x += 1

# BAD: checks if y is zero
if y == 0:
    x = 0

# GOOD: Prevent division by zero error in legacy module
if y == 0:
    x = 0

# GOOD: 86400 is seconds in a day
SECONDS_PER_DAY = 86400
```

## 3. Code Rot (Lies)

A comment that contradicts the code is worse than no comment at all.

```python
# Returns the user's age
def get_user_data():
    return "John Doe"  # The code changed, but comment didn't!

# Pro Tip: Trust the code, verify the comment.
```
