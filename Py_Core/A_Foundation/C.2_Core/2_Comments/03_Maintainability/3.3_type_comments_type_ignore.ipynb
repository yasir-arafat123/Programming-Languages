# Type Comments & Ignores

Talking to the Type Checker (MyPy).

## 🎯 Learning Objectives

- **Modern**: Type Hints (`fails: int`).
- **Legacy**: Type Comments (`# type: int`).
- **Suppressing Errors**: The `# type: ignore` escape hatch.
- **Best Practices**: Being precise with ignores.

## 1. Legacy Type Comments

Before Python 3.6 (and 3.0), we didn't have annotation syntax. We used comments.

```python
# Python 2 style (Still parsed by MyPy!)
x = 1  # type: int

def add(a, b):
    # type: (int, int) -> int
    return a + b

# Modern Python (Preferred)
x: int = 1
def add_modern(a: int, b: int) -> int:
    return a + b
```

## 2. Type Ignore

Sometimes the type checker is wrong, or you are doing dynamic magic it can't understand. Use `# type: ignore` to silence it.

```python
import random

# Simulation: A library function that lacks type hints
mystery_val = random.choice(["10", 10])

# MyPy would complain here because mystery_val is Union[str, int]
res = mystery_val + 5  # type: ignore

# BETTER: Be specific about the error code
res = mystery_val + 5  # type: ignore[operator]
```

## 3. Warning

Using `ignore` allows bugs to slip through. Use it as a last resort.
