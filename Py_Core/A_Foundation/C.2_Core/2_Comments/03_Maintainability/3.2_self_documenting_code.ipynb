# Self-Documenting Code

The art of writing code that speaks.

## 🎯 Learning Objectives

- **Naming**: `days` vs `d`.
- **Boolean Flags**: `is_valid`.
- **Extraction**: Naming blocks of code by putting them in functions.
- **Magic Numbers**: Named constants.

## 1. Naming Things

Variable names should contain units or intent if ambiguous.

```python
# Bad
d = 10 
# Sleep for d ?? Seconds? Minutes? Days?

# Good
timeout_seconds = 10
elapsed_days = 10
```

## 2. Boolean Logic

Avoid complex conditions in `if` statements. Assign them to a named variable.

```python
user = {"age": 25, "role": "admin"}

# Hard to read
if user["age"] > 18 and user.get("role") == "admin":
    pass

# Self-documenting
is_adult = user["age"] > 18
is_admin = user.get("role") == "admin"

if is_adult and is_admin:
    pass
```
