# Comment Smells

Signs that your code needs refactoring, not explaining.

## 🎯 Learning Objectives

- **Redundant Comments**: Stating the obvious.
- **Journal Comments**: Log entries that belong in Git.
- **Divider Comments**: Trying to group code that should be functions.
- **Excuses**: "Sorry about this..."

## 1. Redundancy (The most common smell)

If the code is as readable as English, the comment is noise.

```python
i = 0
# Set i to 0      <-- BAD
i += 1            # Increment i <-- BAD

user_list.clear() # Clear the list <-- BAD

# Fix: Just delete them.
```

## 2. Journaling (The Time Machine)

Do not keep a history of edits in the file header. That is what Version Control (Git) is for.

```python
# BAD:
# 2023-01-01: Created file (Bob)
# 2023-02-15: Fixed bug in parser (Alice)
# 2023-03-10: Added timeout (Bob)
```

## 3. Divider Comments (ASCII Art)

If you need giant banners to separate sections of a function, your function is too long.

```python
def massive_function():
    # ==========================
    # PHASE 1: INITIALIZATION
    # ==========================
    pass
    
    # ==========================
    # PHASE 2: PROCESSING
    # ==========================
    pass

# Fix: Extract Methods
# def massive_function():
#     initialize()
#     process()
```

## 4. Apology Comments

"Sorry this is ugly", "This is a hack".

If you have time to write an apology, you might have time to fix it. If you can't fix it, explain **Why** it must be this way (e.g., "Performance constrained", "Library bug").
