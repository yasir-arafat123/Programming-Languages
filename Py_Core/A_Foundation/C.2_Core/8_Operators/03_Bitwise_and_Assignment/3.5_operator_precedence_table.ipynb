## 1. Associativity

Most operators are Left-to-Right.

```python
print(10 - 5 - 2)
# (10 - 5) - 2 = 3
```

**Exception**: Exponentiation (`**`) is Right-to-Left.

```python
print(2 ** 3 ** 2)
# 2 ** (3 ** 2) = 2 ** 9 = 512
```
