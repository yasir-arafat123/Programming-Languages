# Bitwise Operators

Manipulating integers at the bit level.

## 🎯 Learning Objectives
- `&`, `|`, `^`.
- `<<`, `>>`.
- `~` (Invert).

## 1. Logic Gates (Integers)

```python
a = 0b1010  # 10
b = 0b1100  # 12

print(bin(a & b))  # 0b1000 (8) - AND
print(bin(a | b))  # 0b1110 (14) - OR
print(bin(a ^ b))  # 0b0110 (6)  - XOR
```

## 2. Shifting

Moving bits left or right (multiply/divide by 2).

```python
print(2 << 1)  # 4 (0b0010 -> 0b0100)
print(10 >> 1) # 5 (0b1010 -> 0b0101)
```
