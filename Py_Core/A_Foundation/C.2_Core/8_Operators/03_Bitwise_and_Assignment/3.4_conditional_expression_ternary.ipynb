# Conditional Expression (Ternary Operator)

The one-line `if-else`.

## 🎯 Learning Objectives
- Syntax: `x if C else y`.
- When to use.

## 1. Syntax

```python
status = "Adult" if age >= 18 else "Minor"
```

Equivalent to:
```python
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
```
