# Augmented Assignment

Combining operation and assignment.

## 🎯 Learning Objectives
- Syntax `+=`, `-=`, etc.
- In-place behavior for lists.

## 1. Syntax

```python
x = 5
x += 2  # 7
x *= 2  # 14
```

## 2. The List Trap (Again)

Recap: `a += b` modifies `a` in place if `a` is a list, unlike `a = a + b`.
