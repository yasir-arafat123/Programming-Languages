# Comparison Operators

Evaluating relationships.

## 🎯 Learning Objectives
- Equality (`==`, `!=`).
- Order (`>`, `<`, `>=`, `<=`).

## 1. Comparing Strings

Lexicographical (dictionary) order.

```python
print("Apple" < "Banana")  # True
print("apple" > "Apple")   # True ('a' is 97, 'A' is 65)
```

## 2. Comparing Different Types

Python 3 raises `TypeError` if valid order doesn't exist.

```python
# print(1 < "1")  # TypeError
```

Equality checks are safe:
```python
print(1 == "1")  # False
```
