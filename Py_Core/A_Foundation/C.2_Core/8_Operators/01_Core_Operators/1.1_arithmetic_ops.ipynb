# Arithmetic Operators (Refresher)

Standard math operations.

## 🎯 Learning Objectives
- `+`, `-`, `*`, `/`.
- `//` (integer division).
- `%` (remainder).
- `**` (power).

## 1. The Big 7

| Operator | Name | Example | Result |
| :--- | :--- | :--- | :--- |
| `+` | Addition | `5 + 2` | `7` |
| `-` | Subtraction | `5 - 2` | `3` |
| `*` | Multiplication | `5 * 2` | `10` |
| `/` | Division | `5 / 2` | `2.5` |
| `//` | Floor Division | `5 // 2` | `2` |
| `%` | Modulo | `5 % 2` | `1` |
| `**` | Exponentiation | `5 ** 2` | `25` |
