# Identity Operators

Are these the same object?

## 🎯 Learning Objectives
- `is` operator.
- `is not` operator.
- `id()` function.

## 1. Syntax

```python
x = None
if x is None:
    print("It is None")
```

## 2. The Interning Optimization

Small integers are cached by Python.

```python
a = 5
b = 5
print(a is b)  # True (Usually)

x = 1000
y = 1000
print(x is y)  # False (Usually, depends on implementation)
```
