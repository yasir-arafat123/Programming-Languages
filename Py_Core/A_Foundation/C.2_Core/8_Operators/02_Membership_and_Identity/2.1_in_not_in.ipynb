# Membership Operators

Checking presence in a collection.

## 🎯 Learning Objectives
- `in` operator.
- `not in` operator.
- Performance differences (List vs Set).

## 1. Syntax

```python
nums = [1, 2, 3]
if 2 in nums:
    print("Found it")
```

Works on strings too:
```python
if "py" in "python":
    print("Found substring")
```

## 2. Speed

- **List**: O(n) - Must check every item.
- **Set/Dict**: O(1) - Instant lookup using hashes.

Always convert to set if doing millions of lookups.
