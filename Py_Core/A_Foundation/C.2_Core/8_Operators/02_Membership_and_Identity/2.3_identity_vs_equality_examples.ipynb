# Identity (`is`) vs Equality (`==`)

The most common interview question.

## 🎯 Learning Objectives
- Distinguishing value vs memory address.
- When to use which.

## 1. The Scenario

You have two distinct lists with the same numbers.

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]
```

````python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

print(list1 == list2)  # True (Values match)
print(list1 is list2)  # False (Different objects in memory)
````

## 2. Aliasing

```python
list3 = list1
print(list1 is list3)  # True (Aliasing)
```
