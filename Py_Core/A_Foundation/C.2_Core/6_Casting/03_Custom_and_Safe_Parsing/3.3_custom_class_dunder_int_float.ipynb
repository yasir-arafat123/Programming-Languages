# Custom Class Casting

Making your objects convertible.

## 🎯 Learning Objectives
- `__int__`.
- `__float__`.
- `__bool__`.

## 1. Implementing `__int__`

```python
class Dog:
    def __init__(self, age):
        self.age = age
    
    def __int__(self):
        return self.age

d = Dog(5)
print(int(d))  # 5
```

## 2. Implementing `__bool__`

Control truthiness.

```python
class Box:
    def __init__(self, items):
        self.items = items
    
    def __bool__(self):
        return len(self.items) > 0
```
