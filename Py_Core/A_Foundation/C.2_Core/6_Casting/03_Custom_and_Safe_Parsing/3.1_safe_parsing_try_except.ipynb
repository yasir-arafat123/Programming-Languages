# Safe Parsing

Don't trust user input.

## 🎯 Learning Objectives

- **Try-Except-Else**: The standard pattern for parsing.
- **Validation**: Checking before casting.
- **Defaults**: Providing fallback values.

## 1. The Pattern

Wrap casting in a try block to handle garbage input gracefully.

```
def safe_int(value, default=0):
    try:
        return int(value)
    except (ValueError, TypeError):
        # ValueError: "abc"
        # TypeError: None
        return default

print(f"Valid: {safe_int('100')}")
print(f"Invalid: {safe_int('foo')}")
print(f"None: {safe_int(None, default=-1)}")
```
