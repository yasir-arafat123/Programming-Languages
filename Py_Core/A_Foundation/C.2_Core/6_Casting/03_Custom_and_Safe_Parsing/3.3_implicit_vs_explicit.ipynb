# Implicit vs Explicit Casting

**"Explicit is better than implicit."**

## 🎯 Learning Objectives

- **Implicit**: Automatic promotion (Int -> Float).
- **Explicit**: Manual calling of constructors.
- **Philosophy**: Why Python avoids things like `"1" + 1`.

## 1. Implicit Promotion

Python is okay with this because it loses no data.

```
res = 1 + 2.5
print(f"{res} (Type: {type(res)})")
```

## 2. Explicit Requirement

Python refuses to guess when data might be lost or ambiguous.

```
# Implicit (Languages like JS do this)
# "1" + 1 -> "11"? or 2?

# Python forces you to decide:
print(int("1") + 1) # 2 (Math)
print("1" + str(1)) # 11 (Text)
```
