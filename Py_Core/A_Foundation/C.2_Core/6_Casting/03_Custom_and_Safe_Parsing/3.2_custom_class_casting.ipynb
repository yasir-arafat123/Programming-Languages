# Custom Casting Magic

Teaching your objects how to be numbers.

## 🎯 Learning Objectives

- **`__int__`**: Called by `int(obj)`.
- **`__float__`**: Called by `float(obj)`.
- **`__bool__`**: Called by `bool(obj)` or `if obj:`.

## 1. Implementing the Protocols

Useful for wrapper classes (like a `Money` class).

```
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def __int__(self):
        return int(self.celsius)

    def __float__(self):
        return float(self.celsius)

    def __bool__(self):
        # Absoulute zero is False?
        return self.celsius > -273.15

t = Temperature(25.5)
print(f"As int: {int(t)}")
print(f"As float: {float(t)}")
```
