# Try-Except for Casting

Defensive programming when converting user input.

## 🎯 Learning Objectives
- Catching `ValueError`.
- Providing default values.
- Loop until valid.

## 1. Basic Protection

```python
user_input = "abc"

try:
    val = int(user_input)
except ValueError:
    print("That's not a number!")
    val = 0  # Default
```

## 2. Helper Function

```python
def safe_int(value, default=0):
    try:
        return int(value)
    except (ValueError, TypeError):
        return default
```
