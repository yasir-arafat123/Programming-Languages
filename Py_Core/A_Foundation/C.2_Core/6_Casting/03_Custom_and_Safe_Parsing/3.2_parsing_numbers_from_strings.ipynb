# Parsing Numbers from Strings

Real-world data is messy.

## 🎯 Learning Objectives
- Cleaning strings (`replace`, `strip`).
- Handling currency (`$`).
- Handling commas (`,`).

## 1. Cleaning

```python
price = "$1,234.50"

# Step 1: Remove symbol
clean = price.replace("$", "")

# Step 2: Remove commas
clean = clean.replace(",", "")

# Step 3: Cast
value = float(clean)  # 1234.5
```

## 2. Using Locale

For international apps, `1.234` might mean "one thousand". Use the `locale` module for robust parsing.
