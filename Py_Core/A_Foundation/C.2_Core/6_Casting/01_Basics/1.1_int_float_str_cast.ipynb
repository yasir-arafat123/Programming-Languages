# Basic Type Casting

Explicitly converting data types.

## 🎯 Learning Objectives
- `int()` rules.
- `float()` rules.
- `str()` rules.

## 1. int()

Truncates floats, parses strings.

```python
print(int(3.99))   # 3 (Truncates, doesn't round)
print(int("10"))   # 10
# int("10.5")      # ValueError (Can't parse float string to int directly)
```

## 2. float()

```python
print(float(10))     # 10.0
print(float("3.14")) # 3.14
```

## 3. str()

Works on almost everything.

```python
print(str(100))      # "100"
print(str([1, 2]))   # "[1, 2]"
```
