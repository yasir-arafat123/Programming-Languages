# Basic Casting

Forcing types to change.

## 🎯 Learning Objectives

- **Mechanism**: Casting is establishing a *new* object from an old one.
- **Numeric**: `int()`, `float()` rules.
- **String**: `str()` conversion.
- **Loss**: When casting loses data.

## 1. Int to Float (and back)

Float to Int truncates (cuts off decimals).

```
f = 10.99
i = int(f)
print(f"Original: {f}, Casted: {i}") # 10

# Int to Float is safe (mostly)
print(float(10)) # 10.0
```

## 2. Strings to Numbers

Must be valid syntax, or `ValueError` is raised.

```
s = "123"
n = int(s)
print(n + 1)

# s2 = "12.5"
# int(s2) # ValueError: invalid literal for int() with base 10: '12.5'

# Fix: Cast to float *then* int
print(int(float("12.5"))) # 12
```
