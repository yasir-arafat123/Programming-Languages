# bool() Rules (Truthy vs Falsy)

Everything in Python is True or False.

## 🎯 Learning Objectives
- What triggers `False`.
- Everything else is `True`.
- Implicit boolean checks.

## 1. The Falsy List

The following evaluate to `False`:

- `None`
- `False`
- Zero (`0`, `0.0`, `0j`)
- Empty sequences (`""`, `[]`, `()`, `{}`)

## 2. Implicit Checks

You usually don't need `bool(x)`. `if x:` does it automatically.

```python
name = ""
if not name:  # Checks if name is empty
    print("Name is required")
```

## 3. Gotchas

```python
if [0]: 
    print("True!") # List is not empty, even if content is 0
```
