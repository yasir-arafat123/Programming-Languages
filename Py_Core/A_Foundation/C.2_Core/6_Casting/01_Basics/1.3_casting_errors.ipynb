# Common Casting Errors

When explicit is better than implicit.

## 🎯 Learning Objectives

- **TypeError**: Mixing types in operations.
- **ValueError**: Invalid content for casting.
- **None**: Trying to cast None.

## 1. The Concatenation Trap

Python won't auto-cast `int` to `str` during concatenation.

```
age = 25
# print("I am " + age) # TypeError

print("I am " + str(age)) # Correct
```

## 2. Casting None

`int(None)` is an error. `str(None)` is `"None"`.

```
try:
    int(None)
except TypeError as e:
    print(f"Caught: {e}")

print(f"Str None: {str(None)}")
```
