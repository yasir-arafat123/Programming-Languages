# Boolean Casting (Truthiness)

What is True?

## 🎯 Learning Objectives

- **Falsy**: What evaluates to False (`0`, `""`, `[]`, `None`).
- **Truthy**: Everything else.
- **Usage**: Implicit casting in `if` statements.

## 1. The Falsy List

Memorize these. Everything else is True.

```
falsies = [False, None, 0, 0.0, 0j, "", [], (), set(), {}]

for item in falsies:
    print(f"{repr(item)} is {bool(item)}")
```

## 2. Implicit Casting

`if variable:` automatically calls `bool(variable)`.

```
name = ""
if not name: 
    print("Name is empty")
```
