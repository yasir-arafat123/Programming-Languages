# Common Casting Errors

When conversion fails.

## 🎯 Learning Objectives
- `ValueError: invalid literal`.
- `TypeError`.
- Handling mixed inputs.

## 1. The Classic: int("float")

```python
int("10.5")  # ValueError
```

**Fix**: Cast to float first, then int.
```python
int(float("10.5"))  # 10
```

## 2. Casting None

```python
int(None)  # TypeError
str(None)  # "None" (Dangerous if not expected)
```
