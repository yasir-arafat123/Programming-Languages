# Map & Filter Casting

Handling lazy iterators.

## 🎯 Learning Objectives

- **Laziness**: `map` and `filter` return iterators, not lists.
- **Materialization**: Foreing them into memory with `list()`.
- **Consumption**: An iterator can only be read once.

## 1. The Print Issue

If you print a map object, you see `<map object at ...>`.

```
nums = [1, 2, 3]
m = map(str, nums)

print(m)
print(list(m)) # ['1', '2', '3']
```

## 2. One-Time Use

Iterators are exhausted after use.

```
m = map(str, nums)
first_list = list(m)
second_list = list(m) # Empty! Iterator is dead.

print(f"First: {first_list}")
print(f"Second: {second_list}")
```
