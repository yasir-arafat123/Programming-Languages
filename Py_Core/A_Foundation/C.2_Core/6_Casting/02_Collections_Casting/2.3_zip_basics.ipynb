# Zip

Iterating two lists at once.

## 🎯 Learning Objectives

- **`zip()`**: Combines N iterables into tuples.
- **Length**: Stops at the shortest list.
- **Unzip**: Using `*` to reverse it.

## 1. Basic Zipping

Useful for matrices.

```python
names = ["Alice", "Bob"]
ids = [1, 2]

combined = list(zip(names, ids))
print(combined)
```

## 2. Unzipping

Using the star unpacking operator.

```python
zipped = [('Alice', 1), ('Bob', 2)]
unzipped_names, unzipped_ids = zip(*zipped)

print(unzipped_names)
print(unzipped_ids)
```
