# Collection Casting

Changing the container, keeping the data.

## 🎯 Learning Objectives

- **Iterable Protocol**: Constructors (`list()`, `set()`) accept *any* iterable.
- **Deduplication**: `list(set(x))`.
- **Dict**: `dict()` from pairs.

## 1. List <-> Tuple

Cheap way to switch mutability.

```
t = (1, 2, 3)
l = list(t)
l.append(4)
t2 = tuple(l)
print(t2)
```

## 2. To Set (Unique)

Fastest way to remove duplicates.

```
nums = [1, 2, 2, 3, 3, 3]
unique = list(set(nums))
print(unique) # Order is NOT guaranteed anymore!
```

## 3. Creating Dicts

`dict()` constructor accepts a list of (key, value) pairs.

```
pairs = [("a", 1), ("b", 2)]
d = dict(pairs)
print(d)
```
