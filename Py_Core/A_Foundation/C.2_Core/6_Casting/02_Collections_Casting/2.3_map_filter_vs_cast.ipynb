# Map/Filter vs Cast

Iterators are lazy; casting wakes them up.

## 🎯 Learning Objectives
- Lazy evaluation.
- Forcing execution with `list()`.
- Memory consistency.

## 1. The Lazy Iterator

Functions like `map`, `filter`, `zip`, and `range` return iterators, not lists.

```python
m = map(str, [1, 2, 3])
print(m)  # <map object at ...>
```

## 2. Materializing (Casting)

To see the data, cast it.

```python
print(list(m))  # ['1', '2', '3']
```

⚠️ **Warning**: You can only consume an iterator once! `list(m)` again will be `[]`.
