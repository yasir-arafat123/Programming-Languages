# Casting Collections

Converting between lists, tuples, and sets.

## 🎯 Learning Objectives
- `list()`: To mutable.
- `tuple()`: To immutable.
- `set()`: Deduplication.
- Order preservation.

## 1. De-duplication (set conversion)

Fastest way to remove duplicates.

```python
items = [1, 2, 2, 3]
unique = list(set(items))
# [1, 2, 3] (Warning: Order is lost!)
```

## 2. Freezing (tuple conversion)

To use a list as a dictionary key.

```python
key = tuple([1, 2, 3])
my_dict = {key: "value"}
```

## 3. String to List

```python
list("abc")  # ['a', 'b', 'c']
```
