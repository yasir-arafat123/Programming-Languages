# Dict Casting Patterns

Creating dictionaries from other types.

## 🎯 Learning Objectives
- `dict()` constructor.
- List of tuples.
- `zip()` trick.

## 1. From List of Tuples

The standard input format for `dict()`.

```python
pairs = [("a", 1), ("b", 2)]
d = dict(pairs)  # {'a': 1, 'b': 2}
```

## 2. Keyword Arguments

Clean syntax for simple string keys.

```python
d = dict(name="Alice", age=30)
```

## 3. Zipping Lists

Merging two parallel lists.

```python
keys = ["name", "age"]
vals = ["Alice", 30]

d = dict(zip(keys, vals))
```
