# Constants

Variables that probably shouldn't vary.

## 🎯 Learning Objectives

- **Convention**: `ALL_CAPS`.
- **Mechanism**: There is no enforcement at runtime.
- **Type Hints**: Using `Final` from `typing`.

## 1. The Honor System

In Python, if you see a variable in `UPPER_CASE`, you promise not to change it. But Python won't stop you.

```python
MAX_CONNECTIONS = 100

# You CAN do this, but you shouldn't:
MAX_CONNECTIONS = 200 
print("I just broke the rules.")
```

## 2. Using `typing.Final` (Python 3.8+)

This doesn't stop runtime changes, but it flags errors in your static type checker (MyPy/VSCode).

```python
from typing import Final

PI: Final = 3.14159

# IDEs will underline the next line in red:
PI = 3.0  # Type Check Error: Cannot assign to Final[float]
```
