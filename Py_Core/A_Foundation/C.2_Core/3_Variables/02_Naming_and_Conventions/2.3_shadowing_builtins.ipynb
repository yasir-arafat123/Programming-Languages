# Shadowing Built-ins

Overwriting the standard library.

## 🎯 Learning Objectives

- **The Problem**: Naming a variable `list` or `str`.
- **The Consequence**: Loosing access to the constructor.
- **The Fix**: `builtins` module.

## 1. Accidental Shadowing

It's valid Python to assign to `print`, but it's a disaster for the session.

```python
# Common mistake in data processing
# list = [1, 2, 3] 

# Later...
# empty = list()  # TypeError: 'list' object is not callable
```

## 2. Recovery (`builtins`)

If you mess up, or if you *really* need to access the original function inside a shadowed scope (like a function implementing a wrapper), import `builtins`.

```python
import builtins

# Let's break print purposely
print_backup = print
print = "I am a string now"

try:
    print("Hello")
except TypeError:
    builtins.print("Standard print failed, but builtins.print works!")

# Restore it
print = print_backup
```
