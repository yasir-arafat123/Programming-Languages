# Variable Naming Rules

Choosing names that don't crash the compiler.

## 🎯 Learning Objectives

- **Syntax**: Letters, numbers, underscores.
- **Case Sensitivity**: `Age` vs `age`.
- **Unicode**: Non-ASCII identifiers (and why to avoid them).

## 1. Syntax Review

Must not start with a number. No hyphens.

```python
user_id = 5     # Valid
_private = 10   # Valid

# 1st_user = 5  # SyntaxError (Starts with digit)
# user-id = 5   # SyntaxError (Hyphen is subtraction)
```

## 2. Case Sensitivity

Python distinguishes case. This leads to common bugs where you define `Path` but try to use `path`.

```python
Score = 100
score = 50

print(f"Big Score: {Score}")
print(f"Little score: {score}")
```

## 3. Unicode and Normalization

Python 3 normalizes Unicode identifiers using NFKC (Normalization Form KC).

This means distinct characters that look alike might be treated as the SAME variable.

```python
import unicodedata

# Did you know?
a = 10
ａ = 20 # This uses the Fullwidth Latin Small Letter A (U+FF41)

print(f"Normal a: {a}")      # Surprise! It prints 20.
print(f"Fancy a: {ａ}")

# Why? Because normalization maps them to the same identifier.
```
