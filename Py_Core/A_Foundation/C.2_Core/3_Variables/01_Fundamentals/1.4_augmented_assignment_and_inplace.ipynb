# Augmented Assignment

`+=`, `-=`, and friends.

## 🎯 Learning Objectives

- **Syntax**: `x += 1`.
- **Optimization**: How it differs from `x = x + 1`.
- **In-Place**: The `__iadd__` method for mutable objects.

## 1. Immutable Behavior (floats/ints)

For numbers/strings, `x += 1` is exactly `x = x + 1`. It creates a NEW object.

````python
x = 10
original_id = id(x)
x += 1
print(f"Is same object? {id(x) == original_id}") # False
````

## 2. Mutable Behavior (lists)

For lists, `+=` calls `extend()`. It modifies the object **IN PLACE**. This is faster.

````python
my_list = [1, 2]
original_id = id(my_list)

my_list += [3, 4]  # Same as my_list.extend([3, 4])

print(f"Is same object? {id(my_list) == original_id}") # True
````

## 3. The Tuple Trap

What happens if you have a tuple containing a list? `(list)`

```python
t = ([1],)
t[0] += [2]  # Throws TypeError (tuple immutable) BUT updates list (list mutable)
```

This is a classic Python interview question.
