# Multiple Assignment

Efficiency techniques for initialization.

## 🎯 Learning Objectives

- **Parallel Assignment**: `x, y = 1, 2`.
- **Chained Assignment**: `x = y = 0`.
- **The Mutable Trap**: Why `x = y = []` is dangerous.

## 1. Parallel Assignment (Tuple Unpacking)

Safest way to initialize multiple variables.

```python
x, y, z = 1, 2, 3
print(f"{x}, {y}, {z}")
```

## 2. Low-Level Swapping

Because the right-hand side is evaluated *before* assignment, you can swap without a temp variable.

```python
x, y = y, x
print(f"x: {x}, y: {y}")
```

## 3. Chained Assignment (Danger Zone)

This sets multiple variables to the **SAME OBJECT**.

```python
# Safe for immutables
a = b = 0

# Dangerous for mutables
list_a = list_b = []
list_a.append(1)

print(f"list_b: {list_b}")  # [1] !! list_b changed too!
```
