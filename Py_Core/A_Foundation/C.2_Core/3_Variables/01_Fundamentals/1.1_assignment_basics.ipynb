# Assignment Basics

How Python connects names to objects.

## 🎯 Learning Objectives

- **Assignment**: `x = 10` is binding, not copying.
- **Dynamic Typing**: Changing the type of a variable.
- **Identity**: Using `id()` to verify object reuse.
- **Interning**: Small integer optimization.

## 1. Assignment is Binding

Read `x = 10` as: "Create an object `10`, and attach label `x` to it."

```python
x = 10
print(f"x: {x}, type: {type(x)}, id: {id(x)}")
```

## 2. Reassignment (Moving the Tag)

When you assign a new value to an existing variable, you are just moving the tag to a new balloon. The old balloon is abandoned (garbage collected if no other tags exist).

```python
x = 20
print(f"x: {x}, id: {id(x)}")  # ID changes!

x = "Hello"  # Dynamic Typing: Variable can hold references to any type
print(f"x: {x}, type: {type(x)}")
```

## 3. The `id()` Function

This returns the memory address (in CPython). Two variables are the same if their `id()` is the same.

```python
a = 5
b = 5
print(f"a is b? {a is b}")  # True (Python caches small integers)
```
