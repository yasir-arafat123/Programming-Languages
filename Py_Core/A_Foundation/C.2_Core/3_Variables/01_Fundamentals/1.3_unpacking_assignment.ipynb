# Unpacking Assignment

Advanced ways to destructure sequences.

## 🎯 Learning Objectives

- **Star (`*`)**: Grabbing the rest.
- **Underscore (`_`)**: Ignoring values.

## 1. Star Operator

Only one star allowed per assignment.

```python
head, *tail = [1, 2, 3, 4]
print(f"Head: {head}, Tail: {tail}")  # Tail is a list [2, 3, 4]
```

## 2. Middle Unpacking

Very useful for CSV parsing logic.

```python
first, *middle, last = "PRE_DATA_POST".split("_")
print(f"{first} | {middle} | {last}")
```
