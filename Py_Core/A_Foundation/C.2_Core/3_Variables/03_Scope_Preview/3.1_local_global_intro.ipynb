# Local vs Global

Where variables live and die.

## 🎯 Learning Objectives

- **Local Scope**: Inside functions. Born when called, dies when returned.
- **Global Scope**: Module level. Lives as long as the program runs.
- **Lifetime**: Why accessing a local variable outside crashes.

## 1. Local Scope (The Temporary Workspace)

Variables created inside a function are invisible to the outside world.

```python
def secret_keeper():
    secret = "I am hidden"
    print(f"Inside: {secret}")

secret_keeper()

# print(secret)  # NameError: name 'secret' is not defined
```

## 2. Global Scope (The Public Workspace)

Variables defined at the top level are visible everywhere (for reading).

```python
config_val = 42

def reader():
    print(f"I can see global: {config_val}")

reader()
```

## 3. The Shadowing Trap (Reading vs Writing)

If you try to *write* to a global variable inside a function without declaration, Python creates a new LOCAL variable instead.

```python
x = 10

def confuse():
    x = 20  # This creates a NEW local x. Global x is untouched.
    print(f"Local x: {x}")

confuse()
print(f"Global x: {x}")
```
