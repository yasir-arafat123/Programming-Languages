# The `global` Keyword

Breaking the barrier between local and global.

## 🎯 Learning Objectives

- **Usage**: Modifying global variables inside functions.
- **Dangers**: Why Global State is the root of all evil (concurrency issues, testing pain).

## 1. Modifying Globals

Use `global variable_name` to tell Python: "I don't mean a new local variable, I mean the one outside."

```python
counter = 0

def increment():
    global counter  # Bind to the outer scope
    counter += 1

increment()
increment()
print(f"Counter: {counter}")
```

## 2. Why to Avoid This

Functions that rely on global variables are "impure". They depend on hidden state. This makes them:
1.  Hard to test (you have to reset the global before every test).
2.  Hard to debug (who changed the global?)

**Better approach**: Pass arguments and return values.

```python
# Better Functional Style
def increment_struct(current_val):
    return current_val + 1

val = 0
val = increment_struct(val)
print(val)
```
