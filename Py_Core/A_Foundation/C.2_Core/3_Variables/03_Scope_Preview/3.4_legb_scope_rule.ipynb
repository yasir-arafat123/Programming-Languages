# LEGB Rule (Advanced)

The full picture of Python's Name Resolution.

## 🎯 Learning Objectives

- **LEGB**: Local -> Enclosing -> Global -> Built-in.
- **Class Scope**: Why methods can't see class variables directly.
- **List Comprehension Scope**: How it changed in Python 3.

## 1. Visualizing LEGB

Python checks scopes strictly in this order. It stops at the first match.

```python
# Built-in Scope (len is here)

x = "Global"

def outer():
    x = "Enclosing"
    
    def inner():
        x = "Local"
        print(x)
    
    inner()

outer()
```

## 2. The Class Scope Quirk

Classes behave differently. Methods inside a class cannot see variables defined in the class body without using `self.` or `ClassName.`.

```python
class Demo:
    VAR = 10
    
    def method(self):
        # print(VAR) # NameError!
        print(self.VAR) # OK
        print(Demo.VAR) # OK

Demo().method()
```

## 3. Comprehension Scope Leaking (Python 2 vs 3)

In Python 2, `[x for x in range(3)]` would leak `x` into the global scope. In Python 3, comprehensions have their own private scope.

```python
x = "I am global"
res = [x for x in range(5)] # reuse x inside loop

print(x) # Still "I am global" (Python 3 is safe!)
```
