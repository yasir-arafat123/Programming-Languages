# The `nonlocal` Keyword

*Introduced in Python 3*

Modifying variables in the immediate outer function (Enclosing scope).

## 🎯 Learning Objectives

- **Nested Functions**: Functions inside functions.
- **Enclosing Scope**: The space between Local and Global.
- **Closures**: Functions that remember their environment.
- **Nonlocal**: The keyword to write to enclosing variables.

## 1. The Problem

Similar to global, if you write to a variable in a nested function, it creates a new local one, failing to update the parent's variable.

```python
def outer():
    x = "Outer"
    
    def inner():
        x = "Inner"  # Shadows outer x
        print(f"Inner sees: {x}")
    
    inner()
    print(f"Outer sees: {x}")

outer()
```

## 2. The Solution (`nonlocal`)

Use it to target the nearest enclosing scope (excluding globals).

```python
def make_counter():
    count = 0
    
    def counter():
        nonlocal count
        count += 1
        return count
    
    return counter

c = make_counter()
print(c()) # 1
print(c()) # 2
# 'count' is hidden inside the closure of 'c'!
```

## 3. Closures

This pattern is the basis for **Decorators**. The inner function "closes over" the variables it needs from the outer scope, keeping them alive even after `outer()` finishes.
