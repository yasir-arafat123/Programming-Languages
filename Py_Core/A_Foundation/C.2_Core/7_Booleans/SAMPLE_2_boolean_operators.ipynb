# ⚖️ Boolean Operators: and, or, not

**"True and False: The building blocks of all logic."**

Boolean operators are the foundation of conditional logic in programming. Python's implementation goes beyond simple AND/OR/NOT gates—it includes short-circuit evaluation, truthiness, and operator precedence that affect correctness and performance.

## 🎯 Learning Objectives

- Master the three boolean operators: `and`, `or`, `not`
- Understand short-circuit evaluation and its performance implications
- Learn operator precedence and when parentheses are required
- Explore truthiness beyond `True`/`False`
- Apply DeMorgan's Laws for logical simplification

## 1. Theory: Boolean Algebra in Python

### The Foundation

Boolean algebra was invented by George Boole in the 1850s. In Python:
- `and` returns the **first falsy value** or the **last value**
- `or` returns the **first truthy value** or the **last value**
- `not` converts to boolean and inverts

### Truth Tables

| A | B | A and B | A or B | not A |
| :--- | :--- | :--- | :--- | :--- |
| False | False | False | False | True |
| False | True | False | True | True |
| True | False | False | True | False |
| True | True | True | True | False |

### Why This Matters

In real code:
- **Form validation**: `if username and password and agreed_to_terms:`
- **Default values**: `config = user_config or DEFAULT_CONFIG`
- **Guard clauses**: `if not user.is_admin: return`

## 2. Basic Usage (Beginner Level)

Simple comparisons with boolean values.

````
# AND: Both must be True
print(True and True)    # True
print(True and False)   # False
print(False and False)  # False

# OR: At least one must be True
print(True or False)    # True
print(False or False)   # False

# NOT: Inverts the value
print(not True)         # False
print(not False)        # True
````

````
# Practical example: Access control
is_logged_in = True
has_permission = True
is_banned = False

can_access = is_logged_in and has_permission and not is_banned
print(f"User can access: {can_access}")
````

## 3. Intermediate: Short-Circuit Evaluation

**Critical Concept**: Python doesn't evaluate the right side if the left side already determines the result.

### AND Short-Circuit
If left is `False`, right is never evaluated (because `False and X` is always `False`).

### OR Short-Circuit
If left is `True`, right is never evaluated (because `True or X` is always `True`).

````
# Demonstrating short-circuit with side effects
def expensive_operation():
    print("⚠️  Expensive operation called!")
    return True

# AND: If first is False, second never runs
print("--- Testing AND short-circuit ---")
result = False and expensive_operation()
print(f"Result: {result}")
# Notice: "Expensive operation called!" was NOT printed

print("\n--- Testing OR short-circuit ---")
result = True or expensive_operation()
print(f"Result: {result}")
# Again: function not called
````

### Real-World Application: Null Safety

````
user = None  # Simulating missing user

# BAD: This crashes
# if user.is_admin:  # AttributeError: 'NoneType' object has no attribute 'is_admin'

# GOOD: Short-circuit saves us
if user and user.is_admin:
    print("Admin access")
else:
    print("Access denied or no user")

# Explanation: 'user' is None (falsy), so Python never evaluates 'user.is_admin'
````

## 4. Advanced: Return Values & Truthiness

### The Secret: Operators Return Operands, Not Just True/False

````
# AND returns first falsy OR last value
print(0 and 5)           # 0 (first falsy)
print(3 and 7)           # 7 (both truthy, returns last)
print("" and "hello")    # "" (empty string is falsy)

# OR returns first truthy OR last value
print(0 or 5)            # 5 (first truthy)
print(False or 0 or 3)   # 3 (first truthy)
print(False or 0 or "")  # "" (all falsy, returns last)
````

### Pro Pattern: Default Values

````
# Before (verbose)
user_name = None
if user_name is None:
    user_name = "Guest"

# After (Pythonic)
user_name = None
display_name = user_name or "Guest"  # "Guest"
print(f"Hello, {display_name}")

# Another example
config = {}
timeout = config.get("timeout") or 30  # Default to 30 if not set
print(f"Timeout: {timeout} seconds")
````

## 5. Operator Precedence

**Order** (highest to lowest):
1. `not`
2. `and`
3. `or`

Think: **N**ot **A**nd **O**r (NATO)

````
# Without parentheses
result = True or True and False
# Evaluated as: True or (True and False)
# = True or False
# = True
print(f"Result: {result}")

# With parentheses (explicit is better)
result = (True or True) and False
# = True and False
# = False
print(f"Explicit result: {result}")

# Best practice: Use parentheses for clarity even when not required
````

## 6. DeMorgan's Laws (Logical Simplification)

Two powerful transformation rules:

1. `not (A and B)` ≡ `(not A) or (not B)`
2. `not (A or B)` ≡ `(not A) and (not B)`

````
# Example: Refactoring complex conditions
is_weekday = True
is_holiday = False

# BEFORE: Hard to read
if not (is_weekday and not is_holiday):
    print("Don't work!")

# AFTER: Apply DeMorgan's Law
# not (is_weekday and not is_holiday)
# = (not is_weekday) or (not (not is_holiday))
# = (not is_weekday) or is_holiday

if not is_weekday or is_holiday:
    print("Don't work!")
````

## 7. Common Pitfalls

### Pitfall 1: Chaining Comparisons Incorrectly

````
x = 5

# WRONG (but doesn't error - always True!)
if x == 3 or 4 or 5:
    print("BUG: This always prints!")
# Why? Interprets as: (x == 3) or 4 or 5
# 4 is truthy, so result is always True

# CORRECT
if x == 3 or x == 4 or x == 5:
    print("Correct check")

# BETTER (Pythonic)
if x in (3, 4, 5):
    print("Most readable")
````

### Pitfall 2: Forgetting that `and`/`or` Return Values

````
# Unexpected behavior with 0
value = 0
result = value or "default"  # "default" (because 0 is falsy)
print(result)

# If 0 is a valid value, use explicit None check:
value = 0
result = value if value is not None else "default"  # 0
print(result)
````

## 8. Best Practices

### 1. Prefer `in` for Membership
```python
# BAD
if status == "active" or status == "pending":

# GOOD
if status in {"active", "pending"}:
```

### 2. Use `any()` and `all()` for Iterables
```python
# BAD
if cond1 and cond2 and cond3 and cond4:

# GOOD
if all([cond1, cond2, cond3, cond4]):
```

### 3. Avoid Deeply Nested Conditions
```python
# BAD
if A:
    if B:
        if C:
            do_something()

# GOOD (Guard clauses)
if not A: return
if not B: return
if not C: return
do_something()
```

## 9. Summary

| Concept | Key Takeaway |
| :--- | :--- |
| **and** | Returns first falsy OR last value |
| **or** | Returns first truthy OR last value |
| **not** | Converts to bool and inverts |
| **Short-Circuit** | Right side may not execute (performance + safety) |
| **Precedence** | not > and > or |
| **Truthiness** | Empty containers/0/None are falsy |

### Performance Insight
Short-circuiting can save significant computation:
```python
# If cache_hit is True 99% of the time:
if cache_hit or expensive_database_query():
    # Database query skipped 99% of the time!
```
