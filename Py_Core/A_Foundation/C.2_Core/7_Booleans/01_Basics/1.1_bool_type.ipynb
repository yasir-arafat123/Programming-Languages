## 1. Int Inheritance

Fun fact: `bool` is a subclass of `int`.

```python
print(int(True))   # 1
print(int(False))  # 0

print(True + True)  # 2 (Don't do this, but it works)
```

## 2. Keywords

`True` and `False` are keywords (capitalized). `true` works in JS, but not Python.
