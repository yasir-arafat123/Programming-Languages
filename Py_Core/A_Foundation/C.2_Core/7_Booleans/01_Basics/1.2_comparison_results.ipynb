# Comparison Results

Asking questions about data.

## 🎯 Learning Objectives
- Equality (`==` vs `is`).
- Inequality (`!=`).
- Ordering (`>`, `<`).

## 1. Equality (`==`)

Checks if **values** are equal.

```python
print([1, 2] == [1, 2])  # True (Same content)
```

## 2. Identity (`is`)

Checks if **objects** are the same memory location.

```python
a = [1, 2]
b = [1, 2]
print(a is b)  # False (Different distinct lists)
```

**Rule**: Use `is` for `None` check (`if x is None`). Comparison with `==`.

## 3. Chained Comparison

Python magic.

```python
age = 25
if 18 <= age < 65:
    print("Working age")
```
