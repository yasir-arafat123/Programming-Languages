# All & Any Patterns

Checking entire collections at once.

## 🎯 Learning Objectives
- `all()`: True if everything is True.
- `any()`: True if at least one is True.
- Usage with comprehensions.

## 1. any()

```python
errors = [False, False, True]
if any(errors):
    print("Halt!")
```

## 2. all()

```python
checks = [True, True, True]
if all(checks):
    print("Ignition")
```

## 3. Real World

Checking validity.

```python
password = "password123"
requirements = [
    len(password) > 8,
    any(char.isdigit() for char in password), 
    any(char.isupper() for char in password)
]

if all(requirements):
    print("Strong password")
else:
    print("Weak")
```
