# Operator Precedence (Boolean)

Who wins in a fight?

## 🎯 Learning Objectives
- The order: `not`, `and`, `or`.
- Using parentheses for clarity.

## 1. The Hierarchy

Priority from highest to lowest:

1. `not` (Negation)
2. `and` (Conjunction)
3. `or` (Disjunction)

```python
False or True and not False
# Step 1 (not): False or True and True
# Step 2 (and): False or True
# Step 3 (or) : True
```

## 2. Best Practice

Don't rely on memory. Use parentheses.

```python
False or (True and (not False))  # Explicit is better than implicit
```
