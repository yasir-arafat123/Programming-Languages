# Truthiness Rules Refresher

Quick recap of implicit boolean context.

## 🎯 Learning Objectives
- What values are false?
- Using this for Pythonic checks.

## 1. The "Falsy" Setup

Memorize this list. Everything else is `True`.

1. `False`
2. `None`
3. `0`
4. Empty collections: `""`, `[]`, `{}`, `()`
5. Empty sets/ranges.

## 2. Pythonic Checks

❌ **Non-Pythonic**:
```python
if len(my_list) > 0:
    process(my_list)
```

✅ **Pythonic**:
```python
if my_list:
    process(my_list)
```
