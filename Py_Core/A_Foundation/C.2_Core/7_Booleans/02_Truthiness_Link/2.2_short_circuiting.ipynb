# Short Circuiting

The lazy logic optimization.

## 🎯 Learning Objectives
- How `and`/`or` stop early.
- Using `or` for defaults.
- Using `and` for guards.

## 1. OR Short Circuit

Returns the first Truthy value it finds.

```python
name = user_input or "Anonymous"
```

If `user_input` is "Alice", it returns "Alice".
If `user_input` is "" (Falsy), it evaluates the right side ("Anonymous").

## 2. AND Short Circuit

Stops at the first Falsy value.

```python
result = safe_to_proceed and do_heavy_calculation()
```

If `safe_to_proceed` is False, `do_heavy_calculation()` is NEVER CALLED.
