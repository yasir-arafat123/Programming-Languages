# MyPy & Pyright Preview

The tools that actually check your hints.

## 🎯 Learning Objectives
- What is `mypy`?
- Running type checks.
- Catching errors.

## 1. Installation

```bash
pip install mypy
```

## 2. Usage

**script.py**:
```python
x: int = "hello"  # Error!
```

**Terminal**:
```bash
mypy script.py
# Output: Incompatible types in assignment (expression has type "str", variable has type "int")
```

## 3. Pyright

Pyright is another checker, built by Microsoft. It powers the VS Code "Pylance" extension.
