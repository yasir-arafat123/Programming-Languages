# Basic Annotations

How to write type hints.

## 🎯 Learning Objectives
- Variables: `x: int`.
- Functions: `-> return_type`.
- Lists/Dicts: `list[int]`.

## 1. Variables

```python
age: int = 25
name: str = "Alice"
scores: list[int] = [10, 20, 30]
```

## 2. Functions

```python
def greet(name: str) -> str:
    return f"Hello {name}"
```

## 3. Collections (Python 3.9+)

Use built-in types.

```python
def process_scores(data: dict[str, int]) -> None:
    pass
```

(In older Python, `from typing import Dict, List` was required).
