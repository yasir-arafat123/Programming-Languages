# Why Type Hints?

Python is dynamic, so why add types?

## 🎯 Learning Objectives
- Catch bugs early (Static analysis).
- IDE Autocomplete.
- Documentation.

## 1. Catching Bugs

```python
def add(a: int, b: int) -> int:
    return a + b

add("1", 2)  # Type checker yells at you BEFORE you run it
```

## 2. Better Autocomplete

If you tell VS Code that `user` is a `User` object, it can show you `.name` and `.email` when you type `user.`.

## 3. It's Optional

Python ignores type hints at runtime. They are comments that tools understand.
You can add them gradually to an existing codebase.
