# String Formatting

The evolution of text injection.

## 🎯 Learning Objectives

- **Legacy (`%`)**: The C-style printf (avoid for new code).
- **Classic (`.format()`)**: The Python 3.0 way.
- **Modern (f-strings)**: The Python 3.6+ way (Preferred).

## 1. History Lesson

**Operator `%`** borrows from C. It's limited and can be buggy with tuples.

```python
name = "Alice"
print("Hello %s" % name)
# print("Values: %s" % (1, 2)) # Error! Needs explicit tuple wrapper or it confuses arguments.
```

## 2. `.format()` Method

More powerful, allows reordering arguments.

```python
print("First: {0}, Second: {1}".format("A", "B"))
print("Named: {name}, Age: {age}".format(name="Bob", age=20))
```

## 3. f-strings (The Winner)

Fastest and most readable. See 3.2_f_strings_quickstart.ipynb for deep dive.

```python
import math
print(f"Math is easy: {math.pi:.3f}")
```
