# Serialization (Pickle vs JSON)

Saving objects to disk/network.

## 🎯 Learning Objectives

- **Serialization**: Object -> Bytes.
- **Deserialization**: Bytes -> Object.
- **JSON**: Text-based, cross-language, safe.
- **Pickle**: Python-specific, binary, **UNSAFE**.

## 1. JSON (JavaScript Object Notation)

The standard for web APIs. Supports basic types (dict, list, str, int, bool, None).

```python
import json

data = {"name": "Alice", "skills": ["Python", "Rust"]}

# Dump to string
json_str = json.dumps(data)
print(f"JSON String: {json_str}")

# Load from string
print(f"Back to dict: {json.loads(json_str)}")
```

## 2. Pickle (The Dangerous Power)

Pickle can serialize almost ANY Python object (classes, functions). But it executes arbitrary code during loading.

```python
import pickle

class Secret:
    def __init__(self, key):
        self.key = key

obj = Secret("123-ABC")

# Serialize
binary_blob = pickle.dumps(obj)
print(f"Pickle Blob: {binary_blob}")

# Deserialize
restored = pickle.loads(binary_blob)
print(f"Restored Key: {restored.key}")
```

## 3. Security Warning

**NEVER** load a pickle file from an untrusted source. Hackers can construct a pickle that runs `os.system("rm -rf /")` when loaded.
