# Str vs Repr

The two faces of an object.

## 🎯 Learning Objectives

- **`__str__`**: For users. Readable, pretty.
- **`__repr__`**: For developers. Unambiguous, preferably valid code.
- **Default**: If `str` is missing, Python uses `repr`.

## 1. The Distinction

Consider a datetime object.

```python
import datetime
now = datetime.datetime.now()

print("--- STR (print uses this) ---")
print(str(now))

print("\n--- REPR (IDEs use this) ---")
print(repr(now))
```

## 2. Implementing Custom Classes

Let's make a class that demonstrates both.

```python
class User:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        
    def __str__(self):
        return f"User(Name: {self.name})"
        
    def __repr__(self):
        # Goal: repr should look like the code to recreate the object
        return f"User(name='{self.name}', id={self.id})"

u = User("Alice", 42)

print(f"User sees: {u}")
print(f"Dev sees: {repr(u)}")
```

## 3. The `eval(repr(x))` Goal

Ideally, `eval(repr(x)) == x`. This allows robust debugging.
