# Type Introspection

Checking what things are at runtime.

## 🎯 Learning Objectives

- **`type(x)`**: Get the exact class.
- **`isinstance(x, Type)`**: Check inheritance (The Correct Way).
- **`issubclass(A, B)`**: Check class hierarchy.
- **ABCs**: Abstract Base Classes for generic checking.

## 1. Type vs Isinstance

Always prefer `isinstance` because it supports inheritance.

```python
class Animal: pass
class Dog(Animal): pass

d = Dog()

# BAD (Too strict)
print(f"Is exact Animal? {type(d) is Animal}") # False

# GOOD (Supports inheritance)
print(f"Is instance of Animal? {isinstance(d, Animal)}") # True
```

## 2. Checking Multiple Types

You can pass a tuple to `isinstance`.

```python
val = 10
if isinstance(val, (int, float)):
    print("It is a number")
```

## 3. Abstract Base Classes (ABCs)

Instead of checking if something is a `list`, check if it is a `Sequence`. This works for custom list-like objects too.

```python
from collections.abc import Sequence

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)

print(isinstance(my_list, Sequence))  # True
print(isinstance(my_tuple, Sequence)) # True
```
