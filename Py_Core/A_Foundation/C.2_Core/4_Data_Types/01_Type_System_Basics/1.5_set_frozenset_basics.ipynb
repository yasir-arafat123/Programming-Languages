# Sets and Frozensets

Unordered collections of unique elements.

## 🎯 Learning Objectives

- **Properties**: Unique, Unordered, Fast Lookup (O(1)).
- **Operations**: Union, Intersection, Difference.
- **Frozenset**: The immutable set (can be used as a dict key).
- **Hashability**: Why you can't put a list in a set.

## 1. Basic Operations

Sets automatically remove duplicates.

```python
s = {1, 2, 2, 3}
print(s)  # {1, 2, 3}

s.add(4)
s.remove(1) # Raises KeyError if missing (use discard to be safe)
print(s)
```

## 2. Set Theory Math

Sets support powerful operator overloading.

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(f"Union (|): {a | b}") # {1, 2, 3, 4, 5}
print(f"Intersection (&): {a & b}") # {3}
print(f"Difference (-): {a - b}") # {1, 2}
print(f"Symmetric Diff (^): {a ^ b}") # {1, 2, 4, 5}
```

## 3. The Requirement: Hashability

Set elements must be hashable (immutable).

```python
try:
    invalid_set = {[1, 2], 3} # List is unhashable
except TypeError as e:
    print(f"Error: {e}")
```

## 4. Frozenset

If you need a set inside a set, or a set as a dict key, use `frozenset`.

```python
fs = frozenset([1, 2, 3])
# fs.add(4) # AttributeError

meta_set = {fs, frozenset([4, 5])}
print(meta_set)
```
