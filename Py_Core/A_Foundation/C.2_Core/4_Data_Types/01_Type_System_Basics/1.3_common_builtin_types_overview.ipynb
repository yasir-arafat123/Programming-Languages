The standard toolbox.

## 🎯 Learning Objectives

- **Scalars**: `int`, `float`, `bool`, `NoneType`.

- **Mutable vs Immutable**: The Golden Classification.

## 1. The Hierarchy

Every object in Python fits into this taxonomy.

| Category | Types | Characteristics |
| :--- | :--- | :--- |
| **Numeric** | `int`, `float`, `complex` | Mathematical numbers |
| **Sequence** | `str`, `list`, `tuple` | Ordered, indexable |
| **Map** | `dict` | Key-Value pairs |
| **Set** | `set`, `frozenset` | Unique, unordered |
| **Other** | `bool`, `NoneType` | Logic and existence |

## 2. Mutability Check

Knowing what is mutable is crucial for avoiding aliases bugs.

```python
types = {
    "int": 1,
    "str": "hello",
    "tuple": (1, 2),
    "list": [1, 2],
    "set": {1, 2},
    "dict": {"a": 1}
}

for name, obj in types.items():
    try:
        hash(obj) # Objects that are mutable generally generally unhashable
        print(f"{name}: Immutable (Safe)")
    except TypeError:
        print(f"{name}: Mutable (Careful!)")
```
