# Dynamic Typing

Python is **Dynamically Typed** and **Strongly Typed**.

## 🎯 Learning Objectives

- **Dynamic**: Types are checked at runtime, not compile time.
- **Strong**: No implicit type coercion (e.g., `"1" + 1` fails).
- **Duck Typing**: "If it walks like a duck..."
- **Protocol**: The modern way to define "Ducks".

## 1. Dynamic vs Static

In C++ (Static), a variable `int x` can only hold integers.
In Python (Dynamic), a variable `x` is just a nametag. It can point to an int, then a string.

```python
x = 10
print(f"x is {type(x)}")

x = "Hello"
print(f"x is now {type(x)}")
```

## 2. Strong vs Weak

JavaScript is **Weakly Typed**: `"1" + 1` equals `"11"` (implicit casting).
Python is **Strongly Typed**: It refuses to mix incompatible types.

```python
try:
    print("1" + 1)
except TypeError as e:
    print(f"Error: {e}")
```

## 3. Duck Typing

**"If it walks like a duck and quacks like a duck, then it must be a duck."**

In Python, we don't usually check "Is this a File?". We check "Does it have a `.read()` method?".

```python
class Duck:
    def quack(self):
        print("Quack!")

class Person:
    def quack(self):
        print("I'm imitating a duck!")

def make_it_quack(thing):
    thing.quack()  # We don't care about the type, only the method.

make_it_quack(Duck())
make_it_quack(Person())
```

## 4. Under the Hood: `__class__` Assignment

You can actually change the type of an object at runtime by assigning to `__class__`. (Dangerous magic!).

```python
class A:
    def hello(self): print("A")
class B:
    def hello(self): print("B")

obj = A()
obj.hello() # A

obj.__class__ = B
obj.hello() # B (The object mutated its type!)
```
