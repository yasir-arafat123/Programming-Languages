# Bytes, Bytearray & MemoryView

Working with raw data (images, sockets, files).

## 🎯 Learning Objectives

- **`bytes`**: Immutable sequence of numbers 0-255.
- **`bytearray`**: Mutable version of bytes.
- **`memoryview`**: Zero-copy access to binary buffers.
- **Encoding**: Converting str to bytes.

## 1. Bytes (Immutable)

Network packets and file contents are bytes, not strings.

```python
b = b"Hello"
print(f"First byte: {b[0]}") # 72 (ASCII for 'H')

# b[0] = 65  # TypeError: bytes are immutable
```

## 2. Bytearray (Mutable)

Like a list, but for bytes.

```python
ba = bytearray(b"Hello")
ba[0] = 74 # Change 'H' to 'J'
print(ba) # bytearray(b'Jello')
```

## 3. MemoryView (Zero-Copy)

The `memoryview` object allows Python code to access the internal data of an object that supports the buffer protocol without copying.

```python
data = bytearray(b"X" * 1000000) # Big array

# Slice with copy (Slow)
part = data[0:5] 

# Slice with memoryview (Fast - just a pointer)
mv = memoryview(data)
part_view = mv[0:5]
part_view[0] = 65 # Changes the ORIGINAL data!

print(data[:5]) # b'AXXXX'
```
