# Math Constants

The universe's numbers.

## 🎯 Learning Objectives
- `math.pi`, `math.e`, `math.tau`.
- `math.inf` (Infinity).
- `math.nan` (Not a Number).

## 1. Constants

```python
import math

print(math.pi)   # 3.1415...
print(math.e)    # 2.718...
print(math.tau)  # 6.28... (2 * pi)
```

## 2. Infinity

Useful for initial comparison comparisons.

```python
max_val = -math.inf
min_val = math.inf

print(100 > -math.inf)  # True
```

## 3. NaN

NaN != NaN. It poisons comparisons.

```python
nan = math.nan
print(nan == nan)  # False!
print(math.isnan(nan))  # True
```
