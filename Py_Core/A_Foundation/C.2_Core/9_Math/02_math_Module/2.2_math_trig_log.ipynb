# Trig & Logarithms

Standard scientific calculator functions.

## 🎯 Learning Objectives
- Trig (sin, cos, tan) works in RADIANS.
- Logarithms (ln, log10).

## 1. Trigonometry

Always convert degrees to radians first.

```python
import math

deg = 180
rad = math.radians(deg)  # 3.1415...
print(math.sin(rad))     # 0.0 (approximately)
```

## 2. Logarithms

- `math.log(x)`: Natural log (base e).
- `math.log(x, base)`: Custom base.
- `math.log10(x)`: Base 10.

```python
print(math.log(math.e))  # 1.0
print(math.log10(100))   # 2.0
```
