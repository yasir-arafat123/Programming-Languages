# Floor & Ceil

Forcing direction.

## 🎯 Learning Objectives
- `math.floor`.
- `math.ceil`.
- `math.trunc`.

## 1. Floor & Ceil

```python
import math

print(math.floor(2.9))  # 2
print(math.ceil(2.1))   # 3
```

## 2. Truncate vs Floor

For positive numbers, they are the same.
For negative numbers, they differ.

```python
print(math.floor(-2.5))  # -3 (Down towards negative infinity)
print(math.trunc(-2.5))  # -2 (Towards zero)
```
