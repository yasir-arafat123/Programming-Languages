# Aggregations (sum, min, max)

Processing iterables.

## 🎯 Learning Objectives
- Basic usage.
- Using `key` argument for `min`/`max`.
- `sum` start value.

## 1. Basics

```python
nums = [1, 2, 3, 4]
print(sum(nums))  # 10
print(min(nums))  # 1
print(max(nums))  # 4
```

## 2. The `key` Argument

Find max based on a specific logic.

```python
words = ["apple", "banana", "pear"]
# Which word is longest?
longest = max(words, key=len)
print(longest)  # "banana"
```

## 3. Summing non-zero start

```python
print(sum([1, 2, 3], start=10))  # 16
```
