# Built-in Math Functions

Math without imports.

## 🎯 Learning Objectives
- `abs()`: Absolute value.
- `round()`: Rounding rules.
- `pow()`: Exponentiation (regular vs modular).

## 1. abs()

Distance from zero.

```python
print(abs(-5))   # 5
print(abs(3+4j)) # 5.0 (Magnitude)
```

## 2. pow(base, exp, mod)

Standard power:
```python
print(pow(2, 3))  # 8 (Same as 2**3)
```

Modular power (Efficiency hack for crypto):
```python
print(pow(2, 3, 5))  # (2**3) % 5 = 8 % 5 = 3
# Much faster than (2**3) % 5 for huge numbers
```

## 3. round()

Rounds to nearest even number for .5 cases.

```python
print(round(2.5))  # 2
print(round(3.5))  # 4
print(round(3.14159, 2))  # 3.14
```
