# Divmod & Modulo (Refresher)

Working with remainders.

## 🎯 Learning Objectives
- `divmod()` utility.
- Use cases (Time conversion).

## 1. divmod(x, y)

Returns `(x // y, x % y)`.

```python
q, r = divmod(10, 3)
print(q)  # 3
print(r)  # 1
```

## 2. Practical: Seconds to Minutes

```python
total_seconds = 125
minutes, seconds = divmod(total_seconds, 60)
print(f"{minutes}m {seconds}s")  # 2m 5s
```
