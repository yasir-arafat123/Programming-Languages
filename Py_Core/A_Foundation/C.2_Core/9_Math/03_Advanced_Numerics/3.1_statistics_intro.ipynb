# Statistics Intro

Basic aggregation using the `statistics` module.

## 🎯 Learning Objectives
- `mean` (Average).
- `median` (Middle).
- `mode` (Most common).

## 1. Usage

```python
import statistics

data = [1, 2, 2, 3, 4, 10]
print(statistics.mean(data))    # 3.66...
print(statistics.median(data))  # 2.5
print(statistics.mode(data))    # 2
```

## 2. Robustness

The `statistics` module handles integers, floats, decimals, and fractions correctly.
