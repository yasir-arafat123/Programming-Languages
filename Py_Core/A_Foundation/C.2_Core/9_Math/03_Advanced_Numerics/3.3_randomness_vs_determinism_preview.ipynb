# Randomness vs Determinism

Generating pseudo-random numbers.

## 🎯 Learning Objectives
- `random` module.
- Seeding for reproducibility.
- `secrets` module (Crypto safe).

## 1. Basic Random

```python
import random

print(random.randint(1, 10))  # 1 to 10 (inclusive)
print(random.choice(['win', 'lose']))
```

## 2. Reproducibility

Setting the seed ensures the same sequence every time (good for tests).

```python
random.seed(42)
print(random.random())  # Always 0.639...
```

## 3. Cryptography

Never use `random` for passwords. Use `secrets`.

```python
import secrets
token = secrets.token_hex(16)
```
