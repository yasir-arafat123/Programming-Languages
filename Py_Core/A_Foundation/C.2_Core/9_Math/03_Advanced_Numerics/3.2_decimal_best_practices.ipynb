## 1. Quantizing (Money)

```python
from decimal import Decimal, ROUND_HALF_UP

price = Decimal('10.235')
# Round to 2 decimal places
cost = price.quantize(Decimal('0.01'), rounding=ROUND_HALF_UP)
print(cost)  # 10.24
```

## 2. Serializing

Always convert Decimial to string before sending to JSON APIs, as standard JSON doesn't support generic high-precision numbers.
