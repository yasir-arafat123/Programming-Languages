# The Math Module

The standard library for numerics.

## 🎯 Learning Objectives

- **Constants**: `pi`, `e`, `tau`.
- **Functions**: `sqrt`, `log`, `sin`, `cos`.
- **Combinatorics**: `comb`, `perm`.

## 1. Constants

More precision than you can memorize.

```
import math
print(f"Pi: {math.pi}")
print(f"Euler: {math.e}")
```

## 2. Logarithms

Be careful: `log(x)` is base **e** (natural log). Use `log10` for common log.

```
print(math.log(math.e)) # 1.0
print(math.log10(100))  # 2.0
```
