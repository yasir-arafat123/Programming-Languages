# Big Integers

When 64 bits aren't enough.

## 🎯 Learning Objectives

- **Limit**: The 4300-digit conversion limit (ExDoS).
- **Performance**: Time complexity of operations.
- **Applications**: RSA Encryption.

## 1. The Integer String Conversion Limit

To prevent Denial of Service attacks, modern Python limits `str(int)` to 4300 digits by default.

```
import sys
print(f"Limit: {sys.get_int_max_str_digits()}")

# You can raise it if needed
# sys.set_int_max_str_digits(10000)
```

## 2. Speed Test

Multiplication is efficient (Karatsuba algorithm).

```
import time

a = 10**1000
start = time.time()
b = a * a
end = time.time()

print(f"Squared a 1000-digit number in {end - start:.6f} seconds")
```
