# `cmath` Module

Functions for complex numbers.

## 🎯 Learning Objectives
- `math` vs `cmath`.
- Square roots of negative numbers.
- Polar coordinates.

## 1. Domain Errors

Standard `math` fails on complex domains.

```python
import math
# math.sqrt(-1)  # ValueError
```

## 2. Using cmath

```python
import cmath
print(cmath.sqrt(-1))  # 1j
```

## 3. Polar Coordinates

```python
z = 1 + 1j
r, phi = cmath.polar(z)
print(f"Radius: {r}, Angle: {phi}")
```
