# The CMath Module

Math for complex numbers.

## 🎯 Learning Objectives

- **Usage**: Functions that return complex results from real inputs.
- **Domain**: Handling negative roots.

## 1. Square Root of Negative

`math.sqrt(-1)` raises ValueError. `cmath.sqrt(-1)` returns `1j`.

```
import math
import cmath

try:
    math.sqrt(-1)
except ValueError:
    print("Math module failed (expected)")

print(f"CMath result: {cmath.sqrt(-1)}")
```

## 2. Phase and Polar Coords

Converting between rectangular (x, y) and polar (r, phi).

```
z = 1 + 1j
print(cmath.polar(z))
```
