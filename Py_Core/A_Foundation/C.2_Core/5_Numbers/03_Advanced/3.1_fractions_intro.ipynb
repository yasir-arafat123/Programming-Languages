# Fractions Intro

Exact rational numbers.

## 🎯 Learning Objectives
- `fractions.Fraction` class.
- Automatic simplification.
- Mixed arithmetic.

## 1. Usage

```python
from fractions import Fraction

f = Fraction(3, 4)
print(f + 1)      # Fraction(7, 4)
print(f + 0.25)   # 1.0 (Converts to float if mixed)
```

## 2. From String

```python
f = Fraction('1/3')
print(f * 3)  # Fraction(1, 1)
```
