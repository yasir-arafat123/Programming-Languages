# Random vs Secrets

Predictable vs Unpredictable randomness.

## 🎯 Learning Objectives

- **`random`**: Pseudo-random (Mersenne Twister). Fast, reproducible (via seed), **INSECURE**.
- **`secrets`**: Cryptographically strong. Slow, secure.
- **Usage**: When to use which.

## 1. Random (Simulation/Games)

Use this for shuffling cards in a casual game or statistical sampling.

```
import random

random.seed(42)  # Make it reproducible
print(random.random())
print(random.choice(["A", "B", "C"]))

# WARNING: Never use this for passwords!
```

## 2. Secrets (Security)

Use this for tokens, passwords, and keys.

```
import secrets

token = secrets.token_hex(16)
print(f"Secure Token: {token}")
```
