# Float Precision

Why `0.1 + 0.2 != 0.3`.

## 🎯 Learning Objectives
- Binary representation of decimals.
- The machine epsilon.
- Comparison techniques (`math.isclose`).

## 1. The Problem

Computers use binary (base-2). Some decimals (like 1/10) repeat endlessly in binary (like 1/3 in base-10).

```python
print(0.1 + 0.2)  # 0.30000000000000004
print(0.1 + 0.2 == 0.3)  # False!
```

## 2. The Solution

Never compare floats with `==`. Use a tolerance (epsilon).

```python
import math

a = 0.1 + 0.2
b = 0.3
print(math.isclose(a, b))  # True
```
