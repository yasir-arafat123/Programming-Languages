# Decimal Intro

For when money matters.

## 🎯 Learning Objectives
- `decimal.Decimal` class.
- Precision context.
- Use strings, not floats, for initialization.

## 1. Why Decimal?

Fixed-point and floating-point arithmetic with user-defined precision.

```python
from decimal import Decimal

a = Decimal('0.1')
b = Decimal('0.2')
print(a + b)  # Decimal('0.3')
```

## 2. Initialization Trap

Don't initialize with floats!

```python
# BAD
d = Decimal(0.1)
print(d)  # 0.10000000000000000555...

# GOOD
d = Decimal('0.1')
print(d)  # 0.1
```
