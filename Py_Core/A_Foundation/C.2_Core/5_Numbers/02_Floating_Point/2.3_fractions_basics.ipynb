# The Fraction Type

Rational numbers (Numerator / Denominator).

## 🎯 Learning Objectives

- **Usage**: Exact representation of $1/3$.
- **Reduction**: Automatic GCD simplification.
- **Conversion**: From float.

## 1. Creating Fractions

Integers, strings, or other fractions.

```
from fractions import Fraction

half = Fraction(1, 2)
quarter = Fraction("1/4")

print(half + quarter) # 3/4
```

## 2. Automatic Reduction

Python automatically finds the Greatest Common Divisor.

```
print(Fraction(16, 32)) # 1/2
```

## 3. Limit Denominator

Useful for approximating PI.

```
import math
pi = Fraction(math.pi).limit_denominator(100)
print(pi) # 22/7 (Classic approximation)
```
