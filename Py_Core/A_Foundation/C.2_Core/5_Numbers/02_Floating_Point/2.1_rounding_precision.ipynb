# Rounding Numbers

It's not as simple as "0.5 rounds up".

## 🎯 Learning Objectives

- **`round()`**: Python uses "Round Half to Even" (Banker's Rounding).
- **Floor/Ceil**: `math.floor()`, `math.ceil()`.
- **Truncate**: `math.trunc()` vs `int()`.

## 1. Banker's Rounding

To minimize error accumulation in large sums, Python rounds x.5 to the nearest **even** number.

```
print(round(0.5)) # 0
print(round(1.5)) # 2
print(round(2.5)) # 2 (Surprise!)
print(round(3.5)) # 4
```

## 2. Explicit Floor/Ceil

If you need "Round Up" behavior, use `math`.

```
import math

print(math.floor(2.9)) # 2
print(math.ceil(2.1))  # 3
```

## 3. Formatting Rounding

String formatting `:.2f` rounds using standard rules usually, but is display-only.
