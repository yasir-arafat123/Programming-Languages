# Rounding Gotchas

Python doesn't always round .5 up.

## 🎯 Learning Objectives
- `round()` behavior.
- Bankers' Rounding (Round half to even).
- `math.ceil` vs `math.floor`.

## 1. Bankers' Rounding

To avoid bias, Python rounds `.5` to the nearest **even** number.

```python
print(round(0.5))  # 0
print(round(1.5))  # 2
print(round(2.5))  # 2
print(round(3.5))  # 4
```

This is standard in IEEE 754.

## 2. Forcing Direction

Use `math` module.

```python
import math

print(math.floor(2.9))  # 2
print(math.ceil(2.1))   # 3
```
