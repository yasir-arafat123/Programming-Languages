# The Decimal Type

For when money is involved.

## 🎯 Learning Objectives

- **Why**: Avoiding floating point errors (`0.1 + 0.2`).
- **Initialization**: Always pass strings, not floats!
- **Precision**: Setting global context precision.

## 1. The Float Error Revisited

Using floats for money results in missing pennies.

```
price = 19.99
qty = 3
print(price * qty) # 59.970000000000006
```

## 2. Using Decimal

Import from `decimal`. **Crucial Rule**: Pass exact strings.

```
from decimal import Decimal, getcontext

# BAD
d_bad = Decimal(0.1)
print(f"Initialized with float: {d_bad}")

# GOOD
d_a = Decimal("0.1")
d_b = Decimal("0.2")
print(f"Initialized with string: {d_a + d_b}") # 0.3
```

## 3. Controlling Context

You can decide how many significant digits you want globally.

```
getcontext().prec = 6
print(Decimal("1") / Decimal("7"))
```
