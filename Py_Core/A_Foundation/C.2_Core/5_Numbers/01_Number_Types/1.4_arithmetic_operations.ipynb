# Arithmetic Operations

Math operators and their nuances.

## 🎯 Learning Objectives

- **Division**: True (`/`) vs Floor (`//`).
- **Modulo**: `%` and how it handles negatives.
- **Power**: `**` vs `pow()`.
- **Precedence**: PEMDAS rules.

## 1. The Two Divisions

In Python 3, `/` always returns a float (True Division).

```
print(10 / 2)   # 5.0 (Float)
print(10 // 3)  # 3 (Floor Integer)
print(-10 // 3) # -4 (Floors towards negative infinity!)
```

## 2. Modulo (%)

Remainder of division. Useful for checking even/odd or cycling indices.

```
print(10 % 3) # 1

# Negative modulo (follows the sign of divisor in Python)
print(-10 % 3) # 2 (Wait, why? Because -10 = 3*(-4) + 2)
```

## 3. Power

`pow(base, exp, mod)` is more efficient for modular exponentiation (cryptography).

```
print(2 ** 3)
print(pow(2, 3, 5)) # (2^3) % 5 -> 8 % 5 -> 3
```
