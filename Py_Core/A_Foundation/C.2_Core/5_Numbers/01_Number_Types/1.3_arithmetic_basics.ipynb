# Arithmetic Basics

The standard math operators.

## 🎯 Learning Objectives
- `+`, `-`, `*`.
- `/` (True division).
- `**` (Exponentiation).

## 1. The Basics

```python
print(2 + 2)  # 4
print(5 - 3)  # 2
print(4 * 5)  # 20
```

## 2. True Division (`/`)

Always returns a `float`, even if clean.

```python
print(10 / 2)  # 5.0 (Not 5)
print(10 / 3)  # 3.333...
```

## 3. Exponentiation (`**`)

```python
print(2 ** 3)  # 8
print(9 ** 0.5)  # 3.0 (Square root)
```
