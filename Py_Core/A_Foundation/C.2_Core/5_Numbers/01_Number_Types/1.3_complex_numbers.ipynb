# Complex Numbers

Python has native support for $a + bi$.

## 🎯 Learning Objectives

- **Syntax**: Using `j` instead of `i`.
- **Attributes**: `.real` and `.imag`.
- **Conjugate**: `.conjugate()` method.
- **Applications**: Electrical engineering, Signal processing.

## 1. Syntax

Use the suffix `j` or `J` for the imaginary part.

```
z = 3 + 4j
print(f"z: {z}")
print(f"Type: {type(z)}")
```

## 2. Operations

Math works as expected.

```
z1 = 1 + 1j
z2 = 1 - 1j

print(f"Add: {z1 + z2}")
print(f"Multiply: {z1 * z2}")
print(f"Magnitude (abs): {abs(z1)}") # sqrt(1^2 + 1^2)
```

## 3. Attributes

Access components directly.

```
print(z.real)
print(z.imag)
print(z.conjugate()) # 3 - 4j
```
