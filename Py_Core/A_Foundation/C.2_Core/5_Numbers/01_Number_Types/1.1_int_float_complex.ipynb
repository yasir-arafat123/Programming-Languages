# Number Types

Python's numeric hierarchy.

## 🎯 Learning Objectives
- `int` (Arbitrary precision).
- `float` (Double precision).
- `complex` (Built-in support).

## 1. Integers (int)

Unlimited size (only limited by RAM).

```python
x = 10
huge = 10**100  # A googol
print(type(x))  # <class 'int'>
```

## 2. Floating Point (float)

Equivalent to C's `double` (64-bit).

```python
pi = 3.14159
science = 1.2e-10
```

## 3. Complex Numbers (complex)

Built right into the language syntax using `j`.

```python
z = 1 + 2j
print(z.real)  # 1.0
print(z.imag)  # 2.0
```
