# Floating Point Numbers

Real numbers (approximations).

## 🎯 Learning Objectives

- **IEEE 754**: The standard for representating reals in binary.
- **The Glitch**: Why `0.1 + 0.2 != 0.3`.
- **Equality**: Using `math.isclose()`.
- **Special Values**: `nan`, `inf`.

## 1. The notorious 0.3 bug

Computers store floats in base-2 (binary). Just like 1/3 is repeating in base-10 (0.333...), 0.1 is repeating in base-2.

```
print(0.1 + 0.2)
print(0.1 + 0.2 == 0.3) # False!
```

## 2. Correct Comparison

Never compare floats with `==`. Use a tolerance (epsilon).

```
import math

a = 0.1 + 0.2
b = 0.3

print(math.isclose(a, b)) # True
```

## 3. Special Values

Infinity and Not A Number.

```
inf = float('inf')
nan = float('nan')

print(inf > 10**100) # True
print(nan == nan)    # False! By definition, NaN is not equal to anything, even itself.
```
