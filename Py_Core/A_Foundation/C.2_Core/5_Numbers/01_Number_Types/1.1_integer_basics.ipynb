# Integers

Whole numbers with superpowers.

## 🎯 Learning Objectives

- **Arbitrary Precision**: Why Python ints never overflow.
- **Bases**: Binary (`0b`), Octal (`0o`), Hex (`0x`).
- **Internals**: How Python stores big numbers.
- **Underscores**: `1_000_000` syntax.

## 1. Arbitrary Precision

In C/Java, an integer is usually 32 or 64 bits. If you exceed 2^63, it overflows (wraps around). 
In Python, an integer expands to use as much memory as needed.

```
import sys

# A normal int
x = 2**64
print(f"2^64: {x}")
print(f"Size in bytes: {sys.getsizeof(x)}")

# A massive int
y = 2**1000
print(f"2^1000 has {len(str(y))} digits")
print(f"Size in bytes: {sys.getsizeof(y)}")
```

## 2. Bases and Literals

You can write integers in different bases.

```
print(0b1010)   # Binary (10)
print(0o12)     # Octal (10)
print(0xA)      # Hex (10)

# Converting back
n = 255
print(bin(n))
print(hex(n))
```

## 3. Underscores for Readability

You can put underscores anywhere in a number literal (except start/end) to group digits. Python ignores them.

```
salary = 1_000_000
hex_addr = 0xDEAD_BEEF
binary_flags = 0b1111_0000

print(salary)
```
