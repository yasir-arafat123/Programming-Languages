# Numeric Literals

Different ways to write numbers.

## 🎯 Learning Objectives
- Underscores for readability (`1_000`).
- Bases: Hex, Octal, Binary.
- Scientific notation.

## 1. Underscores

Visual separator only. Python ignores them.

```python
billion = 1_000_000_000
print(billion)  # 1000000000
```

## 2. Bases

```python
binary = 0b1010  # 10
octal = 0o10     # 8
hex = 0xFF       # 255
```

## 3. Scientific Notation

```python
electron_mass = 9.109e-31
avogadro = 6.022e23
```
