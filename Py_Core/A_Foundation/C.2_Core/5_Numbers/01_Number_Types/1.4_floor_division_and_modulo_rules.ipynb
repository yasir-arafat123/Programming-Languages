# Floor Division & Modulo

Integer math operations.

## 🎯 Learning Objectives
- Floor division `//`.
- Modulo `%`.
- Relationships (`divmod`).

## 1. Floor Division (`//`)

Rounds **down** to the nearest integer.

```python
print(10 // 3)   # 3
print(-10 // 3)  # -4 (Not -3! Rounds towards negative infinity)
```

## 2. Modulo (`%`)

Returns the remainder.

```python
print(10 % 3)  # 1
```

Useful for checking even/odd:
```python
is_even = (x % 2 == 0)
```

## 3. divmod()

Calculates both at once.

```python
q, r = divmod(10, 3)
print(q, r)  # 3, 1
```
