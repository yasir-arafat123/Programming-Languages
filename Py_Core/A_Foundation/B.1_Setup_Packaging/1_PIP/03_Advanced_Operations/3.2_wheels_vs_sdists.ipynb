# Wheels vs Source Distributions

Not all packages install the same way. Understanding the difference saves debugging time.

## 🎯 Learning Objectives
- Understand `.whl` (wheel) files.
- Understand `.tar.gz` (sdist) files.
- Know when compilation is required.

## 1. Wheel (.whl) - Precompiled

A **wheel** is a ZIP file containing ready-to-use Python code.

- ✅ Fast installation (just unzip)
- ✅ No compiler needed
- ❌ Platform-specific (Windows wheel ≠ Linux wheel)

## 2. Source Distribution (sdist) - Raw Code

A **sdist** is a `.tar.gz` containing the original source code.

- ❌ Slow installation (must build from scratch)
- ❌ Requires C compiler for packages with C extensions
- ✅ Works on any platform

## 3. What PIP Does

When you run `pip install numpy`:

1. PIP looks for a **wheel** matching your Python version + OS.
2. If found, downloads and unzips it (fast).
3. If NOT found, downloads the **sdist** and compiles it (slow).

## 4. Checking What You Downloaded

````
# When installing, PIP shows:
# "Downloading numpy-1.24.0-cp312-cp312-win_amd64.whl"  ← Wheel
# vs
# "Building wheel for numpy (setup.py)"  ← Sdist (had to build)
````

## 5. Forcing Wheel-Only Installs

```bash
pip install --only-binary :all: numpy
```

This **fails** if no wheel exists (useful for CI/CD to avoid long builds).
