# Dependency Resolution & Conflicts

Understanding how PIP resolves complex dependency graphs prevents hours of debugging.

## 🎯 Learning Objectives
- Understand the new backtracking resolver (PIP 20.3+).
- Diagnose dependency conflicts.
- Use `pipdeptree` for visualization.

## 1. The Dependency Hell

You want:
- `packageA` requires `requests>=2.0,<3.0`
- `packageB` requires `requests>=3.0`

**Impossible to satisfy both!**

## 2. The Modern Resolver

Since PIP 20.3 (late 2020), PIP uses a **backtracking resolver**.

It:
1. Reads ALL requirements upfront.
2. Tries combinations.
3. **Fails fast** with clear error messages if conflict detected.

## 3. Error Example

```
ERROR: Cannot install packageA and packageB because these package versions have conflicting dependencies.

The conflict is caused by:
    packageA depends on requests<3.0
    packageB depends on requests>=3.0
```

## 4. Visualizing Dependencies

Install `pipdeptree`:

````
# %pip install pipdeptree
# !pipdeptree
````

Output shows a tree:
```
requests==2.28.0
  ├── certifi [required: >=2017.4.17]
  ├── charset-normalizer [required: <4,>=2]
  └── urllib3 [required: <1.27,>=1.21.1]
```

## 5. Resolving Conflicts

Options:
1. **Upgrade** one of the conflicting packages.
2. **Downgrade** to an older compatible version.
3. **Remove** the conflicting package (if not critical).
4. **Open an issue** asking maintainers to loosen version constraints.
