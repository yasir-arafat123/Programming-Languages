# Private Indexes & Trusted Hosts

Corporate environments often host their own package mirrors or private packages.

## 🎯 Learning Objectives
- Use custom package indexes.
- Bypass SSL verification (carefully!).
-Combine multiple indexes.

## 1. Custom Index URL

```bash
pip install --index-url https://pypi.company.com/simple/ mypackage
```

This tells PIP to look at `company.com` instead of `pypi.org`.

## 2. Multiple Indexes (Extra Index)

```bash
pip install --extra-index-url https://pypi.company.com/simple/ mypackage
```

This checks BOTH `pypi.org` AND `company.com`.

## 3. Trusted Hosts (Skip SSL)

If your company mirror has a self-signed certificate:

```bash
pip install --trusted-host pypi.company.com mypackage
```

⚠️ **Warning**: Only use this on internal networks you trust!

## 4. Permanent Configuration

In `pip.conf`:
```ini
[global]
index-url = https://pypi.company.com/simple
extra-index-url = https://pypi.org/simple
trusted-host = pypi.company.com
```
