# Hashes & Package Integrity

Prevent supply-chain attacks by verifying package checksums.

## 🎯 Learning Objectives
- Generate hash-pinned requirements.
- Use `--require-hashes`.
- Understand the security model.

## 1. The Threat

Scenario: An

 attacker compromises PyPI and replaces `requests==2.28.0` with malicious code.

Your `requirements.txt`:
```text
requests==2.28.0
```
...will install the **poisoned** version because you ONLY verified version, not content.

## 2. Solution: Hash Pinning

```bash
pip install --require-hashes -r requirements.txt
```

This forces PIP to verify the SHA256 hash of every file.

## 3. Generating Hash-Pinned Requirements

```bash
pip hash requests==2.28.0
```

Output:
```
requests==2.28.0 \
    --hash=sha256:abc123...
```

## 4. Example Secure requirements.txt

```text
requests==2.28.0 \
    --hash=sha256:64763187c1c05e2b0aa6fa6d64b8e5891d6d2d1e...
certifi==2022.12.7 \
    --hash=sha256:35824b4c3a97115964b408844d64aa14db1cc518f6ce72329d0b22f9c1e8e8fe
```

## 5. Automation with Tools

Manually maintaining hashes is painful. Use tools like:
- **hashin** (`pip install hashin`)
- **pip-tools** (covered later)
