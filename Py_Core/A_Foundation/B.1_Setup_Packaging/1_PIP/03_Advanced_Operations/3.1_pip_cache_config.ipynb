# PIP Cache & Configuration

Speed up installs and customize PIP's behavior.

## 🎯 Learning Objectives
- Understand the PIP cache directory.
- Configure PIP globally via `pip.conf`.
- Clear cache when troubleshooting.

## 1. The PIP Cache

PIP caches downloaded packages to avoid re-downloading.

````
# Show cache location
!pip cache dir
````

Typical locations:
- **Windows**: `C:\Users\<name>\AppData\Local\pip\Cache`
- **macOS**: `~/Library/Caches/pip`
-Linux**: `~/.cache/pip`

## 2. Cache Commands

````
# List cached files
!pip cache list

# Show cache info
!pip cache info

# Remove all cache
# !pip cache purge
````

## 3. Configuration Files

PIP reads settings from `pip.conf` (Linux/Mac) or `pip.ini` (Windows).

**Locations** (in priority order):
1. **Per-virtualenv**: `$VIRTUAL_ENV/pip.conf`
2. **User**: `~/.config/pip/pip.conf` (Linux), `~/Library/Application Support/pip/pip.conf` (Mac)
3. **Global**: `/etc/pip.conf`

## 4. Example Configuration

```ini
[global]
timeout = 60
index-url = https://pypi.org/simple

[install]
no-cache-dir = false
```

## 5. View Current Configuration

````
!pip config list
````
