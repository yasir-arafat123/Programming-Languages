# Editable Installs (Development Mode)

When developing a package, you don't want to reinstall it after every code change. Editable installs solve this.

## 🎯 Learning Objectives
- Install packages in development mode.
- Understand how `-e` works.
- Know when to use it.

## 1. The Problem

You're building `mypackage`. Normal workflow:
1. Edit code in `/home/user/mypackage/`
2. Run `pip install .` to test
3. Make another change
4. Run `pip install .` **again** 😫

## 2. The Solution: Editable Install

```bash
pip install -e .
```

Or with path:
```bash
pip install -e /path/to/mypackage
```

## 3. How It Works

Instead of copying files to `site-packages/`, PIP creates a **symlink** (or `.pth` file) pointing to your source directory.

Changes in your source code are **immediately** reflected when you `import mypackage`.

## 4. Verifying Editable Install

````
# After pip install -e .
!pip list
# You'll see:
# mypackage    0.1.0    /home/user/projects/mypackage
````

## 5. Common in requirements-dev.txt

```text
# requirements-dev.txt
-e .
pytest
black
mypy
```

This installs your package in editable mode + dev tools.
