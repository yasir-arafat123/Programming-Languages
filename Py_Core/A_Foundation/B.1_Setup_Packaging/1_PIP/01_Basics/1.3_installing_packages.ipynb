# Installing Packages

The bread and butter of PIP: getting code from the internet into your project.

## 🎯 Learning Objectives
- Install packages from PyPI.
- Install from Git repositories.
- Install from local directories.

## 1. Basic Installation

```python
# Install a package
%pip install requests
```

### What Just Happened?

1. PIP connected to PyPI.
2. Downloaded `requests` and its dependencies (`urllib3`, `certifi`, etc.).
3. Installed them to `site-packages/`.

## 2. Installing Specific Versions

```python
# Install exact version
%pip install requests==2.28.0

# Install minimum version
%pip install "requests>=2.28.0"

# Install with upper bound
%pip install "requests>=2.28.0,<3.0.0"
```

## 3. Installing from Git

```python
# Install from GitHub (bleeding edge)
# %pip install git+https://github.com/psf/requests.git
```

## 4. Installing Multiple Packages

```python
# Space-separated
%pip install requests beautifulsoup4 pandas
```

## 5. Verifying Installation

```python
import requests
print(f"requests version: {requests.__version__}")
print(f"Installed at: {requests.__file__}")
```
