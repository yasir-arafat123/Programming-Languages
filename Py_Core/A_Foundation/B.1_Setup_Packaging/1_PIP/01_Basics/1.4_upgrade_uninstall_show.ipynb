# Upgrade, Uninstall & Show

Managing installed packages: updating, removing, and inspecting.

## 🎯 Learning Objectives
- Upgrade packages safely.
- Uninstall packages and dependencies.
- Inspect package metadata.

## 1. Listing Installed Packages

```python
!pip list
```

## 2. Showing Package Details

```python
!pip show requests
```

This displays:
- Version
- Location (where it's installed)
- Dependencies (`Requires:`)
- What depends on it (`Required-by:`)

## 3. Upgrading a Package

```python
%pip install --upgrade requests
```

**⚠️ Warning**: Upgrading one package might break another if they have conflicting requirements.

## 4. Checking for Outdated Packages

```python
!pip list --outdated
```

## 5. Uninstalling a Package

```python
# pip uninstall requests -y
# The -y flag skips confirmation
```

**Important**: Uninstalling a package **does NOT** uninstall its dependencies.

If you installed `requests` (which pulled in `urllib3` and `certifi`), uninstalling `requests` leaves those orphaned.
