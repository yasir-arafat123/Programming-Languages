# Troubleshooting Common PIP Errors

When PIP breaks, these are the top 5 culprits.

## 🎯 Learning Objectives
- Fix permission errors.
- Resolve SSL certificate issues.
- Debug network/proxy problems.

## 1. Permission Denied

**Error**:
```
ERROR: Could not install packages due to an OSError: [Errno 13] Permission denied
```

**❌ Wrong Fix**: `sudo pip install`

**✅ Right Fix**: Use a virtual environment OR:
```bash
pip install --user package_name
```

## 2. SSL Certificate Verify Failed

**Error**:
```
SSL: CERTIFICATE_VERIFY_FAILED
```

**Causes**:
- Corporate firewall (MITM proxy)
- Outdated CA certificates

**Fix**:
```bash
pip install --trusted-host pypi.org --trusted-host files.pythonhosted.org package_name
```

Or upgrade certifi:
```bash
pip install --upgrade cert ifi
```

## 3. Network Timeout

**Error**:
```
ReadTimeoutError: HTTPSConnectionPool(host='pypi.org', port=443)
```

**Fix**:
```bash
pip install --timeout=1000 package_name
```

## 4. No Matching Distribution Found

**Error**:
```
ERROR: Could not find a version that satisfies the requirement packagename
```

**Causes**:
- Typo in package name
- Package doesn't support your Python version
-  Package doesn't support your OS

**Fix**: Check [pypi.org](https://pypi.org/) for correct name and compatibility.

## 5. Legacy Resolver Deprecated

**Error**:
```
DEPRECATION: pip's legacy dependency resolver...
```

**Fix**: Upgrade PIP:
```bash
python -m pip install --upgrade pip
```
