# PIP Best Practices

Production-grade PIP usage patterns.

## 🎯 Learning Objectives
- Follow the "Golden Rules" of PIP.
- Avoid common anti-patterns.
- Establish a professional workflow.

## 1. The Golden Rules

### ✅ DO:
1. **Always use virtual environments** (covered next section).
2. **Pin versions in production** (`==` for deployed code).
3. **Use `requirements.txt`** for reproducibility.
4. **Use `python -m pip`** instead of just `pip`.
5. **Separate dev and prod dependencies**.

### ❌ DON'T:
1. **Never use `sudo pip install`**.
2. **Don't install packages globally**.
3. **Don't rely on `pip freeze` blindly** (includes transitive deps).
4. **Don't ignore version constraints**.

## 2. Dependency Files Structure

**Professional project structure**:
```
requirements/
├── base.txt          # Core dependencies
├── dev.txt           # Development tools
├── test.txt          # Testing libraries
└── prod.txt          # Production-only
```

Example `dev.txt`:
```text
-r base.txt    # Include base requirements
black
pytest
mypy
```

## 3. Version Pinning Strategy

**Development** (`requirements-dev.txt`):
```text
requests~=2.28.0    # Allow patch updates
flask>=2.3.0        # Flexible
```

**Production** (`requirements.txt`):
```text
requests==2.28.2    # Exact, frozen
flask==2.3.3
```

## 4. Upgrading Dependencies Safely

 1. Create a separate branch.
2. Update one dependency at a time.
3. Run full test suite.
4. Monitor in staging before production.

## 5. Security Scanning

Check for known vulnerabilities:
```bash
pip install safety
safety check
```

Or use GitHub Dependabot (automatic).
