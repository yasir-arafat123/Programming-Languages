# Constraints Files

Control transitive dependencies without explicitly listing them in requirements.txt.

## 🎯 Learning Objectives
- Understand the constraints file concept.
- Apply upper bounds to transitive dependencies.
- Use `-c constraints.txt`.

## 1. The Problem

Your `requirements.txt`:
```text
requests
```

But `requests` depends on `urllib3`. Tomorrow, `urllib3` might release v2.0 with breaking changes. Your code breaks, but you never mentioned `urllib3` in requirements.

**Old Solution**: Pin everything in requirements.txt (messy).

**Better**: Use constraints.

## 2. Creating constraints.txt

This file specifies version **limits** without actually installing the packages:

```text
# constraints.txt
urllib3<2.0.0
certifi>=2022.0.0
```

## 3. Using Constraints

```bash
pip install -r requirements.txt -c constraints.txt
```

PIP will:
1. Install everything from requirements.txt
2. Ensure transitive deps respect constraints.txt

## 4. Real-World Use Case

Large companies use constraints to:
- Enforce security policies ("Never use package X > 2.0")
- Prevent known-bad versions
- Share common dependency bounds across 100s of projects
