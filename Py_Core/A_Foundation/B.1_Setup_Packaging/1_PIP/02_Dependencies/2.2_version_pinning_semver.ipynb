# Version Pinning & Semantic Versioning

Not all package upgrades are safe. Understanding version constraints prevents breakage.

## 🎯 Learning Objectives
- Master version specifiers (`==`, `>=`, `~=`, `^`).
- Understand Semantic Versioning (SemVer).
- Balance stability vs freshness.

## 1. Semantic Versioning (MAJOR.MINOR.PATCH)

Example: `requests==2.28.1`

- **MAJOR (2)**: Breaking changes. Code **will** break.
- **MINOR (28)**: New features. Backward compatible.
- **PATCH (1)**: Bug fixes only. Always safe.

## 2. Version Specifiers

| Specifier | Meaning | Example |
| :--- | :--- | :--- |
| `==2.28.1` | **Exact** version only | Frozen, ultra-safe, but no bug fixes |
| `>=2.28.0` | **At least** this version | Allows upgrades (risky) |
| `>=2.28,<3.0` | Range | Will get 2.x updates, never 3.x |
| `~=2.28.0` | **Compatible release** | `>=2.28.0, <2.29.0` (patch updates only) |

## 3. The `~=` Operator (Compatible Release)

This is the **sweet spot**:
```text
requests~=2.28.0
```
Translates to: "Give me 2.28.x bug fixes, but not 2.29 or 3.0"

## 4. Real-World Example

❌ **Too Loose**:
```text
flask  # Could get Flask 4.0 tomorrow and break everything
```

❌ **Too Strict**:
```text
flask==2.3.2  # Will never get security patches
```

✅ **Just Right**:
```text
flask~=2.3.0  # Gets 2.3.x patches, not 2.4 or 3.0
```

## 5. Pre-releases

By default, PIP ignores alpha/beta/rc versions.

To install:
```bash
pip install --pre somepackage
```
