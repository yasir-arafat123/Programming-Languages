# Extras & Optional Dependencies

Not all users need all features. Extras let you install optional functionality on demand.

## 🎯 Learning Objectives
- Install packages with extras (`pip install package[dev]`).
- Understand common extras (dev, test, docs).
- Define extras in your own packages.

## 1. What are Extras?

Example: `requests` has optional `security` features:

```bash
pip install requests[security]
```

This installs `requests` PLUS additional security-related packages.

## 2. Common Extras

| Extra | Purpose | Example |
| :--- | :--- | :--- |
| `[dev]` | Development tools (linters, formatters) | `pip install myproject[dev]` |
| `[test]` | Testing libraries (pytest, coverage) | `pip install myproject[test]` |
| `[docs]` | Documentation builders (Sphinx) | `pip install myproject[docs]` |
| `[all]` | Everything | `pip install myproject[all]` |

## 3. Multiple Extras

```bash
pip install myproject[dev,test]
```

## 4. Real Example: Django

```bash
pip install django[argon2,bcrypt]
```

Installs Django + advanced password hashers.

## 5. Defining Extras (Preview)

In your `pyproject.toml`:
```toml
[project.optional-dependencies]
dev = ["black", "ruff", "mypy"]
test = ["pytest", "coverage"]
```
