# requirements.txt Basics

Hardcoding `pip install pandas numpy requests` in documentation is amateur hour. Professionals use `requirements.txt`.

## 🎯 Learning Objectives
- Create a requirements file.
- Install from requirements.txt.
- Understand pip freeze.

## 1. The Problem

Your project depends on 15 packages. Telling a colleague to run:
```bash
pip install pandas numpy requests flask sqlalchemy ...
```
...is error-prone and unmaintainable.

## 2. Creating requirements.txt

**Method A: Manual (Recommended for new projects)**

Create a file named `requirements.txt`:
```text
requests
pandas
flask
```

**Method B: Freeze (For existing environments)**

````python
!pip freeze > requirements.txt
````

This captures **every installed package** with exact versions:
```text
certifi==2024.2.2
charset-normalizer==3.3.2
idna==3.6
requests==2.31.0
urllib3==2.2.0
```

## 3. Installing from requirements.txt

````python
# Dry run (show what would be installed)
!pip install -r requirements.txt --dry-run

# Actually install
# !pip install -r requirements.txt
````

## 4. Best Practices

### Don't Mix Direct Dependencies with Transitive Ones

❌ **Bad** (from pip freeze):
```text
requests==2.31.0
urllib3==2.2.0    # This is a dependency OF requests
certifi==2024.2.2 # This too
```

✅ **Good** (manual):
```text
requests==2.31.0  # Only what YOU directly use
```

## 5. Comments in requirements.txt

```text
# Web framework
flask==3.0.0

# Data processing
pandas>=2.0.0
numpy  # Latest version
```
