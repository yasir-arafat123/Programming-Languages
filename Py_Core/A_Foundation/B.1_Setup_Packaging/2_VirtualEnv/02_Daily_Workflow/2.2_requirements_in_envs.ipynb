# Requirements Files in Virtual Environments

Combining venvs with requirements.txt for perfect reproducibility.

## 🎯 Learning Objectives
- Create clean requirements from a venv.
- Recreate a venv from requirements.
- Understand the freeze workflow.

## 1. The Standard Workflow

```bash
# 1. Create venv
python -m venv .venv

# 2. Activate
source .venv/bin/activate  # or Windows equivalent

# 3. Install packages
pip install requests pandas flask

# 4. Freeze to requirements
pip freeze > requirements.txt
```

## 2. Clean Freeze (Pro Method)

**Problem**: `pip freeze` captures EVERYTHING, including transitive dependencies.

**Solution**: Use `pipreqs` or manually maintain requirements.txt with only direct dependencies:

```text
# requirements.txt (manual, clean)
requests~=2.28.0
pandas~=2.0.0
flask~=3.0.0
```

## 3. Recreating an Environment

**On another machine** or 6 months later:

```bash
# 1. Create fresh venv
python -m venv .venv

# 2. Activate
source .venv/bin/activate

# 3. Install from requirements
pip install -r requirements.txt
```

You now have **identical** dependencies.

## 4. Development vs Production Split

```text
# requirements.txt (production)
flask==3.0.3
requests==2.31.0

# requirements-dev.txt (development)
-r requirements.txt  # Include production
pytest
black
mypy
ipython
```

Install dev deps:
```bash
pip install -r requirements-dev.txt
```

## 5. Upgrading Dependencies

```bash
# Upgrade all packages to latest compatible versions
pip install --upgrade -r requirements.txt

# Then re-freeze
pip freeze > requirements.txt
```
