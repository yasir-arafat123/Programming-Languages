# Editable Installs in Development

When building a package, install it in "development mode" so changes apply immediately.

## 🎯 Learning Objectives
- Install your own package in editable mode.
- Understand when to use `-e`.
- Set up a development environment.

## 1. The Developer's Workflow

You're creating `mypackage/`:

```
mypackage/
├── src/
│   └── mypackage/
│       └── __init__.py
├── tests/
├── pyproject.toml
└── README.md
```

**Without `-e`**: Every code change requires `pip install .` (slow!).

**With `-e`**: Code changes are live.

## 2. Installing in Editable Mode

```bash
cd mypackage/
python -m venv .venv
source .venv/bin/activate
pip install -e .
```

The `.` means "current directory".

## 3. How It Works

PIP creates a **link** (`.pth` file) in `site-packages/` pointing to your source:

```
.venv/lib/python3.12/site-packages/
└── mypackage.egg-link  → /home/user/mypackage/src
```

When you `import mypackage`, Python reads from your working directory.

## 4. With Dependencies

Your `pyproject.toml`:
```toml
[project]
name = "mypackage"
dependencies = [
    "requests",
    "pandas"
]

[project.optional-dependencies]
dev = ["pytest", "black"]
```

Install with dev deps:
```bash
pip install -e ".[dev]"
```

## 5. Typical requirements-dev.txt

```text
# requirements-dev.txt
-e .  # Editable install of this project
pytest
pytest-cov
black
mypy
ruff
```

One command setup:
```bash
pip install -r requirements-dev.txt
```
