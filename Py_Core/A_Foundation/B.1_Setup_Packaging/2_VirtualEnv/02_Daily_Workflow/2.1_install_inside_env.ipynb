# Installing Inside Environments

Common mistake: Activating venv but accidentally installing globally.

## 🎯 Learning Objectives
- Verify pip is using the venv.
- Install packages correctly.
- Avoid the "wrong pip" trap.

## 1. The Critical Check

**Before installing anything**, verify which pip you're using:

```
!which pip  # Linux/Mac
# !where pip  # Windows

# Should show: /path/to/project/.venv/bin/pip
# NOT: /usr/local/bin/pip
```

## 2. Installing Packages

With venv activated:

```
# In notebook (preferred)
%pip install requests pandas

# Or in terminal
# pip install requests pandas
```

## 3. Verification

Check where the package installed:

```
import requests
print(requests.__file__)
# Should show: /path/to/project/.venv/lib/python3.12/site-packages/requests/__init__.py
```

## 4. Common Mistake: Forgot to Activate

**Symptom**: You install a package, but `import` fails.

**Cause**: You installed to global Python, but your IDE/notebook is using the venv.

**Fix**: 
1. Activate the venv
2. Reinstall the package

## 5. Listing Installed Packages (Venv Only)

```
!pip list --local
# Shows only packages in THIS venv, not global
```
