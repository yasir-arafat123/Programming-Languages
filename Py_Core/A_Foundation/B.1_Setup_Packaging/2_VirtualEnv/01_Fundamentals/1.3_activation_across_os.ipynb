# Activation Across Operating Systems

The most common stumbling block for beginners: different activation commands per OS.

## 🎯 Learning Objectives
- Master OS-specific activation commands.
- Troubleshoot common activation errors.
- Use IDE integrations.

## 1. The Activation Matrix

| OS | Shell | Command |
| :--- | :--- | :--- |
| **Windows** | PowerShell | `.venv\Scripts\Activate.ps1` |
| **Windows** | CMD | `.venv\Scripts\activate.bat` |
| **Windows** | Git Bash | `source .venv/Scripts/activate` |
| **Linux/Mac** | Bash/Zsh | `source .venv/bin/activate` |
| **Linux/Mac** | Fish | `source .venv/bin/activate.fish` |

## 2. Windows PowerShell Execution Policy Error

**Error**:
```
Activate.ps1 cannot be loaded because running scripts is disabled on this system.
```

**Fix**:
```powershell
Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser
```

## 3. How Activation Works

The `activate` script modifies your `PATH` environment variable:

**Before**:
```
PATH=/usr/bin:/usr/local/bin
```

**After**:
```
PATH=/path/to/project/.venv/bin:/usr/bin:/usr/local/bin
```

Now typing `python` runs `.venv/bin/python` instead of `/usr/bin/python`.

## 4. IDE Integration

**VS Code**: Automatically detects `.venv` and offers to select it as the interpreter.

**PyCharm**: File → Settings → Project Interpreter → Add → Existing Environment → Select `.venv/bin/python`.

## 5. Pro Tip: Automatic Activation

Add to `.bashrc` or `.zshrc`:
```bash
# Auto-activate venv if present
if [ -f ".venv/bin/activate" ]; then
    source .venv/bin/activate
fi
```

Or use `direnv` for sophisticated auto-activation.
