# Creating & Using Virtual Environments

The `venv` module is built into Python 3.3+. No external tools needed.

## 🎯 Learning Objectives
- Create a virtual environment with `venv`.
- Understand the directory structure.
- Verify venv is active.

## 1. Creating a Virtual Environment

```bash
python -m venv .venv
```

**Breakdown**:
- `python`: The Python version you want to use
- `-m venv`: Run the venv module
- `.venv`: Name of the directory to create

**Naming Convention**: `.venv` or `venv` (I prefer `.venv` because `.` hides it on Unix)

## 2. Directory Structure

After creation:
```
.venv/
├── bin/  (Linux/Mac) OR Scripts/  (Windows)
│   ├── activate
│   ├── pip
│   └── python  (symlink to system python)
├── include/
├── lib/
│   └── python3.12/
│       └── site-packages/  ← Packages install here
└── pyvenv.cfg  (metadata)
```

## 3. Activating the Environment

**Windows (PowerShell)**:
```powershell
.venv\Scripts\Activate.ps1
```

**Windows (CMD)**:
```cmd
.venv\Scripts\activate.bat
```

**Linux/Mac**:
```bash
source .venv/bin/activate
```

## 4. Verifying Activation

Your prompt changes:
```
(.venv) C:\Projects\myapp>
```

Check which Python:

````
import sys
print(sys.executable)
# Should show: /path/to/project/.venv/bin/python
````

## 5. Deactivating

```bash
deactivate
```

The `(.venv)` prefix disappears.
