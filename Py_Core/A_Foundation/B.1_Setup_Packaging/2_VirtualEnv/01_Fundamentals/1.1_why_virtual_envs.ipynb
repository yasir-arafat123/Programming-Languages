# Why Virtual Environments?

The #1 best practice in Python development. No exceptions.

## 🎯 Learning Objectives
- Understand the "dependency hell" problem.
- Learn why global installs are dangerous.
- See real-world isolation scenarios.

## 1. The Problem: Global Package Hell

**Scenario**:
- Project A needs `django==3.2`
- Project B needs `django==4.2`

If you install both globally:
```bash
pip install django==3.2  # Project A works
pip install django==4.2  # Project A breaks!
```

**You can only have ONE version installed globally.**

## 2. The Solution: Isolated Environments

A virtual environment creates a **separate** Python installation for each project.

```
~/.local/  (global - BAD)
    └── lib/python3.12/site-packages/
        └── django (version 4.2)

project_a/
    └── .venv/  (isolated - GOOD)
        └── lib/python3.12/site-packages/
            └── django (version 3.2)

project_b/
    └── .venv/  (isolated - GOOD)
        └── lib/python3.12/site-packages/
            └── django (version 4.2)
```

## 3. Additional Benefits

1. **Reproducibility**: `requirements.txt` works reliably.
2. **No sudo required**: You own the venv directory.
3. **Easy cleanup**: Delete the folder, dependencies gone.
4. **Testing**: Test against multiple Python versions.

## 4. Real-World Example

Data scientist workflow:
```
~/projects/
├── ml_project_2023/  (uses tensorflow 2.10)
│   └── .venv/
├── ml_project_2025/  (uses tensorflow 2.15)
│   └── .venv/
└── legacy_analysis/   (uses pandas 1.3)
    └── .venv/
```

All coexist peacefully!

## 5. The Golden Rule

**ONE project = ONE virtual environment.**

Never skip this step. Ever.
