# Common Virtual Environment Mistakes

Even pros make these errors. Learn to avoid them.

## 🎯 Learning Objectives
- Diagnose forgotten activation.
- Fix "package not found" errors.
- Handle venv corruption.

## 1. Mistake: Forgot to Activate

**Symptom**: `ImportError: No module named 'requests'`

**Cause**: You installed to the venv, but your IDE/terminal is using global Python.

**Fix**:
```bash
source .venv/bin/activate  # Linux/Mac
.venv\Scripts\activate  # Windows
```

**Prevention**: Check for `(.venv)` in your prompt.

## 2. Mistake: Using Global pip

**Symptom**: Packages install but aren't visible in venv.

**Diagnosis**:

````
!which pip  # Linux/Mac
# Should show: /path/to/project/.venv/bin/pip
# NOT: /usr/local/bin/pip
````

**Fix**: Deactivate, reactivate, reinstall.

## 3. Mistake: Moving the Project Folder

**Symptom**: After moving `/home/user/old_path/project` to `/home/user/new_path/project`, the venv breaks.

**Cause**: Venvs contain **hardcoded paths**.

**Fix**: Delete `.venv` and recreate:
```bash
rm -rf .venv
python -m venv .venv
pip install -r requirements.txt
```

## 4. Mistake: Committing .venv to Git

**Problem**: `.venv/` can be 100MB+. It bloats your repo and doesn't work across OS.

**Fix**: Add to `.gitignore`:
```gitignore
.venv/
venv/
*.pyc
__pycache__/
```

## 5. Mistake: Corrupt Venv

**Symptom**: Random import errors, pip crashes.

**Fix**: Nuclear option (always works):
```bash
rm -rf .venv
python -m venv .venv
source .venv/bin/activate
pip install --upgrade pip
pip install -r requirements.txt
```
