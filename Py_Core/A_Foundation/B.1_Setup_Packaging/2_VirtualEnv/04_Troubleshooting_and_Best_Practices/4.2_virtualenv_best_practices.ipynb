# Virtual Environment Best Practices

Production-grade venv workflows.

## 🎯 Learning Objectives
- Follow the golden rules.
- Establish professional habits.
- Integrate with CI/CD.

## 1. The Golden Rules

### ✅ DO:
1. **Create a venv for EVERY project** (no exceptions).
2. **Use `.venv` as the directory name** (IDE auto-detection).
3. **Add `.venv/` to `.gitignore`**.
4. **Keep `requirements.txt` updated**.
5. **Test by recreating the venv monthly**.

### ❌ DON'T:
1. **Don't share venvs across projects**.
2. **Don't commit `.venv/` to version control**.
3. **Don't install packages globally** (except dev tools like `pip`, `virtualenv`).
4. **Don't move venvs between machines** (create fresh).

## 2. Project Initialization Checklist

```bash
# 1. Create project directory
mkdir myproject && cd myproject

# 2. Initialize git
git init

# 3. Create .gitignore
echo ".venv/" >> .gitignore
echo "__pycache__/" >> .gitignore
echo "*.pyc" >> .gitignore

# 4. Create venv
python -m venv .venv

# 5. Activate
source .venv/bin/activate

# 6. Upgrade pip
pip install --upgrade pip

# 7. Create requirements
touch requirements.txt
```

## 3. Directory Structure

```
myproject/
├── .venv/           (not in git)
├── .git/
├── .gitignore
├── src/
├── tests/
├── requirements.txt
├── requirements-dev.txt
└── README.md
```

## 4. CI/CD Pattern

**GitHub Actions** (`.github/workflows/test.yml`):
```yaml
- name: Set up Python
  uses: actions/setup-python@v4
  with:
    python-version: '3.12'

- name: Create venv and install deps
  run: |
    python -m venv .venv
    source .venv/bin/activate
    pip install -r requirements.txt

- name: Run tests
  run: |
    source .venv/bin/activate
    pytest
```

## 5. Documentation

Add to **README.md**:
```markdown
## Setup

1. Clone the repository
2. Create virtual environment: `python -m venv .venv`
3. Activate: `source .venv/bin/activate` (Linux/Mac) or `.venv\Scripts\activate` (Windows)
4. Install dependencies: `pip install -r requirements.txt`
5. Run tests: `pytest`
```
