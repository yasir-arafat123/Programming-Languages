# virtualenv (The Third-Party Tool)

`venv` is built-in, but `virtualenv` is more powerful.

## 🎯 Learning Objectives
- Understand venv vs virtualenv differences.
- Use virtualenv for advanced features.
- Know when to use each.

## 1. venv vs virtualenv

| Feature | venv (built-in) | virtualenv |
| :--- | :---: | :---: |
| Python 3.3+ | ✅ Included | ❌ pip install virtualenv |
| Speed | Slower | **Faster** |
| Python 2 support | ❌ | ✅ |
| Plugin system | ❌ | ✅ |

## 2. Installing virtualenv

````
%pip install virtualenv
````

## 3. Creating an Environment

```bash
virtualenv .venv
```

Activation is identical to `venv`.

## 4. Advanced Features

### Specify Python Version
```bash
virtualenv -p python3.9 .venv39
```

### Copy Instead of Symlink
```bash
virtualenv --always-copy .venv
```

(Useful for Docker containers)

## 5. When to Use Each

**Use `venv` if**:
- You want zero dependencies
- You're on Python 3.3+

**Use `virtualenv` if**:
- You need Python 2 support
- You want faster environment creation
- You use plugins (e.g., `virtualenvwrapper`)
