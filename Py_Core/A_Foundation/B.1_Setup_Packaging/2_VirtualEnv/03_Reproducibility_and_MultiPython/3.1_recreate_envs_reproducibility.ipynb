# Recreating Environments for Reproducibility

The "it works on my machine" problem, solved.

## 🎯 Learning Objectives
- Achieve deterministic builds.
- Use hashed requirements.
- Test environment recreation.

## 1. The Reproducibility Problem

**Day 1 (Your Machine)**:
```bash
pip install flask  # Gets flask 3.0.0
```

**6 Months Later (Production Server)**:
```bash
pip install flask  # Gets flask 3.1.0 (Breaking changes!)
```

Your app crashes in production.

## 2. Solution: Exact Pin Everything

**Production requirements.txt**:
```text
flask==3.0.3
werkzeug==3.0.1
click==8.1.7
itsdangerous==2.1.2
jinja2==3.1.2
markupsafe==2.1.3
```

Generated via `pip freeze`.

## 3. Testing Reproducibility

**Monthly Drill**:
```bash
# Delete old venv
rm -rf .venv

# Create fresh
python -m venv .venv
source .venv/bin/activate

# Install
pip install -r requirements.txt

# Run tests
pytest
```

If tests pass, your requirements are solid.

## 4. Two-File Strategy

**requirements.in** (flexible, for development):
```text
flask>=3.0
requests
```

**requirements.txt** (locked, for production):
```text
flask==3.0.3
werkzeug==3.0.1
...
```

Use `pip-tools` to compile `.in` → `.txt`:
```bash
pip-compile requirements.in
```

## 5. CI/CD Integration

Your GitHub Actions should recreate the venv from scratch every time to catch reproducibility bugs early.
