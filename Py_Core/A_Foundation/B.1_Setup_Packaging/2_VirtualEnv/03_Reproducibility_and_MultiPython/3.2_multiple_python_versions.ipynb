# Multiple Python Versions

Test your code against Python 3.9, 3.10, 3.11, and 3.12.

## 🎯 Learning Objectives
- Install multiple Python versions.
- Create venvs with specific versions.
- Use `pyenv` for version management.

## 1. The Need

You maintain a library. Users have:
- Python 3.9 (legacy systems)
- Python 3.12 (latest adopters)

You need to test against **all** supported versions.

## 2. Windows: Python Launcher

Windows has `py` launcher:
```cmd
py -3.9 -m venv .venv39
py -3.12 -m venv .venv312
```

## 3. Linux/Mac: pyenv

Install `pyenv`:
```bash
curl https://pyenv.run | bash
```

Install multiple versions:
```bash
pyenv install 3.9.18
pyenv install 3.12.1
```

Set local version:
```bash
pyenv local 3.12.1
python --version  # Shows 3.12.1
```

## 4. Creating Version-Specific Venvs

```bash
# Use Python 3.9
~/.pyenv/versions/3.9.18/bin/python -m venv .venv39

# Use Python 3.12
~/.pyenv/versions/3.12.1/bin/python -m venv .venv312
```

## 5. Testing with tox

`tox` automates testing across multiple Python versions:

```ini
# tox.ini
[tox]
envlist = py39,py310,py311,py312

[testenv]
deps = pytest
commands = pytest
```

Run:
```bash
tox
```

It creates 4 venvs and runs tests in each.
