# Import Basics & Name Lookup

Understanding `import` mechanics prevents "ModuleNotFoundError".

## 🎯 Learning Objectives
- Understand how `import` searches for modules.
- Inspect `sys.path`.
- Debug import issues.

## 1. The Import Search Path

When you run `import mymodule`, Python searches in order:

````
import sys
for path in sys.path:
    print(path)
````

**Typical** `sys.path`:
1.  Current directory (`''`)
2.  `PYTHONPATH` environment variable directories
3.  Standard library (`/usr/lib/python3.12`)
4.  Site-packages (`.venv/lib/python3.12/site-packages`)

## 2. Import Algorithm

For `import foo`:

1.  Check if `foo` is already in `sys.modules` (cached).
2.  If not, search in `sys.path` for:
    - `foo.py` (module)
    - `foo/__init__.py` (package)
3.  Once found, execute the file and cache in `sys.modules`.
4.  Create a name `foo` in the current namespace.

## 3. The Module Cache

````
import math
print('math' in sys.modules)  # True

# Importing again uses the cache, does NOT re-execute math.py
import math
````

## 4. Common Import Error

**Error**:
```
ModuleNotFoundError: No module named 'mymodule'
```

**Causes**:
1.  Typo in module name
2.  Module not in `sys.path`
3.  Forgot to install the package (`pip install mymodule`)

**Debug**: Print `sys.path` and verify your module location.

## 5. Modifying sys.path (Not Recommended)

```python
import sys
sys.path.append('/path/to/my/modules')

import mymodule  # Now Python can find it
```

⚠️ **Better**: Use proper package installation (`pip install -e .`).
