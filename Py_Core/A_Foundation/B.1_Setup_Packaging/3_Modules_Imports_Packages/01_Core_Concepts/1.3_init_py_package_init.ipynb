# __init__.py & Package Initialization

The `__init__.py` file transforms a directory into a package. Let's master its power.

## 🎯 Learning Objectives
- Understand `__init__.py` purpose.
- Control what gets exposed from a package.
- Initialize package-level variables.

## 1. The Role of __init__.py

**Without `__init__.py`** (Python 2 / implicit namespaces aside):
```
mypackage/
└── utils.py
```
`import mypackage` → Error: Not recognized as a package.

**With `__init__.py`**:
```
mypackage/
├── __init__.py
└── utils.py
```
`import mypackage` → Success!

## 2. Empty vs Populated __init__.py

### Empty (Minimal)
```python
# mypackage/__init__.py
# (empty file)
```

Usage:
```python
from mypackage.utils import helper_func
```

### Populated (Convenience)
```python
# mypackage/__init__.py
from .utils import helper_func

__all__ = ['helper_func']
```

Usage:
```python
from mypackage import helper_func  # Shorter!
```

## 3. Package Initialization Code

Code in `__init__.py` runs **once** when the package is first imported.

````
# Example __init__.py
print("Package mypackage is loading...")

# Set package-level variable
__version__ = "1.0.0"

# Import submodules
from .utils import *
from .helpers import *
````

## 4. Subpackages

```
mypackage/
├── __init__.py
├── utils.py
└── subpackage/
    ├── __init__.py
    └── tools.py
```

Usage:
```python
from mypackage.subpackage import tools
```

## 5. __all__ Variable

Controls what `from mypackage import *` exports:

```python
__all__ = ['func1', 'func2']  # Only these are exported
```
