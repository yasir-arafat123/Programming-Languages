# Module vs Package vs Distribution

These words are often confused. Let's define them precisely.

## 🎯 Learning Objectives
- Distinguish module from package from distribution.
- Understand real-world examples.
- Know which term to use when.

## 1. Module

A **module** is a single `.py` file.

Example:
```python
# myutils.py
def greet(name):
    return f"Hello, {name}!"
```

Usage:
```python
import myutils
print(myutils.greet("Alice"))
```

## 2. Package

A **package** is a **directory** containing an `__init__.py` file and other modules.

Example:
```
mypackage/
├── __init__.py
├── utils.py
└── helpers.py
```

Usage:
```python
import mypackage.utils
from mypackage import helpers
```

## 3. Distribution

A **distribution** is what you download from PyPI. It's a compressed archive (`.whl` or `.tar.gz`) containing one or more packages.

Example:
```bash
pip install requests
```

This downloads the `requests` **distribution**, which contains the `requests` **package**, which contains multiple **modules** (`api.py`, `sessions.py`, etc.).

## 4. Visual Summary

```
requests (distribution on PyPI)
└── requests/  (package installed in site-packages)
    ├── __init__.py
    ├── api.py  (module)
    ├── sessions.py  (module)
    └── models.py  (module)
```

## 5. Common Confusion

People say "I installed the pandas package".

**Technically**: You installed the `pandas` **distribution**, which contains the `pandas` **package**.
