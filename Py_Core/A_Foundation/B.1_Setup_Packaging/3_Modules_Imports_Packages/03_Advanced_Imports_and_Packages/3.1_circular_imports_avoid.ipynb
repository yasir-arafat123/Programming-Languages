# Circular Imports - How to Avoid

Module A imports B, B imports A. Python crashes. Let's fix it.

## 🎯 Learning Objectives
- Understand circular import errors.
- Refactor to eliminate circles.
- Use late imports as a workaround.

## 1. The Problem

**File: `models.py`**:
```python
from views import render_user

class User:
    def display(self):
        return render_user(self)
```

**File: `views.py`**:
```python
from models import User  # CIRCULAR!

def render_user(user: User):
    return f"<div>{user.name}</div>"
```

**Error**:
```
ImportError: cannot import name 'render_user' from partially initialized module 'views'
```

## 2. Solution 1: Refactor (Best)

Move shared code to a **third module**:

```
models.py
views.py
utils.py  ← Extract common dependencies here
```

## 3. Solution 2: Late Import

**models.py**:
```python
class User:
    def display(self):
        from views import render_user  # Import INSIDE function
        return render_user(self)
```

This delays the import until runtime.

## 4. Solution 3: Type Hints Only

**views.py**:
```python
from typing import TYPE_CHECKING

if TYPE_CHECKING:
    from models import User  # Only for type checkers

def render_user(user: 'User'):  # String annotation
    return f"<div>{user.name}</div>"
```

## 5. Prevention

- **Keep dependencies unidirectional** (models → views, not both ways).
- **Use dependency injection** instead of direct imports.
- **Visualize** your import graph with tools like `pydeps`.
