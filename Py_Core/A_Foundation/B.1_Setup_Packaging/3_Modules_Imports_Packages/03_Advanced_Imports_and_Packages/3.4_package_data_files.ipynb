# Package Data Files

How to include non-Python files (images, JSON, templates) in your package.

## 🎯 Learning Objectives
- Include data files in packages.
- Access them at runtime with `importlib.resources`.
- Configure `pyproject.toml` for package data.

## 1. The Problem

Your package needs `config.json`:

```
mypackage/
├── __init__.py
├── utils.py
└── config.json  ← How to include this?
```

## 2. Declaring Package Data

In `pyproject.toml`:
```toml
[tool.setuptools.package-data]
mypackage = ["*.json", "templates/*"]
```

## 3. Accessing Data (Modern Way)

Use `importlib.resources` (Python 3.9+):

````python
from importlib import resources
import json

# Read config.json from mypackage/
data = resources.files('mypackage').joinpath('config.json').read_text()
config = json.loads(data)
print(config)
````

## 4. Old Way (Deprecated)

❌ **Don't use** `__file__` paths:
```python
import os
config_path = os.path.join(os.path.dirname(__file__), 'config.json')
# Breaks when installed as a ZIP!
```

## 5. MANIFEST.in (Rarely Needed)

If using old `setup.py`:

```
# MANIFEST.in
include mypackage/*.json
recursive-include mypackage/templates *
```
