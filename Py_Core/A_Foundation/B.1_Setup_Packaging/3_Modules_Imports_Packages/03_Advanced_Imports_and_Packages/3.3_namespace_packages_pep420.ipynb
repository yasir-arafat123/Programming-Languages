# Namespace Packages (PEP 420)

Packages without `__init__.py`. Advanced topic for large projects.

## 🎯 Learning Objectives
- Understand namespace packages.
- Know when to use them.
- See real-world examples (Zope, Azure SDK).

## 1. Traditional Package

```
mycompany/
├── __init__.py  ← Required
└── utils.py
```

## 2. Namespace Package

```
mycompany/
└── utils.py  ← No __init__.py!
```

Python 3.3+ treats this as a **namespace package**.

## 3. The Use Case

**Problem**: Different teams maintain different parts of `mycompany` package.

```
repo1/
└── mycompany/
    └── database/

repo2/
└── mycompany/
    └── api/
```

Installing both creates:
```
site-packages/mycompany/database/
site-packages/mycompany/api/
```

Users can do:
```python
from mycompany.database import connect
from mycompany.api import server
```

## 4. Real-World: Azure SDK

```bash
pip install azure-storage-blob
pip install azure-identity
```

Both share the `azure` namespace:
```python
from azure.storage.blob import BlobClient
from azure.identity import DefaultAzureCredential
```

## 5. When NOT to Use

❌ **Single-repo projects**: Just use regular packages.

✅ **Multi-repo, shared namespace**: Namespace packages shine.
