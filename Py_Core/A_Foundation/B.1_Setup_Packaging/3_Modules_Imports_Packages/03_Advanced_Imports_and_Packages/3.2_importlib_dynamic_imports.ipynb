# importlib & Dynamic Imports

Import modules at runtime based on user input or configuration.

## 🎯 Learning Objectives
- Use `importlib.import_module()`.
- Reload modules dynamically.
- Build plugin systems.

## 1. The Need

**Problem**: You don't know which module to import until runtime.

Example: Plugin loader based on config file.

## 2. importlib.import_module()

````
import importlib

module_name = "math"  # Could come from config
math_module = importlib.import_module(module_name)

print(math_module.sqrt(16))  # 4.0
````

## 3. Importing Specific Attributes

```python
module = importlib.import_module('package.submodule')
func = getattr(module, 'function_name')
result = func()
```

## 4. Reloading Modules (Development)

**Problem**: Module is cached after first import.

**Solution** (only during development!):

````
import importlib
import mymodule

# Edit mymodule.py...

importlib.reload(mymodule)  # Re-executes the file
````

## 5. Plugin System Example

```python
# config.yaml: plugin: myapp.plugins.email

import importlib
import yaml

with open('config.yaml') as f:
    config = yaml.safe_load(f)

plugin_module = importlib.import_module(config['plugin'])
plugin_module.run()
```
