# Relative Imports

Importing using `.` notation for package-internal imports.

## 🎯 Learning Objectives
- Use `.` and `..` in imports.
- Know when relative imports are appropriate.
- Understand limitations.

## 1. The Dot Notation

```python
from . import sibling   # Same directory
from .. import parent  # Parent directory
from ..sibling import foo  # Sibling of parent
```

## 2. Example Project

```
myproject/
├── __init__.py
├── utils.py
└── database/
    ├── __init__.py
    ├── models.py
    └── queries.py
```

**In `database/queries.py`**:
```python
from . import models  # Relative (same package)
from .. import utils  # Go up one level
```

## 3. Advantages

✅ **Package portability**: Rename the package,imports still work.

✅ **Less typing** (within the same package).

## 4. Limitations

❌ **Cannot run as script**:
```bash
python database/queries.py  # Error:
# ImportError: attempted relative import with no known parent package
```

✅ **Must run as module**:
```bash
python -m myproject.database.queries  # Works!
```

## 5. Best Practice

- **Use absolute imports** by default.
- **Use relative imports** only for internal package structure (sibling modules in same package).
