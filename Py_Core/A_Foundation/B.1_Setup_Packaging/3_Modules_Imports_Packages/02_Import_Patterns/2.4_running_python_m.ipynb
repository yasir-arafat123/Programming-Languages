# Running with python -m

The difference between `python script.py` and `python -m module`.

## 🎯 Learning Objectives
- Understand `-m` flag.
- Know when to use itexecution.
- Fix import errors with `-m`.

## 1. Direct Execution

```bash
python myproject/api/views.py
```

- Python adds `myproject/api/` to `sys.path`.
- Absolute imports like `from myproject.database import models` **fail**.

## 2. Module Execution (-m)

```bash
python -m myproject.api.views
```

- Python adds current directory to `sys.path`.
- Absolute imports **work**!

## 3. Real-World Examples

**http.server**:
```bash
python -m http.server 8000
``` 

**pytest**:
```bash
python -m pytest tests/
```

**pip** (recommended):
```bash
python -m pip install requests
```

## 4. The __main__.py Pattern

```
myproject/
└── __main__.py
```

Allows:
```bash
python -m myproject  # Runs __main__.py
```

## 5. Best Practice

- **Develop packages**: Always use `python -m`.
- **Simple scripts**: `python script.py` is fine.
