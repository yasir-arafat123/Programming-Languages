# sys.path & PYTHONPATH

How Python finds modules, and how to extend the search path.

## 🎯 Learning Objectives
- Inspect and modify `sys.path`.
- Set `PYTHONPATH` environment variable.
- Understand best practices.

## 1. sys.path

A list of directories Python searches for modules.

````
import sys
print(sys.path)
````

## 2. Modifying sys.path (Runtime)

```python
import sys
sys.path.insert(0, '/path/to/my/modules')
# Now you can import from that directory
```

⚠️ **Warning**: This is a hack. Don't abuse it.

## 3. PYTHONPATH Environment Variable

**Linux/Mac**:
```bash
export PYTHONPATH=/path/to/modules:/another/path
python myscript.py
```

**Windows (PowerShell)**:
```powershell
$env:PYTHONPATH = "C:\path\to\modules"
python myscript.py
```

## 4. When to Use

**Legitimate Uses**:
- Temporary testing during development
- Corporate environments with custom module locations

**Better Alternatives**:
- Use `pip install -e .` for local development
- Publish packages to PyPI or private index

## 5. Debugging Import Issues

```python
import sys
print('\n'.join(sys.path))
# Check if your module directory is listed
```
