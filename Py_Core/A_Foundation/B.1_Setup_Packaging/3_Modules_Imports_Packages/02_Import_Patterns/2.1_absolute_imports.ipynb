# Absolute Imports

The standard, recommended way to import in Python.

## 🎯 Learning Objectives
- Understand absolute import syntax.
- Know when to use absolute imports.
- Avoid common pitfalls.

## 1. What is an Absolute Import?

An **absolute import** specifies the full path from the project root.

```python
from myproject.utils import helper
from myproject.database.models import User
```

## 2. Example Project

```
myproject/
├── __init__.py
├── utils.py
├── database/
│   ├── __init__.py
│   └── models.py
└── api/
    ├── __init__.py
    └── views.py
```

In `api/views.py`, use:
```python
from myproject.database.models import User  # Absolute
```

## 3. Advantages

✅ **Clear**: Anyone reading the code knows exactly where the import comes from.

✅ **Refactor-safe**: Moving files doesn't break imports (as long as you update appropriately).

✅ **PEP 8 recommended**.

## 4. Running Code with Absolute Imports

**Problem**: If you run `python api/views.py` directly, Python doesn't know about `myproject`.

**Solution**: Run as a module from the project root:
```bash
python -m myproject.api.views
```

## 5. Common Mistake

❌ **Wrong** (inside myproject/):
```python
from database.models import User  # Fragile!
```

✅ **Right**:
```python
from myproject.database.models import User
```
