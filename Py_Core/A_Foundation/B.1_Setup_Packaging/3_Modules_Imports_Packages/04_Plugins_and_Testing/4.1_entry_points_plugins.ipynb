# Entry Points & Plugins

Create installable command-line tools and extensible plugin systems.

## 🎯 Learning Objectives
- Create CLI tools with `console_scripts`.
- Build plugin systems.
- Discover installed plugins dynamically.

## 1. Console Scripts

**Problem**: How does `pip install requests` create a `pip` command?

**Solution**: Entry points in `pyproject.toml`:

```toml
[project.scripts]
mytool = "mypackage.cli:main"
```

After install:
```bash
mytool  # Runs mypackage.cli.main()
```

## 2. Example CLI Tool

**File: `mypackage/cli.py`**:
```python
def main():
    print("Hello from mytool!")

if __name__ == "__main__":
    main()
```

**pyproject.toml**:
```toml
[project.scripts]
mytool = "mypackage.cli:main"
```

**After `pip install` creating `mytool` command globally**.

## 3. Plugin System

**Core App** defines plugin interface.

**pyproject.toml** (plugin):
```toml
[project.entry-points."myapp.plugins"]
email = "myplugin.email:EmailPlugin"
```

**Core App discovers plugins**:
```python
from importlib.metadata import entry_points

plugins = entry_points(group='myapp.plugins')
for plugin in plugins:
    cls = plugin.load()
    instance = cls()
    instance.run()
```

## 4. Real Example: Pytest Plugins

Pytest discovers plugins via:
```toml
[project.entry-points.pytest11]
myplugin = "myplugin.plugin"
```
