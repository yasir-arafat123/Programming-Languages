# Imports in Testing & Mocking

Control imports during testing to isolate code.

## 🎯 Learning Objectives
- Mock imported modules with `unittest.mock`.
- Patch imports for testing.
- Test code with external dependencies.

## 1. The Problem

**Code**:
```python
# myapp/api.py
import requests

def fetch_data():
    return requests.get('https://api.example.com').json()
```

**Test Challenge**: Don't want to hit real API during tests.

## 2. Mocking with unittest.mock

```python
from unittest.mock import patch, MagicMock
from myapp.api import fetch_data

@patch('myapp.api.requests.get')
def test_fetch_data(mock_get):
    # Setup mock
    mock_response = MagicMock()
    mock_response.json.return_value = {'key': 'value'}
    mock_get.return_value = mock_response
    
    # Test
    result = fetch_data()
    assert result == {'key': 'value'}
    mock_get.assert_called_once()
```

## 3. Where to Patch

❌ **Wrong**: `@patch('requests.get')`

✅ **Right**: `@patch('myapp.api.requests.get')`

**Rule**: Patch where it's imported, not where it's defined.
