# Versioning Strategies

How to manage version numbers professionally.

## 🎯 Learning Objectives
- Understand SemVer.
- Learn CalVer alternative.
- Automate version bumping.

## 1. Semantic Versioning (SemVer)

**Format**: `MAJOR.MINOR.PATCH`

- **MAJOR**: Breaking changes (2.0.0 → 3.0.0)
- **MINOR**: New features (2.1.0 → 2.2.0)
- **PATCH**: Bug fixes (2.1.1 → 2.1.2)

**Use**: Most libraries (requests, flask, pandas)

## 2. Calendar Versioning (CalVer)

**Format**: `YYYY.MM.MICRO`

Examples:
- Ubuntu: `22.04`, `24.04`
- pip: `24.0`

**Use**: Enterprise software, time-based releases.

## 3. Pre-release Versions

```
1.0.0a1  # Alpha 1
1.0.0b2  # Beta 2
1.0.0rc1 # Release Candidate 1
1.0.0    # Final
```

## 4. Version Bumping with `bump-my-version`

```bash
pip install bump-my-version
```

**pyproject.toml**:
```toml
[tool.bumpversion]
current_version = "1.2.3"
commit = true
tag = true

[[tool.bumpversion.files]]
filename = "pyproject.toml"
search = "version = \"{current_version}\""
replace = "version = \"{new_version}\""
```

**Usage**:
```bash
bump-my-version bump patch  # 1.2.3 → 1.2.4
bump-my-version bump minor  # 1.2.4 → 1.3.0
bump-my-version bump major  # 1.3.0 → 2.0.0
```

## 5. Git Tags

```bash
git tag -a v1.0.0 -m "Release version 1.0.0"
git push origin v1.0.0
```

CI/CD can auto-publish when tags are pushed.
