# Building Wheels & Source Distributions

Create distributable packages for PyPI.

## 🎯 Learning Objectives
- Build wheels and sdists.
- Understand distribution formats.
- Use `python -m build`.

## 1. Install Build Tool

```bash
pip install build
```

## 2. Building Distributions

```bash
python -m build
```

Creates `dist/` directory with:
```
dist/
├── mypackage-1.0.0.tar.gz     # Source distribution
└── mypackage-1.0.0-py3-none-any.whl  # Wheel
```

## 3. Wheel vs Sdist

**Wheel (`.whl`)**:
- Binary distribution
- Fast to install (unzip)
- Platform-specific for compiled code

**Sdist (`.tar.gz`)**:
- Source code archive
- Slower to install (must build)
- Universal

## 4. Testing Built Package

```bash
# Create test venv
python -m venv test_env
source test_env/bin/activate

# Install from wheel
pip install dist/mypackage-1.0.0-py3-none-any.whl

# Test import
python -c "import mypackage; print(mypackage.__version__)"
```

## 5. Clean Build

```bash
# Remove old builds
rm -rf dist/ build/ *.egg-info

# Rebuild
python -m build
```
