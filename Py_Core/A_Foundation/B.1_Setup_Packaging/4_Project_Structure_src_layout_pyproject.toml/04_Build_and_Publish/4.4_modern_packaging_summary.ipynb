The state of Python packaging in 2025.

## 🎯 Learning Objectives
- Recap best practices.

## 1. Modern Stack (2025)

| Task | Tool |
| :--- | :--- |
| **Config** | `pyproject.toml` (PEP 621) |
| **Layout** | src/ layout |
| **Build** | `python -m build` |
|
 **Publish** | `twine` |
| **Venv** | `python -m venv` |
| **Formatter** | `black` or `ruff format` |
| **Linter** | `ruff` |
| **Type** | `mypy` |
| **Test** | `pytest` |

## 2. The Golden Rules

✅ **DO**:
1. Use `pyproject.toml` for all config
2. Use src/ layout
3. Install in editable mode: `pip install -e ".[dev]"`
4. Pin dependencies in applications, loose in libraries
5. Test with `pytest --cov`
6. Upload to TestPyPI before real PyPI

❌ **DON'T**:
1. Use `setup.py` for new projects
2. Commit `.venv/` or `dist/` to git
3. Use `sudo pip`
4. Hardcode paths
