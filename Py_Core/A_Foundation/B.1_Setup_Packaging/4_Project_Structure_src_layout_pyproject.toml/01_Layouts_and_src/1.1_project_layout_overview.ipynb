# Project Layout Overview

Structure matters. Let's compare the two main approaches.

## 🎯 Learning Objectives
- Understand flat layout.
- Understand src layout.
- Know when to use each.

## 1. Flat Layout

```
myproject/
├── mypackage/
│   ├── __init__.py
│   └── utils.py
├── tests/
├── pyproject.toml
└── README.md
```

**Pros**:
- Simple
- Easy to navigate

**Cons**:
- Can accidentally import uninstalled code during development
- Tests might pass locally but fail in production

## 2. src Layout (Recommended)

```
myproject/
├── src/
│   └── mypackage/
│       ├── __init__.py
│       └── utils.py
├── tests/
├── pyproject.toml
└── README.md
```

**Pros**:
- Forces proper installation (`pip install -e .`)
- Tests always run against installed code
- Industry standard

**Cons**:
- One extra directory level

## 3. Why src Layout is Better

**Flat** layout problem:
```bash
python -m pytest  # Imports mypackage/ directly (not installed)
```

**src** layout prevents this:
```bash
python -m pytest  # ImportError unless pip install -e .
```

This catches missing `MANIFEST.in` files and packaging bugs **before** you publish.

## 4. Official Recommendation

**PyPA (Python Packaging Authority)** recommends src layout for all new projects.
