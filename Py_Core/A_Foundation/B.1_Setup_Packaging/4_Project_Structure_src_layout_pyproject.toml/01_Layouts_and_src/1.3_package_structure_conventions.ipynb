# Package Structure Conventions

Professional Python projects follow predictable structures.

## 🎯 Learning Objectives
- Understand standard directory layout.
- Know what goes where.
- Follow community conventions.

## 1. Complete Project Structure

```
myproject/
├── .git/
├── .github/
│   └── workflows/          # CI/CD
├── .venv/                  # (Excluded from git)
├── docs/
│   ├── conf.py
│   └── index.rst
├── src/
│   └── mypackage/
│       ├── __init__.py
│       └── utils.py
├── tests/
│   ├── __init__.py
│   └── test_utils.py
├── examples/              # Usage examples
├── scripts/               # Dev scripts
├── .gitignore
├── LICENSE
├── pyproject.toml
├── README.md
└── requirements-dev.txt
```

## 2. Directory Purposes

| Directory | Purpose |
| :--- | :--- |
| `src/` | Package source code |
| `tests/` | Unit/integration tests |
| `docs/` | Sphinx documentation |
| `examples/` | Tutorial code samples |
| `scripts/` | Build/deploy scripts |
| `.github/` | GitHub Actions workflows |

## 3. Root Files

| File | Purpose |
| :--- | :--- |
| `pyproject.toml` | Modern packaging config |
| `README.md` | Project description |
| `LICENSE` | Legal terms |
| `.gitignore` | Exclude files from git |
| `requirements-dev.txt` | Dev dependencies |

## 4. Test Directory Structure

**Mirror your src structure**:
```
src/mypackage/api/views.py
tests/api/test_views.py
```

## 5. Essential .gitignore

```gitignore
# Virtual environments
.venv/
venv/

# Python
__pycache__/
*.pyc
*.pyo
*.egg-info/
dist/
build/

# IDE
.vscode/
.idea/

# OS
.DS_Store
Thumbs.db
```
