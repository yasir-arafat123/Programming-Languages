# The src Layout Concept

Deep dive into why professional projects use `src/`.

## 🎯 Learning Objectives
- Understand the "import shadowing" problem.
- Set up src layout correctly.
- Configure tools for src layout.

## 1. The Import Shadowing Problem

**Flat layout**:
```
myproject/
└── mypackage/
    └── utils.py
```

Running `python` from `myproject/` adds it to `sys.path`.

```python
import mypackage  # Works WITHOUT installation!
```

This hides packaging bugs.

## 2. src Layout Prevents This

```
myproject/
└── src/
    └── mypackage/
```

Running `python` from `myproject/` does NOT add `src/` to path.

```python
import mypackage  # ImportError UNLESS pip install -e .
```

Forces proper installation during development.

## 3. Setting Up src Layout

**pyproject.toml**:
```toml
[tool.setuptools.packages.find]
where = ["src"]
```

This tells setuptools to look in `src/` for packages.

## 4. Development Workflow

```bash
# 1. Create venv
python -m venv .venv

# 2. Activate
source .venv/bin/activate

# 3. Editable install
pip install -e .

# Now imports work!
python -c "import mypackage"
```

## 5. Tool Configuration

**pytest.ini** or **pyproject.toml**:
```toml
[tool.pytest.ini_options]
pythonpath = "src"
```

This helps pytest find your package.
