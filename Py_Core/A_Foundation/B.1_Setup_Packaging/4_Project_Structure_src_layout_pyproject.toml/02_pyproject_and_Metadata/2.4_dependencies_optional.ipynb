# Dependencies & Optional Dependencies

Declaring what your package needs to run.

## 🎯 Learning Objectives
- Declare runtime dependencies.
- Create optional dependency groups.
- Pin versions appropriately.

## 1. Core Dependencies

```toml
[project]
dependencies = [
    "requests>=2.28.0,<3.0.0",
    "pandas~=2.0.0",
    "click",
]
```

These are **always** installed with your package.

## 2. Optional Dependencies

```toml
[project.optional-dependencies]
dev = [
    "pytest>=7.0",
    "black",
    "mypy",
]
docs = [
    "sphinx",
    "sphinx-rtd-theme",
]
test = [
    "pytest",
    "pytest-cov",
]
```

Install with:
```bash
pip install mypackage[dev]
pip install mypackage[dev,docs]
```

## 3. Version Constraints Best Practices

**For Applications** (you control the environment):
```toml
dependencies = [
    "flask==3.0.3",  # Exact pin
]
```

**For Libraries** (users control the environment):
```toml
dependencies = [
    "flask>=3.0,<4.0",  # Flexible range
]
```

## 4. Python Version Requirement

```toml
[project]
requires-python = ">=3.9"
```

Prevents installation on Python 3.8 or earlier.

## 5. Extras Pattern for Features

```toml
[project.optional-dependencies]
postgres = ["psycopg2"]
mysql = ["pymysql"]
all = [
    "psycopg2",
    "pymysql",
]
```

Users choose database support:
```bash
pip install mypackage[postgres]
```
