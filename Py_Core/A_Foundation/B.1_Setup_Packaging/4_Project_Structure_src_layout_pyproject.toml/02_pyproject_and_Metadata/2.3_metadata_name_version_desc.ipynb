# Metadata: Name, Version, Description

Detailed guide to package metadata fields.

## 🎯 Learning Objectives
- Choose a good package name.
- Follow semantic versioning.
- Write effective descriptions.

## 1. Package Name

```toml
[project]
name = "my-awesome-package"
```

**Rules**:
- Lowercase
- Hyphens allowed (converted to underscores for imports)
- Must be unique on PyPI
- Check availability: `pip search my-awesome-package` (deprecated) or search pypi.org

## 2. Versioning (SemVer)

```toml
version = "1.2.3"
# MAJOR.MINOR.PATCH
```

- **MAJOR**: Breaking changes
- **MINOR**: New features (backward compatible)
- **PATCH**: Bug fixes

**Examples**:
- `0.1.0`: Initial development
- `1.0.0`: First stable release
- `1.1.0`: Added new function
- `2.0.0`: Removed deprecated API

## 3. Description & Readme

```toml
description = "A one-line description for PyPI"
readme = "README.md"
```

- `description`: Shown in `pip search`
- `readme`: Displayed on PyPI project page

## 4. Authors & Maintainers

```toml
authors = [
    {name = "Alice Smith", email = "alice@example.com"},
    {name = "Bob Jones"},
]
maintainers = [
    {name = "Charlie Brown", email = "charlie@example.com"},
]
```

## 5. URLs

```toml
[project.urls]
Homepage = "https://example.com"
Documentation = "https://docs.example.com"
Repository = "https://github.com/user/repo"
"Bug Tracker" = "https://github.com/user/repo/issues"
```
