# setup.py → setup.cfg → pyproject.toml

The evolution of Python packaging configuration.

## 🎯 Learning Objectives
- Understand the historical progression.
- Know which to use in 2025.
- Migrate from old styles.

## 1. setup.py Era (2000-2020)

```python
# setup.py
from setuptools import setup

setup(
    name="mypackage",
    version="1.0.0",
    packages=["mypackage"],
    install_requires=["requests"],
)
```

**Problems**:
- Executable code (security risk)
- Hard to parse by tools
- Can run arbitrary code during install

## 2. setup.cfg Era (2016-2021)

```ini
[metadata]
name = mypackage
version = 1.0.0

[options]
packages = find:
install_requires = requests
```

**Improvement**: Static, declarative.

## 3. pyproject.toml Era (2021+)

```toml
[project]
name = "mypackage"
version = "1.0.0"
dependencies = ["requests"]
```

**Advantages**:
- Standardized format (PEP 621)
- Used by ALL modern Python tools
- TOML is cleaner than INI

## 4. What to Use in 2025

✅ **Always use `pyproject.toml`**.

❌ **Don't** create new projects with `setup.py` or `setup.cfg`.

## 5. Migration

Old projects can still use `setup.py`, but gradually migrate to `pyproject.toml`.

Both can coexist during transition.
