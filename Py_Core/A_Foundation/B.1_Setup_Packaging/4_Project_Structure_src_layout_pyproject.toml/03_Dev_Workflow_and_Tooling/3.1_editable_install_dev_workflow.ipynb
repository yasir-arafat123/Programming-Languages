# Editable Install & Development Workflow

The daily workflow for package development.

## 🎯 Learning Objectives
- Set up editable installs.
- Iterate on code efficiently.
- Manage dev dependencies.

## 1. The Development Loop

```bash
# One-time setup
python -m venv .venv
source .venv/bin/activate
pip install -e ".[dev]"

# Daily workflow
# 1. Edit code in src/mypackage/
# 2. Run tests: pytest
# 3. Repeat (no reinstall needed!)
```

## 2. Editable Install Explained

`pip install -e .` creates a **symlink** from `site-packages/` to your `src/` directory.

Changes to source code are **immediately** visible without reinstalling.

## 3. Dev Dependencies Setup

**pyproject.toml**:
```toml
[project.optional-dependencies]
dev = [
    "pytest",
    "pytest-cov",
    "black",
    "ruff",
    "mypy",
]
```

Install everything:
```bash
pip install -e ".[dev]"
```

## 4. Pre-commit Hooks

Auto-format and lint before committing:

```yaml
# .pre-commit-config.yaml
repos:
  - repo: https://github.com/psf/black
    rev: 24.1.1
    hooks:
      - id: black
```

Install:
```bash
pip install pre-commit
pre-commit install
```
