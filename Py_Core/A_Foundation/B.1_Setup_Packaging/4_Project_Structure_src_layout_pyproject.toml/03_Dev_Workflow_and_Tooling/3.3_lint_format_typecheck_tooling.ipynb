# Lint, Format & Typecheck Tooling

Modern Python quality tools.

## 🎯 Learning Objectives
- Format code with black.
- Lint with ruff.
- Typecheck with mypy.

## 1. Black (Formatter)

**Install**:
```bash
pip install black
```

**Run**:
```bash
black src/ tests/
```

**pyproject.toml**:
```toml
[tool.black]
line-length = 88
target-version = ['py39', 'py310', 'py311', 'py312']
```

## 2. Ruff (Linter)

**Install**:
```bash
pip install ruff
```

**Run**:
```bash
ruff check src/
ruff format src/  # Also formats (alternative to black)
```

**pyproject.toml**:
```toml
[tool.ruff]
line-length = 88
select = ["E", "F", "I"]  # Error, Flake8, Isort
```

## 3. Mypy (Type Checker)

**Install**:
```bash
pip install mypy
```

**Run**:
```bash
mypy src/
```

**pyproject.toml**:
```toml
[tool.mypy]
python_version = "3.12"
warn_return_any = true
strict = true
```

## 4. Makefile for Convenience

```makefile
.PHONY: format lint typecheck test

format:
	black src/ tests/

lint:
	ruff check src/

typecheck:
	mypy src/

test:
	pytest

all: format lint typecheck test
```

Run:
```bash
make all
```
