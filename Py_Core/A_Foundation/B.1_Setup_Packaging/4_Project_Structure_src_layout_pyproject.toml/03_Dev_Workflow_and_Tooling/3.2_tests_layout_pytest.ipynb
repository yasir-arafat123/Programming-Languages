# Tests Layout & pytest

Professional test organization.

## 🎯 Learning Objectives
- Structure the tests/ directory.
- Write pytest tests.
- Run tests with coverage.

## 1. Test Directory Structure

```
myproject/
├── src/mypackage/
│   ├── __init__.py
│   ├── utils.py
│   └── api/
│       └── views.py
└── tests/
    ├── conftest.py        # Shared fixtures
    ├── test_utils.py
    └── api/
        └── test_views.py
```

**Mirror the src structure** in tests.

## 2. Basic pytest Test

```python
# tests/test_utils.py
from mypackage.utils import add

def test_add():
    assert add(2, 3) == 5

def test_add_negatives():
    assert add(-1, -1) == -2
```

Run:
```bash
pytest
```

## 3. Fixtures (conftest.py)

```python
# tests/conftest.py
import pytest

@pytest.fixture
def sample_data():
    return {"key": "value"}

# tests/test_something.py    
def test_with_fixture(sample_data):
    assert sample_data["key"] == "value"
```

## 4. Coverage Reporting

```bash
pytest --cov=mypackage --cov-report=html
```

Creates `htmlcov/index.html` showing which lines are tested.

## 5. pyproject.toml Configuration

```toml
[tool.pytest.ini_options]
testpaths = ["tests"]
pythonpath = "src"
addopts = "--cov=mypackage --cov-report=term-missing"
```
