# 🔀 if/elif/else - Conditional Logic

**"The program must decide."**

Conditional statements are the decision-making backbone of programming. Python's `if/elif/else` is simple on the surface but hides powerful idioms like EAFP (Easier to Ask Forgiveness than Permission) and guard clauses.

## 🎯 Learning Objectives

- Master the `if/elif/else` syntax and truthiness evaluation
- Learn Python idioms: EAFP vs LBYL (Look Before You Leap)
- Understand guard clauses for reducing nesting
- Explore the ternary operator for concise conditions
- Identify common anti-patterns and refactorings

## 1. Theory: Control Flow Graphs

Every `if` statement creates a **branch** in your program's execution path:

```
Start → [Condition?] → Yes → Block A → End
                    ↓
                    No → Block B → End
```

### Truthiness Review
Python evaluates conditions using "truthiness":
- **Falsy**: `False`, `None`, `0`, `0.0`, `""`, `[]`, `{}`, `set()`
- **Truthy**: Everything else

### Why This Matters
```python
# Instead of:
if len(my_list) > 0:

# Pythonic:
if my_list:  # Non-empty lists are truthy
```

## 2. Basic Syntax (Beginner Level)

````
age = 20

if age >= 18:
    print("Adult")
else:
    print("Minor")

# Multi-way branching
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"Grade: {grade}")
````

## 3. Intermediate: Python Idioms

### EAFP vs LBYL

**LBYL (Look Before You Leap)**: Check conditions first

**EAFP (Easier to Ask Forgiveness than Permission)**: Try first, handle exceptions

Python culture prefers **EAFP**.

````
# LBYL (Common in C/Java)
data = {"name": "Alice"}

if "age" in data:
    age = data["age"]
else:
    age = None

# EAFP (Pythonic)
try:
    age = data["age"]
except KeyError:
    age = None

# Even better (dict method)
age = data.get("age")  # Returns None if missing

print(f"Age: {age}")
````

### Why EAFP?
1. **Race conditions**: In multithreading, conditions can change between check and use
2. **Readability**: Shows the "happy path" code first
3. **Performance**: Exceptions are fast in Python when not raised frequently

## 4. Advanced: Guard Clauses

**Problem**: Deeply nested if-statements ("arrow code")

**Solution**: Early returns (guard clauses)

````
# BEFORE: Deep nesting (hard to read)
def process_user_bad(user):
    if user is not None:
        if user.is_active:
            if user.has_permission("edit"):
                # Main logic buried here
                return "Success"
            else:
                return "No permission"
        else:
            return "User not active"
    else:
        return "No user"

# AFTER: Guard clauses (easy to read)
def process_user_good(user):
    if user is None:
        return "No user"
    if not user.is_active:
        return "User not active"
    if not user.has_permission("edit"):
        return "No permission"
    
    # Main logic at top level (left-aligned)
    return "Success"
````

## 5. Ternary Operator (Conditional Expression)

````
# Syntax: value_if_true if condition else value_if_false

age = 15
status = "adult" if age >= 18 else "minor"
print(status)

# Useful for default values
user_name = None
display = user_name if user_name else "Guest"
# Even better: use 'or'
display = user_name or "Guest"

print(f"Welcome, {display}")
````

### Warning: Don't Abuse Ternary

````
# BAD: Nested ternary (unreadable)
grade = "A" if score >= 90 else "B" if score >= 80 else "C"

# GOOD: Use if/elif for multiple branches
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"
````

## 6. Common Pitfalls

### Pitfall 1: Comparing to True/False

````
is_valid = True

# BAD (verbose)
if is_valid == True:
    print("Valid")

# GOOD (idiomatic)
if is_valid:
    print("Valid")
````

### Pitfall 2: Using `is` for Value Comparison

````
count = 1000

# WRONG (checks identity, not value)
if count is 1000:  # Works accidentally for small ints!
    print("Oops")

# CORRECT
if count == 1000:
    print("Correct")

# EXCEPTION: Only use 'is' for None, True, False
if value is None:
    pass
````

### Pitfall 3: Empty Conditionals

````
# if something:
#     pass  # ❌ Why have this at all?

# Better: Invert the condition
# if not something:
#     do_alternative()
````

## 7. Real-World Application: Input Validation

````
def validate_email(email):
    """Validate email with guard clauses."""
    # Guard clauses handle edge cases first
    if not email:
        return False, "Email is required"
    
    if "@" not in email:
        return False, "Missing @ symbol"
    
    parts = email.split("@")
    if len(parts) != 2:
        return False, "Invalid format"
    
    local, domain = parts
    if not local or not domain:
        return False, "Empty local or domain"
    
    # Main logic (happy path)
    return True, "Valid"

# Test
is_valid, message = validate_email("user@example.com")
print(f"{is_valid}: {message}")
````

## 8. Summary

| Pattern | Use When |
| :--- | :--- |
| **if/elif/else** | Multiple distinct conditions |
| **Ternary** | Simple one-liner (max 1 line) |
| **Guard Clauses** | Validating inputs (early returns) |
| **EAF

P** | Handling optional data or exceptions |

### Best Practices
1. Use truthiness (`if x:` not `if len(x) > 0:`)
2. `is` only for `None`, `True`, `False`
3. Prefer guard clauses over deep nesting
4. Keep ternary simple (1 line max)
5. EAFP when exceptions are rare
