# 🔤 String Methods: split(), join(), strip()

**"Text processing is 80% of programming."**

String manipulation is one of the most common tasks in Python. Master these three methods—`split()`, `join()`, and `strip()`—and you'll handle 90% of text processing needs.

## 🎯 Learning Objectives

- Master the core string transformation methods
- Understand immutability and why strings return new copies
- Learn `split`/`join` as inverse operations
- Handle Unicode and whitespace edge cases
- Apply performance optimizations for large text processing

## 1. Theory: Strings Are Immutable

In Python, strings are **immutable**—you cannot change them in place. Every string method returns a **new string**.

### Why Immutability?
1. **Thread Safety**: Multiple threads can read the same string
2. **Hashability**: Strings can be dict keys
3. ***String Interning**: Python caches small strings in memory for efficiency

### Real-World Impact
```python
# WRONG (does nothing!)
text = "  hello  "
text.strip()  # Returns new string, but we ignore it
print(text)  # Still "  hello  "

# CORRECT
text = text.strip()
```

## 2. split(): Breaking Apart Strings

Converts a string into a list.

````
# Basic usage: split on whitespace (default)
text = "Python is awesome"
words = text.split()
print(words)  # ['Python', 'is', 'awesome']

# Split on custom delimiter
csv_line = "Alice,25,Engineer"
fields = csv_line.split(",")
print(fields)  # ['Alice', '25', 'Engineer']

# Limit splits (maxsplit parameter)
path = "/home/user/documents/file.txt"
parts = path.split("/", maxsplit=2)
print(parts)  # ['', 'home', 'user/documents/file.txt']
````

````
# Special behavior: split() vs split(" ")
messy = "  lots   of    spaces  "

# Default split() handles multiple whitespace
print(messy.split())     # ['lots', 'of', 'spaces']

# But split(' ') treats each space as delimiter
print(messy.split(' '))  # ['', '', 'lots', '', '', 'of', '', '', '', 'spaces', '', '']

# Lesson: Always use split() without args for general whitespace splitting
````

## 3. join(): Reassembling Strings

The inverse of `split()`. Takes a list and creates a string.

````
words = ["Python", "is", "awesome"]

# Join with space
sentence = " ".join(words)
print(sentence)  # "Python is awesome"

# Join with different delimiter
csv_line = ",".join(["Alice", "25", "Engineer"])
print(csv_line)  # "Alice,25,Engineer"

# Empty string join (concatenate)
chars = ['P', 'y', 't', 'h', 'o', 'n']
word = "".join(chars)
print(word)  # "Python"
````

### Performance: Why join() Instead of +=

**Anti-pattern** (slow for large lists):
```python
result = ""
for word in words:
    result += word + " "  # Creates N new strings!
```

**Pro pattern** (fast):
```python
result = " ".join(words)  # One allocation
```

Time complexity: `+=` is O(n²), `join()` is O(n)

````
# Benchmarking the difference
import time

words = ["word"] * 10000

# Slow way
start = time.time()
result = ""
for w in words:
    result += w + " "
slow_time = time.time() - start

# Fast way
start = time.time()
result = " ".join(words)
fast_time = time.time() - start

print(f"Slow (+=): {slow_time:.4f}s")
print(f"Fast (join): {fast_time:.4f}s")
print(f"Speedup: {slow_time / fast_time:.1f}x faster")
````

## 4. strip(): Removing Whitespace

Cleans up leading/trailing characters (whitespace by default).

````
# Basic strip
text = "   hello world   "
print(f"'{text.strip()}'")    # 'hello world'
print(f"'{text.lstrip()}'")   # 'hello world   ' (left only)
print(f"'{text.rstrip()}'")   # '   hello world' (right only)

# Strip custom characters
url = "https://example.com/"
clean = url.strip("https://").rstrip("/")
print(clean)  # "example.com"

# Warning: strip() removes ANY of the characters, not the exact sequence!
weird = "abcHELLOcba"
print(weird.strip("abc"))  # "HELLO" (removes a, b, c from both ends)
````

## 5. Advanced: Unicode and Edge Cases

````
# Non-ASCII whitespace
text = "\u00A0hello\u00A0"  # Non-breaking space (Unicode)
print(f"Length: {len(text)}")
print(f"After strip: '{text.strip()}'")
# strip() handles Unicode whitespace correctly

# Split on multi-char delimiters
text = "name::age::city"
fields = text.split("::")
print(fields)  # ['name', 'age', 'city']

# Empty string edge case
print("".split())  # []
print(",".join([]))  # ""
````

## 6. Common Pitfalls

### Pitfall 1: Forgetting to Reassign

````
name = "  Alice  "
name.strip()  # ❌ WRONG - does nothing
name = name.strip()  # ✅ CORRECT
````

### Pitfall 2: join() Order Confusion

The separator comes FIRST (method-call syntax is backwards from logic).

````
# Weird but correct:
result = ",".join(["a", "b", "c"])  # "a,b,c"

# Mental model: "separator JOIN items"
````

## 7. Real-World Application: Log Parsing

````
# Parse Apache-style log line
log_line = '192.168.1.1 - - [15/Dec/2024:10:45:22] "GET /index.html HTTP/1.1" 200 1234'

# Extract IP and URL
parts = log_line.split()
ip = parts[0]
url = parts[6]  # "/index.html"

print(f"Request from {ip} for {url}")

# Clean up URL path
clean_url = url.strip('"').strip("/")
print(f"Clean path: {clean_url}")
````

## 8. Best Practices

1. **Use `split()` without args** for whitespace splitting
2. **Use `join()` not `+=`** for concatenating multiple strings
3. **Always reassign** when using strip/replace/etc
4. **Be aware of Unicode** - Python 3 handles it correctly
5. **Consider `splitlines()`** for multi-line text instead of `split('\n')`
